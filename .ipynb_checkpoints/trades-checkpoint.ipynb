```python
//@version=6
indicator("Manny's Liquidity Strategy V9", overlay=true, max_lines_count=500, max_labels_count=500, max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS — Easy to understand and adjust
// ═══════════════════════════════════════════

// Levels
showMonthly  = input.bool(true,  "📅 Show Monthly High/Low Lines",        group="📊 Chart Levels")
showWeekly   = input.bool(true,  "📅 Show Weekly High/Low Lines",         group="📊 Chart Levels")
showDaily    = input.bool(true,  "📅 Show Daily High/Low Lines (3 Days)", group="📊 Chart Levels")

// Bias
showWeekBias = input.bool(true,  "📈 Show Weekly Bias (Mon 4AM/8AM GMT)", group="🧭 Bias Levels")
showDayBias  = input.bool(true,  "📈 Show Daily Bias (4AM GMT Candle)",   group="🧭 Bias Levels")

// Features
showEMA      = input.bool(true,  "📉 Show EMA 200 Trend Line",            group="🔧 Features")
showSweeps   = input.bool(true,  "💧 Show Liquidity Sweeps",              group="🔧 Features")
showFVG      = input.bool(true,  "📦 Show Fair Value Gaps (FVG)",         group="🔧 Features")
showDiv      = input.bool(true,  "📐 Show RSI Divergence",                group="🔧 Features")
showBOS      = input.bool(true,  "🔨 Show Break of Structure (BOS)",      group="🔧 Features")
showOB       = input.bool(true,  "🟦 Show Order Blocks",                  group="🔧 Features")
showSignals  = input.bool(true,  "🚦 Show Buy/Sell Signals",              group="🔧 Features")

// Signal Quality
minScore     = input.int(8,      "Minimum Score to Show Signal (out of 12)", group="🎯 Signal Quality", minval=1, maxval=12, tooltip="8 = Good Trade, 9 = Strong Trade, 10+ = A+ Trade. We recommend keeping this at 8 or above.")

// Risk Management
accountSize  = input.float(1000, "💰 Account Size ($)",                   group="💵 Risk Management", tooltip="Enter your total account balance")
riskPercent  = input.float(1.0,  "⚠️ Risk Per Trade (%)",                group="💵 Risk Management", tooltip="1% means you risk 1% of your account per trade. Recommended: 1-2%")
atrLength    = input.int(14,     "ATR Period (Default 14)",               group="💵 Risk Management", tooltip="Used to calculate stop loss and take profit distances. Leave at 14 unless you know what you're doing.")

// ═══════════════════════════════════════════
// TIME HELPERS (GMT)
// ═══════════════════════════════════════════
currentHour = hour(time, "GMT+0")
currentMin  = minute(time, "GMT+0")
currentDOW  = dayofweek(time, "GMT+0")
isMonday    = currentDOW == 2
is4AMCandle = currentHour == 4 and currentMin == 0
is8AMCandle = currentHour == 8 and currentMin == 0
isNewDay    = ta.change(time("D")) != 0
isNewWeek   = ta.change(time("W")) != 0
isNewMonth  = ta.change(time("M")) != 0

// ═══════════════════════════════════════════
// ATR + EMA 200
// ═══════════════════════════════════════════
atr         = ta.atr(atrLength)
ema200      = ta.ema(close, 200)
bullishBias = close > ema200
bearishBias = close < ema200
plot(showEMA ? ema200 : na, "EMA 200 — Trend Direction", color=bullishBias ? color.new(color.green, 0) : color.new(color.red, 0), linewidth=2)

// ═══════════════════════════════════════════
// MONTHLY LEVELS — Purple
// ═══════════════════════════════════════════
month1High = ta.valuewhen(isNewMonth, high[1], 0)
month1Low  = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, month1High, bar_index + 100, month1High, color=color.new(color.purple, 0), width=2, style=line.style_solid)
    mLL := line.new(bar_index, month1Low,  bar_index + 100, month1Low,  color=color.new(color.purple, 0), width=2, style=line.style_solid)
    label.new(bar_index, month1High, "📅 Monthly High", color=color.new(color.purple, 70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, month1Low,  "📅 Monthly Low",  color=color.new(color.purple, 70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)

// Monthly bias — is price heading toward monthly high or low
monthBullAlign = close > month1Low  and bullishBias
monthBearAlign = close < month1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY LEVELS — Blue
// ═══════════════════════════════════════════
week1High = ta.valuewhen(isNewWeek, high[1], 0)
week1Low  = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, week1High, bar_index + 100, week1High, color=color.new(color.blue, 0), width=2, style=line.style_solid)
    wLL := line.new(bar_index, week1Low,  bar_index + 100, week1Low,  color=color.new(color.blue, 0), width=2, style=line.style_solid)
    label.new(bar_index, week1High, "📅 Weekly High", color=color.new(color.blue, 70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, week1Low,  "📅 Weekly Low",  color=color.new(color.blue, 70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

// Weekly bias
weekBullish = close > week1Low  and bullishBias
weekBearish = close < week1High and bearishBias

// ═══════════════════════════════════════════
// DAILY LEVELS — Black (3 days fading)
// ═══════════════════════════════════════════
day1High = ta.valuewhen(isNewDay, high[1], 0)
day1Low  = ta.valuewhen(isNewDay, low[1],  0)
day2High = ta.valuewhen(isNewDay, high[1], 1)
day2Low  = ta.valuewhen(isNewDay, low[1],  1)
day3High = ta.valuewhen(isNewDay, high[1], 2)
day3Low  = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, day1High, bar_index + 100, day1High, color=color.new(color.black, 0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, day1Low,  bar_index + 100, day1Low,  color=color.new(color.black, 0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, day2High, bar_index + 100, day2High, color=color.new(color.black, 40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, day2Low,  bar_index + 100, day2Low,  color=color.new(color.black, 40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, day3High, bar_index + 100, day3High, color=color.new(color.black, 70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, day3Low,  bar_index + 100, day3Low,  color=color.new(color.black, 70), width=1, style=line.style_dotted)
    label.new(bar_index, day1High, "D1H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day1Low,  "D1L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2High, "D2H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2Low,  "D2L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3High, "D3H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3Low,  "D3L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

// Daily bias
dayBullish = close > day1Low  and bullishBias
dayBearish = close < day1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY BIAS — Monday 4AM + 8AM (Gold)
// ═══════════════════════════════════════════
var float weekBiasHigh = na
var float weekBiasLow  = na
var line  wbHL         = na
var line  wbLL         = na

if isMonday and is4AMCandle and showWeekBias
    weekBiasHigh := high
    weekBiasLow  := low

if isMonday and is8AMCandle and showWeekBias
    weekBiasHigh := math.max(weekBiasHigh, high)
    weekBiasLow  := math.min(weekBiasLow,  low)
    line.delete(wbHL)
    line.delete(wbLL)
    wbHL := line.new(bar_index, weekBiasHigh, bar_index + 100, weekBiasHigh, color=color.new(color.yellow, 0), width=2, style=line.style_solid)
    wbLL := line.new(bar_index, weekBiasLow,  bar_index + 100, weekBiasLow,  color=color.new(color.yellow, 0), width=2, style=line.style_solid)
    label.new(bar_index, weekBiasHigh, "⚡ Weekly Bias High", color=color.new(color.yellow, 70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, weekBiasLow,  "⚡ Weekly Bias Low",  color=color.new(color.yellow, 70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY BIAS — 4AM Candle (Green)
// ═══════════════════════════════════════════
var float dayBiasHigh = na
var float dayBiasLow  = na
var line  dbHL        = na
var line  dbLL        = na

if is4AMCandle and showDayBias
    dayBiasHigh := high
    dayBiasLow  := low
    line.delete(dbHL)
    line.delete(dbLL)
    dbHL := line.new(bar_index, dayBiasHigh, bar_index + 100, dayBiasHigh, color=color.new(color.green, 0), width=1, style=line.style_dashed)
    dbLL := line.new(bar_index, dayBiasLow,  bar_index + 100, dayBiasLow,  color=color.new(color.green, 0), width=1, style=line.style_dashed)
    label.new(bar_index, dayBiasHigh, "🕓 4AM High", color=color.new(color.green, 70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, dayBiasLow,  "🕓 4AM Low",  color=color.new(color.green, 70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// Price wicks beyond previous day H/L but closes back
// ═══════════════════════════════════════════
bullSweepWick = day1Low - low
bearSweepWick = high - day1High
bullishSweep  = low < day1Low   and close > day1Low  and bullSweepWick >= atr * 0.3
bearishSweep  = high > day1High and close < day1High and bearSweepWick >= atr * 0.3

var float lastBullSweepLow  = na
var float lastBearSweepHigh = na
if bullishSweep
    lastBullSweepLow  := low
if bearishSweep
    lastBearSweepHigh := high

plotshape(showSweeps and bullishSweep, "💧 Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime, 0), size=size.small, text="Sweep")
plotshape(showSweeps and bearishSweep, "💧 Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,  0), size=size.small, text="Sweep")

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE (BOS)
// High quality — confirmed close + strong candle
// ═══════════════════════════════════════════
swingLen    = 5
swingHigh   = ta.highest(high, swingLen * 2 + 1)[swingLen]
swingLow    = ta.lowest(low,   swingLen * 2 + 1)[swingLen]

// Strong candle body filter
strongBOSCandle = math.abs(close - open) > atr * 0.7

// BOS confirmed = closes beyond swing + strong candle + bias aligned
bullBOS = close > swingHigh and strongBOSCandle and bullishBias and close[1] <= swingHigh
bearBOS = close < swingLow  and strongBOSCandle and bearishBias and close[1] >= swingLow

// Track last BOS for score
var bool recentBullBOS = false
var bool recentBearBOS = false
var int  bosTimer      = 0

if bullBOS
    recentBullBOS := true
    recentBearBOS := false
    bosTimer      := bar_index

if bearBOS
    recentBearBOS := true
    recentBullBOS := false
    bosTimer      := bar_index

// Reset BOS after 10 bars
if bar_index - bosTimer > 10
    recentBullBOS := false
    recentBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low  - atr * 0.5, "🔨 BOS ↑", color=color.new(color.green, 20), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index - 1, swingHigh, bar_index + 10, swingHigh, color=color.new(color.green, 40), width=1, style=line.style_dotted)

if showBOS and bearBOS
    label.new(bar_index, high + atr * 0.5, "🔨 BOS ↓", color=color.new(color.red,   20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index - 1, swingLow,  bar_index + 10, swingLow,  color=color.new(color.red,   40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// ORDER BLOCKS (OB)
// Last opposing candle before strong impulsive move
// ═══════════════════════════════════════════
impulseLen  = 3
bullImpulse = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImpulse = close < close[1] and close[1] < close[2] and close[2] < close[3]

// Bullish OB = last bearish candle before bullish impulse
// Bearish OB = last bullish candle before bearish impulse
bullOBCandle = bullImpulse and close[impulseLen] < open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5
bearOBCandle = bearImpulse and close[impulseLen] > open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5

var float bullOBTop   = na
var float bullOBBot   = na
var float bearOBTop   = na
var float bearOBBot   = na
var bool  bullOBAlive = false
var bool  bearOBAlive = false
var box   bullOBBox   = na
var box   bearOBBox   = na
var bool  bullOBRespected = false
var bool  bearOBRespected = false
var int   bullOBTouches   = 0
var int   bearOBTouches   = 0

// Create new OB zones
if bullOBCandle and bullishBias and showOB
    bullOBTop       := high[impulseLen]
    bullOBBot       := low[impulseLen]
    bullOBAlive     := true
    bullOBRespected := false
    bullOBTouches   := 0
    box.delete(bullOBBox)
    bullOBBox := box.new(bar_index - impulseLen, bullOBTop, bar_index + 100, bullOBBot, bgcolor=color.new(color.blue, 80), border_color=color.new(color.blue, 30), border_width=1)
    label.new(bar_index - impulseLen, bullOBTop, "🟦 Bull OB", color=color.new(color.blue, 60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBCandle and bearishBias and showOB
    bearOBTop       := high[impulseLen]
    bearOBBot       := low[impulseLen]
    bearOBAlive     := true
    bearOBRespected := false
    bearOBTouches   := 0
    box.delete(bearOBBox)
    bearOBBox := box.new(bar_index - impulseLen, bearOBTop, bar_index + 100, bearOBBot, bgcolor=color.new(color.orange, 80), border_color=color.new(color.orange, 30), border_width=1)
    label.new(bar_index - impulseLen, bearOBBot, "🟧 Bear OB", color=color.new(color.orange, 60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

// Check if price returns to OB zone — first touch only
if bullOBAlive and bullOBTouches == 0
    if low <= bullOBTop and close >= bullOBBot
        bullOBRespected := true
        bullOBTouches   := bullOBTouches + 1
        box.set_bgcolor(bullOBBox, color.new(color.green, 70))
        box.set_border_color(bullOBBox, color.new(color.green, 20))
        label.new(bar_index, bullOBBot, "✅ OB Respected", color=color.new(color.green, 30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearOBAlive and bearOBTouches == 0
    if high >= bearOBBot and close <= bearOBTop
        bearOBRespected := true
        bearOBTouches   := bearOBTouches + 1
        box.set_bgcolor(bearOBBox, color.new(color.red, 70))
        box.set_border_color(bearOBBox, color.new(color.red, 20))
        label.new(bar_index, bearOBTop, "✅ OB Respected", color=color.new(color.red, 30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// Invalidate OB if price closes through it
if bullOBAlive and close < bullOBBot
    bullOBAlive     := false
    bullOBRespected := false
    box.set_bgcolor(bullOBBox, color.new(color.gray, 90))

if bearOBAlive and close > bearOBTop
    bearOBAlive     := false
    bearOBRespected := false
    box.set_bgcolor(bearOBBox, color.new(color.gray, 90))

// ═══════════════════════════════════════════
// HIGH QUALITY FVG
// Strong candle + bias aligned + stays visible
// ═══════════════════════════════════════════
candleBody   = math.abs(close[1] - open[1])
strongCandle = candleBody > atr * 1.0

bullFVGraw = low > high[2] and strongCandle and close[1] > open[1] and bullishBias
bearFVGraw = high < low[2] and strongCandle and close[1] < open[1] and bearishBias

var float bullFVGTop      = na
var float bullFVGBot      = na
var float bearFVGTop      = na
var float bearFVGBot      = na
var bool  bullFVGAlive    = false
var bool  bearFVGAlive    = false
var box   bullFVGBox      = na
var box   bearFVGBox      = na
var bool  bullFVGRespected = false
var bool  bearFVGRespected = false

if bullFVGraw and showFVG
    bullFVGTop      := low
    bullFVGBot      := high[2]
    bullFVGAlive    := true
    bullFVGRespected := false
    box.delete(bullFVGBox)
    bullFVGBox := box.new(bar_index - 2, bullFVGTop, bar_index + 100, bullFVGBot, bgcolor=color.new(color.green, 85), border_color=color.new(color.green, 40), border_width=1)
    label.new(bar_index, bullFVGBot, "📦 FVG", color=color.new(color.green, 50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGraw and showFVG
    bearFVGTop      := low[2]
    bearFVGBot      := high
    bearFVGAlive    := true
    bearFVGRespected := false
    box.delete(bearFVGBox)
    bearFVGBox := box.new(bar_index - 2, bearFVGTop, bar_index + 100, bearFVGBot, bgcolor=color.new(color.red, 85), border_color=color.new(color.red, 40), border_width=1)
    label.new(bar_index, bearFVGTop, "📦 FVG", color=color.new(color.red, 50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// FVG respected = price touched AND closed back in direction
if bullFVGAlive and low <= bullFVGTop and close >= bullFVGTop
    bullFVGRespected := true
    box.set_bgcolor(bullFVGBox, color.new(color.green, 70))
    box.set_border_color(bullFVGBox, color.new(color.green, 20))
    label.new(bar_index, bullFVGBot, "✅ FVG Respected", color=color.new(color.green, 30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGAlive and high >= bearFVGBot and close <= bearFVGBot
    bearFVGRespected := true
    box.set_bgcolor(bearFVGBox, color.new(color.red, 70))
    box.set_border_color(bearFVGBox, color.new(color.red, 20))
    label.new(bar_index, bearFVGTop, "✅ FVG Respected", color=color.new(color.red, 30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// Invalidate FVG if price closes through it
if bullFVGAlive and close < bullFVGBot
    bullFVGAlive    := false
    bullFVGRespected := false
    box.set_bgcolor(bullFVGBox, color.new(color.gray, 90))

if bearFVGAlive and close > bearFVGTop
    bearFVGAlive    := false
    bearFVGRespected := false
    box.set_bgcolor(bearFVGBox, color.new(color.gray, 90))

// ═══════════════════════════════════════════
// RSI DIVERGENCE — Strong + Hidden only
// ═══════════════════════════════════════════
rsi = ta.rsi(close, 14)
lkb = 7

priceLL = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
priceHL = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
priceHH = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
priceLH = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)

rsiHL = ta.lowest(rsi,  lkb) > ta.lowest(rsi[lkb],  lkb)
rsiLL = ta.lowest(rsi,  lkb) < ta.lowest(rsi[lkb],  lkb)
rsiLH = ta.highest(rsi, lkb) < ta.highest(rsi[lkb], lkb)
rsiHH = ta.highest(rsi, lkb) > ta.highest(rsi[lkb], lkb)

rsiBullExtreme = rsi < 40
rsiBearExtreme = rsi > 60

strongBullDiv = priceLL and rsiHL
hiddenBullDiv = priceHL and rsiLL
strongBearDiv = priceHH and rsiLH
hiddenBearDiv = priceLH and rsiHH

anyBullDiv = strongBullDiv or hiddenBullDiv
anyBearDiv = strongBearDiv or hiddenBearDiv

if showDiv and strongBullDiv
    label.new(bar_index, low  - atr * 0.3, "◆ Strong Bull Div", color=color.new(color.green,  0), textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and hiddenBullDiv
    label.new(bar_index, low  - atr * 0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,   0), textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and strongBearDiv
    label.new(bar_index, high + atr * 0.3, "◆ Strong Bear Div", color=color.new(color.red,    0), textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hiddenBearDiv
    label.new(bar_index, high + atr * 0.3, "◇ Hidden Bear Div", color=color.new(color.orange, 0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdLine, signalLine, histLine] = ta.macd(close, 12, 26, 9)
macdBullish  = macdLine > signalLine
macdBearish  = macdLine < signalLine
macdMomentum = histLine > histLine[1]

// ═══════════════════════════════════════════
// CONFLUENCE SCORE — 12 POINTS
// ═══════════════════════════════════════════

// LONG SCORE
longScore = 0
longScore := longScore + (bullishBias                            ? 1 : 0)  // 1.  EMA trend
longScore := longScore + (monthBullAlign                         ? 1 : 0)  // 2.  Monthly bias
longScore := longScore + (weekBullish                            ? 1 : 0)  // 3.  Weekly bias
longScore := longScore + (dayBullish                             ? 1 : 0)  // 4.  Daily bias
longScore := longScore + (bullishSweep                           ? 1 : 0)  // 5.  Liquidity sweep
longScore := longScore + (recentBullBOS                          ? 1 : 0)  // 6.  Break of structure
longScore := longScore + (bullOBRespected                        ? 1 : 0)  // 7.  Order block respected
longScore := longScore + (bullFVGRespected                       ? 1 : 0)  // 8.  FVG respected
longScore := longScore + (strongBullDiv ? 2 : hiddenBullDiv ? 1 : 0)      // 9-10. Divergence
longScore := longScore + (macdBullish                            ? 1 : 0)  // 11. MACD
longScore := longScore + (rsiBullExtreme                         ? 1 : 0)  // 12. RSI extreme

// SHORT SCORE
shortScore = 0
shortScore := shortScore + (bearishBias                          ? 1 : 0)
shortScore := shortScore + (monthBearAlign                       ? 1 : 0)
shortScore := shortScore + (weekBearish                          ? 1 : 0)
shortScore := shortScore + (dayBearish                           ? 1 : 0)
shortScore := shortScore + (bearishSweep                         ? 1 : 0)
shortScore := shortScore + (recentBearBOS                        ? 1 : 0)
shortScore := shortScore + (bearOBRespected                      ? 1 : 0)
shortScore := shortScore + (bearFVGRespected                     ? 1 : 0)
shortScore := shortScore + (strongBearDiv ? 2 : hiddenBearDiv ? 1 : 0)
shortScore := shortScore + (macdBearish                          ? 1 : 0)
shortScore := shortScore + (rsiBearExtreme                       ? 1 : 0)

// ═══════════════════════════════════════════
// CONFIDENCE LABEL
// ═══════════════════════════════════════════
confidenceLabel(score) =>
    score >= 11 ? "🔥 A+ PERFECT"  :
     score >= 9  ? "💪 STRONG"      :
     score >= 8  ? "✅ GOOD TRADE"  :
                   "⏳ WAIT"

confidenceColor(score) =>
    score >= 11 ? color.new(color.yellow, 0) :
     score >= 9  ? color.new(color.green,  0) :
     score >= 8  ? color.new(color.teal,   0) :
                   color.new(color.gray,   0)

// ═══════════════════════════════════════════
// RISK MANAGEMENT — 1:3 only
// ═══════════════════════════════════════════
riskAmount = accountSize * (riskPercent / 100)
slBuffer   = longScore >= 9 or shortScore >= 9 ? atr * 0.2 : atr * 0.5

longSL    = not na(lastBullSweepLow)  ? lastBullSweepLow  - slBuffer : close - atr * 1.5
longRisk  = math.max(close - longSL,  0.0001)
longTP    = close + longRisk * 3
longSize  = riskAmount / longRisk

shortSL   = not na(lastBearSweepHigh) ? lastBearSweepHigh + slBuffer : close + atr * 1.5
shortRisk = math.max(shortSL - close, 0.0001)
shortTP   = close - shortRisk * 3
shortSize = riskAmount / shortRisk

// ═══════════════════════════════════════════
// ENTRY SIGNAL LABELS — full breakdown
// ═══════════════════════════════════════════
bullishEntry = longScore  >= minScore
bearishEntry = shortScore >= minScore

longConditions =
     "⬆ BUY SIGNAL — " + str.tostring(longScore) + "/12\n" +
     confidenceLabel(longScore) + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CORE CONDITIONS:\n" +
     (bullishBias      ? "✅ EMA 200: Bullish Trend\n"        : "❌ EMA 200: Not Bullish\n")       +
     (weekBullish      ? "✅ Weekly Bias: Confirmed\n"         : "❌ Weekly Bias: Not Confirmed\n") +
     (dayBullish       ? "✅ Daily Bias: Confirmed\n"          : "❌ Daily Bias: Not Confirmed\n")  +
     (recentBullBOS    ? "✅ Break of Structure: Yes\n"        : "❌ Break of Structure: No\n")     +
     (bullishSweep     ? "✅ Liquidity Sweep: Yes\n"           : "❌ Liquidity Sweep: No\n")        +
     (anyBullDiv       ? "✅ RSI Divergence: " + (strongBullDiv ? "Strong\n" : "Hidden\n") : "❌ RSI Divergence: None\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CONFLUENCE:\n" +
     (monthBullAlign   ? "✅ Monthly Bias: Aligned\n"          : "➖ Monthly Bias: Not Set\n")      +
     (bullOBRespected  ? "✅ Order Block: Respected\n"         : "➖ Order Block: Not Hit\n")       +
     (bullFVGRespected ? "✅ FVG: Respected\n"                 : "➖ FVG: Not Respected\n")         +
     (macdBullish      ? "✅ MACD: Bullish\n"                  : "➖ MACD: Not Bullish\n")          +
     (rsiBullExtreme   ? "✅ RSI Oversold: " + str.tostring(math.round(rsi, 1)) + "\n" : "➖ RSI: " + str.tostring(math.round(rsi, 1)) + "\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRADE LEVELS:\n" +
     "🔴 Stop Loss:  $" + str.tostring(math.round(longSL,         2)) + "\n" +
     "🟢 Take Profit: $" + str.tostring(math.round(longTP,        2)) + "\n" +
     "💰 Risk:       $" + str.tostring(math.round(riskAmount,     2)) + "\n" +
     "🎯 Target:     $" + str.tostring(math.round(riskAmount * 3, 2)) + "\n" +
     "📊 Size:        " + str.tostring(math.round(longSize,       4)) + " units"

shortConditions =
     "⬇ SELL SIGNAL — " + str.tostring(shortScore) + "/12\n" +
     confidenceLabel(shortScore) + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CORE CONDITIONS:\n" +
     (bearishBias      ? "✅ EMA 200: Bearish Trend\n"         : "❌ EMA 200: Not Bearish\n")      +
     (weekBearish      ? "✅ Weekly Bias: Confirmed\n"          : "❌ Weekly Bias: Not Confirmed\n") +
     (dayBearish       ? "✅ Daily Bias: Confirmed\n"           : "❌ Daily Bias: Not Confirmed\n") +
     (recentBearBOS    ? "✅ Break of Structure: Yes\n"         : "❌ Break of Structure: No\n")    +
     (bearishSweep     ? "✅ Liquidity Sweep: Yes\n"            : "❌ Liquidity Sweep: No\n")       +
     (anyBearDiv       ? "✅ RSI Divergence: " + (strongBearDiv ? "Strong\n" : "Hidden\n") : "❌ RSI Divergence: None\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CONFLUENCE:\n" +
     (monthBearAlign   ? "✅ Monthly Bias: Aligned\n"           : "➖ Monthly Bias: Not Set\n")     +
     (bearOBRespected  ? "✅ Order Block: Respected\n"          : "➖ Order Block: Not Hit\n")      +
     (bearFVGRespected ? "✅ FVG: Respected\n"                  : "➖ FVG: Not Respected\n")        +
     (macdBearish      ? "✅ MACD: Bearish\n"                   : "➖ MACD: Not Bearish\n")         +
     (rsiBearExtreme   ? "✅ RSI Overbought: " + str.tostring(math.round(rsi, 1)) + "\n" : "➖ RSI: " + str.tostring(math.round(rsi, 1)) + "\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRADE LEVELS:\n" +
     "🔴 Stop Loss:   $" + str.tostring(math.round(shortSL,        2)) + "\n" +
     "🟢 Take Profit: $" + str.tostring(math.round(shortTP,        2)) + "\n" +
     "💰 Risk:        $" + str.tostring(math.round(riskAmount,     2)) + "\n" +
     "🎯 Target:      $" + str.tostring(math.round(riskAmount * 3, 2)) + "\n" +
     "📊 Size:         " + str.tostring(math.round(shortSize,      4)) + " units"

if showSignals and bullishEntry
    label.new(bar_index, low, longConditions, color=color.new(color.green, 10), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index, longSL,  bar_index + 40, longSL,  color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP,  bar_index + 40, longTP,  color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index + 40, longSL,  "🔴 SL", color=color.new(color.red,   0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index + 40, longTP,  "🟢 TP", color=color.new(color.green, 0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearishEntry
    label.new(bar_index, high, shortConditions, color=color.new(color.red, 10), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index + 40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index + 40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index + 40, shortSL, "🔴 SL", color=color.new(color.red,   0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index + 40, shortTP, "🟢 TP", color=color.new(color.green, 0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DASHBOARD — Clean and user friendly
// ═══════════════════════════════════════════
var table dash = table.new(position.top_right, 2, 16, bgcolor=color.new(color.black, 60), border_color=color.new(color.gray, 50), border_width=1)

// Header
table.cell(dash, 0, 0,  "📊 CONDITION",      text_color=color.white, text_size=size.small, bgcolor=color.new(color.navy,  20))
table.cell(dash, 1, 0,  "STATUS",            text_color=color.white, text_size=size.small, bgcolor=color.new(color.navy,  20))

// Core conditions
table.cell(dash, 0, 1,  "EMA 200 Trend",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 1,  bullishBias ? "🟢 BULL" : "🔴 BEAR",          text_color=bullishBias ? color.green : color.red,    text_size=size.small)
table.cell(dash, 0, 2,  "Monthly Bias",      text_color=color.white, text_size=size.small)
table.cell(dash, 1, 2,  monthBullAlign ? "🟢 BULL" : monthBearAlign ? "🔴 BEAR" : "⏳ WAIT", text_color=monthBullAlign ? color.green : monthBearAlign ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 0, 3,  "Weekly Bias",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 3,  weekBullish ? "🟢 BULL" : weekBearish ? "🔴 BEAR" : "⏳ WAIT",        text_color=weekBullish ? color.green : weekBearish ? color.red : color.yellow,       text_size=size.small)
table.cell(dash, 0, 4,  "Daily Bias",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 4,  dayBullish ? "🟢 BULL" : dayBearish ? "🔴 BEAR" : "⏳ WAIT",          text_color=dayBullish ? color.green : dayBearish ? color.red : color.yellow,         text_size=size.small)
table.cell(dash, 0, 5,  "Break of Structure",text_color=color.white, text_size=size.small)
table.cell(dash, 1, 5,  recentBullBOS ? "🟢 BULL BOS" : recentBearBOS ? "🔴 BEAR BOS" : "⏳ NONE", text_color=recentBullBOS ? color.green : recentBearBOS ? color.red : color.gray, text_size=size.small)
table.cell(dash, 0, 6,  "Liquidity Sweep",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 6,  bullishSweep ? "🟢 BULL" : bearishSweep ? "🔴 BEAR" : "⏳ NONE",      text_color=bullishSweep or bearishSweep ? color.green : color.gray,                  text_size=size.small)
table.cell(dash, 0, 7,  "RSI Divergence",    text_color=color.white, text_size=size.small)
table.cell(dash, 1, 7,  strongBullDiv ? "🟢 Strong Bull" : hiddenBullDiv ? "🟢 Hidden Bull" : strongBearDiv ? "🔴 Strong Bear" : hiddenBearDiv ? "🔴 Hidden Bear" : "⏳ NONE",     text_color=anyBullDiv ? color.green : anyBearDiv ? color.red : color.gray, text_size=size.small)

// Confluence conditions
table.cell(dash, 0, 8,  "── CONFLUENCE ──",  text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 8,  "──────────────",    text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 9,  "Order Block",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 9,  bullOBRespected ? "✅ BULL RESP" : bearOBRespected ? "✅ BEAR RESP" : bullOBAlive ? "🟦 BULL LIVE" : bearOBAlive ? "🟧 BEAR LIVE" : "⏳ NONE",             text_color=bullOBRespected or bearOBRespected ? color.green : bullOBAlive or bearOBAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 0, 10, "Fair Value Gap",    text_color=color.white, text_size=size.small)
table.cell(dash, 1, 10, bullFVGRespected ? "✅ BULL RESP" : bearFVGRespected ? "✅ BEAR RESP" : bullFVGAlive ? "📦 BULL LIVE" : bearFVGAlive ? "📦 BEAR LIVE" : "⏳ NONE",         text_color=bullFVGRespected or bearFVGRespected ? color.green : bullFVGAlive or bearFVGAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 0, 11, "MACD",             text_color=color.white, text_size=size.small)
table.cell(dash, 1, 11, macdBullish ? "🟢 BULL" : "🔴 BEAR",          text_color=macdBullish ? color.green : color.red,    text_size=size.small)
table.cell(dash, 0, 12, "RSI Value",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 12, str.tostring(math.round(rsi, 1)) + (rsiBullExtreme ? " 🟢 Oversold" : rsiBearExtreme ? " 🔴 Overbought" : " ⚪ Neutral"), text_color=rsiBullExtreme ? color.green : rsiBearExtreme ? color.red : color.white, text_size=size.small)

// Risk info
table.cell(dash, 0, 13, "── RISK INFO ──",   text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 13, "──────────────",    text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 14, "Risk / Target",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 14, "💰 $" + str.tostring(math.round(riskAmount, 2)) + " → 🎯 $" + str.tostring(math.round(riskAmount * 3, 2)), text_color=color.yellow, text_size=size.small)

// Signal
table.cell(dash, 0, 15, "🚦 SIGNAL",         text_color=color.white, text_size=size.small, bgcolor=color.new(color.gray, 40))
table.cell(dash, 1, 15,
     bullishEntry ? "⬆ BUY " + str.tostring(longScore)  + "/12 " + confidenceLabel(longScore)  :
     bearishEntry ? "⬇ SELL " + str.tostring(shortScore) + "/12 " + confidenceLabel(shortScore) :
     "⏳ WAIT — " + str.tostring(math.max(longScore, shortScore)) + "/12",
     text_color=bullishEntry ? color.green : bearishEntry ? color.red : color.yellow,
     text_size=size.small)
```

```python
//@version=6
strategy("Manny's Liquidity Strategy V9 — Backtest", overlay=true, max_labels_count=500, max_lines_count=500, max_boxes_count=500, initial_capital=1000, default_qty_type=strategy.percent_of_equity, default_qty_value=100, commission_type=strategy.commission.percent, commission_value=0.1, slippage=2)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
showMonthly  = input.bool(true,  "📅 Show Monthly High/Low Lines",        group="📊 Chart Levels")
showWeekly   = input.bool(true,  "📅 Show Weekly High/Low Lines",         group="📊 Chart Levels")
showDaily    = input.bool(true,  "📅 Show Daily High/Low Lines (3 Days)", group="📊 Chart Levels")
showWeekBias = input.bool(true,  "📈 Show Weekly Bias (Mon 4AM/8AM GMT)", group="🧭 Bias Levels")
showDayBias  = input.bool(true,  "📈 Show Daily Bias (4AM GMT Candle)",   group="🧭 Bias Levels")
showEMA      = input.bool(true,  "📉 Show EMA 200 Trend Line",            group="🔧 Features")
showSweeps   = input.bool(true,  "💧 Show Liquidity Sweeps",              group="🔧 Features")
showFVG      = input.bool(true,  "📦 Show Fair Value Gaps",               group="🔧 Features")
showDiv      = input.bool(true,  "📐 Show RSI Divergence",                group="🔧 Features")
showBOS      = input.bool(true,  "🔨 Show Break of Structure",            group="🔧 Features")
showOB       = input.bool(true,  "🟦 Show Order Blocks",                  group="🔧 Features")
showSignals  = input.bool(true,  "🚦 Show Buy/Sell Signals",              group="🔧 Features")
minScore     = input.int(8,      "Minimum Score to Signal (out of 12)",   group="🎯 Signal Quality", minval=1, maxval=12, tooltip="8 = Good Trade, 9 = Strong, 10+ = A+. Recommend keeping at 8 or above.")
accountSize  = input.float(1000, "💰 Account Size ($)",                   group="💵 Risk Management")
riskPercent  = input.float(1.0,  "⚠️ Risk Per Trade (%)",                group="💵 Risk Management")
atrLength    = input.int(14,     "ATR Period",                            group="💵 Risk Management")

// ═══════════════════════════════════════════
// TIME HELPERS (GMT)
// ═══════════════════════════════════════════
currentHour = hour(time, "GMT+0")
currentMin  = minute(time, "GMT+0")
currentDOW  = dayofweek(time, "GMT+0")
isMonday    = currentDOW == 2
is4AMCandle = currentHour == 4 and currentMin == 0
is8AMCandle = currentHour == 8 and currentMin == 0
isNewDay    = ta.change(time("D")) != 0
isNewWeek   = ta.change(time("W")) != 0
isNewMonth  = ta.change(time("M")) != 0

// ═══════════════════════════════════════════
// ATR + EMA 200
// ═══════════════════════════════════════════
atr         = ta.atr(atrLength)
ema200      = ta.ema(close, 200)
bullishBias = close > ema200
bearishBias = close < ema200
plot(showEMA ? ema200 : na, "EMA 200", color=bullishBias ? color.new(color.green, 0) : color.new(color.red, 0), linewidth=2)

// ═══════════════════════════════════════════
// MONTHLY LEVELS
// ═══════════════════════════════════════════
month1High = ta.valuewhen(isNewMonth, high[1], 0)
month1Low  = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, month1High, bar_index + 100, month1High, color=color.new(color.purple, 0), width=2, style=line.style_solid)
    mLL := line.new(bar_index, month1Low,  bar_index + 100, month1Low,  color=color.new(color.purple, 0), width=2, style=line.style_solid)
    label.new(bar_index, month1High, "📅 Monthly High", color=color.new(color.purple, 70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, month1Low,  "📅 Monthly Low",  color=color.new(color.purple, 70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)

monthBullAlign = close > month1Low  and bullishBias
monthBearAlign = close < month1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY LEVELS
// ═══════════════════════════════════════════
week1High = ta.valuewhen(isNewWeek, high[1], 0)
week1Low  = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, week1High, bar_index + 100, week1High, color=color.new(color.blue, 0), width=2, style=line.style_solid)
    wLL := line.new(bar_index, week1Low,  bar_index + 100, week1Low,  color=color.new(color.blue, 0), width=2, style=line.style_solid)
    label.new(bar_index, week1High, "📅 Weekly High", color=color.new(color.blue, 70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, week1Low,  "📅 Weekly Low",  color=color.new(color.blue, 70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

weekBullish = close > week1Low  and bullishBias
weekBearish = close < week1High and bearishBias

// ═══════════════════════════════════════════
// DAILY LEVELS
// ═══════════════════════════════════════════
day1High = ta.valuewhen(isNewDay, high[1], 0)
day1Low  = ta.valuewhen(isNewDay, low[1],  0)
day2High = ta.valuewhen(isNewDay, high[1], 1)
day2Low  = ta.valuewhen(isNewDay, low[1],  1)
day3High = ta.valuewhen(isNewDay, high[1], 2)
day3Low  = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, day1High, bar_index + 100, day1High, color=color.new(color.black, 0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, day1Low,  bar_index + 100, day1Low,  color=color.new(color.black, 0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, day2High, bar_index + 100, day2High, color=color.new(color.black, 40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, day2Low,  bar_index + 100, day2Low,  color=color.new(color.black, 40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, day3High, bar_index + 100, day3High, color=color.new(color.black, 70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, day3Low,  bar_index + 100, day3Low,  color=color.new(color.black, 70), width=1, style=line.style_dotted)
    label.new(bar_index, day1High, "D1H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day1Low,  "D1L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2High, "D2H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2Low,  "D2L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3High, "D3H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3Low,  "D3L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

dayBullish = close > day1Low  and bullishBias
dayBearish = close < day1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY BIAS
// ═══════════════════════════════════════════
var float weekBiasHigh = na
var float weekBiasLow  = na
var line  wbHL         = na
var line  wbLL         = na

if isMonday and is4AMCandle and showWeekBias
    weekBiasHigh := high
    weekBiasLow  := low

if isMonday and is8AMCandle and showWeekBias
    weekBiasHigh := math.max(weekBiasHigh, high)
    weekBiasLow  := math.min(weekBiasLow,  low)
    line.delete(wbHL)
    line.delete(wbLL)
    wbHL := line.new(bar_index, weekBiasHigh, bar_index + 100, weekBiasHigh, color=color.new(color.yellow, 0), width=2, style=line.style_solid)
    wbLL := line.new(bar_index, weekBiasLow,  bar_index + 100, weekBiasLow,  color=color.new(color.yellow, 0), width=2, style=line.style_solid)
    label.new(bar_index, weekBiasHigh, "⚡ Weekly Bias High", color=color.new(color.yellow, 70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, weekBiasLow,  "⚡ Weekly Bias Low",  color=color.new(color.yellow, 70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY BIAS
// ═══════════════════════════════════════════
var float dayBiasHigh = na
var float dayBiasLow  = na
var line  dbHL        = na
var line  dbLL        = na

if is4AMCandle and showDayBias
    dayBiasHigh := high
    dayBiasLow  := low
    line.delete(dbHL)
    line.delete(dbLL)
    dbHL := line.new(bar_index, dayBiasHigh, bar_index + 100, dayBiasHigh, color=color.new(color.green, 0), width=1, style=line.style_dashed)
    dbLL := line.new(bar_index, dayBiasLow,  bar_index + 100, dayBiasLow,  color=color.new(color.green, 0), width=1, style=line.style_dashed)
    label.new(bar_index, dayBiasHigh, "🕓 4AM High", color=color.new(color.green, 70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, dayBiasLow,  "🕓 4AM Low",  color=color.new(color.green, 70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweepWick = day1Low - low
bearSweepWick = high - day1High
bullishSweep  = low < day1Low   and close > day1Low  and bullSweepWick >= atr * 0.3
bearishSweep  = high > day1High and close < day1High and bearSweepWick >= atr * 0.3

var float lastBullSweepLow  = na
var float lastBearSweepHigh = na
if bullishSweep
    lastBullSweepLow  := low
if bearishSweep
    lastBearSweepHigh := high

plotshape(showSweeps and bullishSweep, "💧 Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime, 0), size=size.small, text="Sweep")
plotshape(showSweeps and bearishSweep, "💧 Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,  0), size=size.small, text="Sweep")

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swingLen        = 5
swingHigh       = ta.highest(high, swingLen * 2 + 1)[swingLen]
swingLow        = ta.lowest(low,   swingLen * 2 + 1)[swingLen]
strongBOSCandle = math.abs(close - open) > atr * 0.7
bullBOS         = close > swingHigh and strongBOSCandle and bullishBias and close[1] <= swingHigh
bearBOS         = close < swingLow  and strongBOSCandle and bearishBias and close[1] >= swingLow

var bool recentBullBOS = false
var bool recentBearBOS = false
var int  bosTimer      = 0

if bullBOS
    recentBullBOS := true
    recentBearBOS := false
    bosTimer      := bar_index
if bearBOS
    recentBearBOS := true
    recentBullBOS := false
    bosTimer      := bar_index
if bar_index - bosTimer > 10
    recentBullBOS := false
    recentBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low  - atr * 0.5, "🔨 BOS ↑", color=color.new(color.green, 20), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index - 1, swingHigh, bar_index + 10, swingHigh, color=color.new(color.green, 40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr * 0.5, "🔨 BOS ↓", color=color.new(color.red,   20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index - 1, swingLow,  bar_index + 10, swingLow,  color=color.new(color.red,   40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
impulseLen  = 3
bullImpulse = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImpulse = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBCandle = bullImpulse and close[impulseLen] < open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5
bearOBCandle = bearImpulse and close[impulseLen] > open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5

var float bullOBTop       = na
var float bullOBBot       = na
var float bearOBTop       = na
var float bearOBBot       = na
var bool  bullOBAlive     = false
var bool  bearOBAlive     = false
var box   bullOBBox       = na
var box   bearOBBox       = na
var bool  bullOBRespected = false
var bool  bearOBRespected = false
var int   bullOBTouches   = 0
var int   bearOBTouches   = 0

if bullOBCandle and bullishBias and showOB
    bullOBTop       := high[impulseLen]
    bullOBBot       := low[impulseLen]
    bullOBAlive     := true
    bullOBRespected := false
    bullOBTouches   := 0
    box.delete(bullOBBox)
    bullOBBox := box.new(bar_index - impulseLen, bullOBTop, bar_index + 100, bullOBBot, bgcolor=color.new(color.blue, 80), border_color=color.new(color.blue, 30), border_width=1)
    label.new(bar_index - impulseLen, bullOBTop, "🟦 Bull OB", color=color.new(color.blue, 60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBCandle and bearishBias and showOB
    bearOBTop       := high[impulseLen]
    bearOBBot       := low[impulseLen]
    bearOBAlive     := true
    bearOBRespected := false
    bearOBTouches   := 0
    box.delete(bearOBBox)
    bearOBBox := box.new(bar_index - impulseLen, bearOBTop, bar_index + 100, bearOBBot, bgcolor=color.new(color.orange, 80), border_color=color.new(color.orange, 30), border_width=1)
    label.new(bar_index - impulseLen, bearOBBot, "🟧 Bear OB", color=color.new(color.orange, 60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bullOBAlive and bullOBTouches == 0
    if low <= bullOBTop and close >= bullOBBot
        bullOBRespected := true
        bullOBTouches   := bullOBTouches + 1
        box.set_bgcolor(bullOBBox, color.new(color.green, 70))
        box.set_border_color(bullOBBox, color.new(color.green, 20))
        label.new(bar_index, bullOBBot, "✅ OB Respected", color=color.new(color.green, 30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearOBAlive and bearOBTouches == 0
    if high >= bearOBBot and close <= bearOBTop
        bearOBRespected := true
        bearOBTouches   := bearOBTouches + 1
        box.set_bgcolor(bearOBBox, color.new(color.red, 70))
        box.set_border_color(bearOBBox, color.new(color.red, 20))
        label.new(bar_index, bearOBTop, "✅ OB Respected", color=color.new(color.red, 30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullOBAlive and close < bullOBBot
    bullOBAlive     := false
    bullOBRespected := false
    box.set_bgcolor(bullOBBox, color.new(color.gray, 90))

if bearOBAlive and close > bearOBTop
    bearOBAlive     := false
    bearOBRespected := false
    box.set_bgcolor(bearOBBox, color.new(color.gray, 90))

// ═══════════════════════════════════════════
// FAIR VALUE GAPS
// ═══════════════════════════════════════════
candleBody    = math.abs(close[1] - open[1])
strongCandle  = candleBody > atr * 1.0
bullFVGraw    = low > high[2] and strongCandle and close[1] > open[1] and bullishBias
bearFVGraw    = high < low[2] and strongCandle and close[1] < open[1] and bearishBias

var float bullFVGTop      = na
var float bullFVGBot      = na
var float bearFVGTop      = na
var float bearFVGBot      = na
var bool  bullFVGAlive    = false
var bool  bearFVGAlive    = false
var box   bullFVGBox      = na
var box   bearFVGBox      = na
var bool  bullFVGRespected = false
var bool  bearFVGRespected = false

if bullFVGraw and showFVG
    bullFVGTop       := low
    bullFVGBot       := high[2]
    bullFVGAlive     := true
    bullFVGRespected := false
    box.delete(bullFVGBox)
    bullFVGBox := box.new(bar_index - 2, bullFVGTop, bar_index + 100, bullFVGBot, bgcolor=color.new(color.green, 85), border_color=color.new(color.green, 40), border_width=1)
    label.new(bar_index, bullFVGBot, "📦 FVG", color=color.new(color.green, 50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGraw and showFVG
    bearFVGTop       := low[2]
    bearFVGBot       := high
    bearFVGAlive     := true
    bearFVGRespected := false
    box.delete(bearFVGBox)
    bearFVGBox := box.new(bar_index - 2, bearFVGTop, bar_index + 100, bearFVGBot, bgcolor=color.new(color.red, 85), border_color=color.new(color.red, 40), border_width=1)
    label.new(bar_index, bearFVGTop, "📦 FVG", color=color.new(color.red, 50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullFVGAlive and low <= bullFVGTop and close >= bullFVGTop
    bullFVGRespected := true
    box.set_bgcolor(bullFVGBox, color.new(color.green, 70))
    box.set_border_color(bullFVGBox, color.new(color.green, 20))

if bearFVGAlive and high >= bearFVGBot and close <= bearFVGBot
    bearFVGRespected := true
    box.set_bgcolor(bearFVGBox, color.new(color.red, 70))
    box.set_border_color(bearFVGBox, color.new(color.red, 20))

if bullFVGAlive and close < bullFVGBot
    bullFVGAlive     := false
    bullFVGRespected := false
    box.set_bgcolor(bullFVGBox, color.new(color.gray, 90))

if bearFVGAlive and close > bearFVGTop
    bearFVGAlive     := false
    bearFVGRespected := false
    box.set_bgcolor(bearFVGBox, color.new(color.gray, 90))

// ═══════════════════════════════════════════
// RSI DIVERGENCE
// ═══════════════════════════════════════════
rsi = ta.rsi(close, 14)
lkb = 7

priceLL = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
priceHL = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
priceHH = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
priceLH = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)

rsiHL = ta.lowest(rsi,  lkb) > ta.lowest(rsi[lkb],  lkb)
rsiLL = ta.lowest(rsi,  lkb) < ta.lowest(rsi[lkb],  lkb)
rsiLH = ta.highest(rsi, lkb) < ta.highest(rsi[lkb], lkb)
rsiHH = ta.highest(rsi, lkb) > ta.highest(rsi[lkb], lkb)

rsiBullExtreme = rsi < 40
rsiBearExtreme = rsi > 60

strongBullDiv = priceLL and rsiHL
hiddenBullDiv = priceHL and rsiLL
strongBearDiv = priceHH and rsiLH
hiddenBearDiv = priceLH and rsiHH

anyBullDiv = strongBullDiv or hiddenBullDiv
anyBearDiv = strongBearDiv or hiddenBearDiv

if showDiv and strongBullDiv
    label.new(bar_index, low  - atr * 0.3, "◆ Strong Bull Div", color=color.new(color.green,  0), textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and hiddenBullDiv
    label.new(bar_index, low  - atr * 0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,   0), textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and strongBearDiv
    label.new(bar_index, high + atr * 0.3, "◆ Strong Bear Div", color=color.new(color.red,    0), textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hiddenBearDiv
    label.new(bar_index, high + atr * 0.3, "◇ Hidden Bear Div", color=color.new(color.orange, 0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdLine, signalLine, histLine] = ta.macd(close, 12, 26, 9)
macdBullish  = macdLine > signalLine
macdBearish  = macdLine < signalLine
macdMomentum = histLine > histLine[1]

// ═══════════════════════════════════════════
// CONFLUENCE SCORE — 12 POINTS
// ═══════════════════════════════════════════
longScore = 0
longScore := longScore + (bullishBias                            ? 1 : 0)
longScore := longScore + (monthBullAlign                         ? 1 : 0)
longScore := longScore + (weekBullish                            ? 1 : 0)
longScore := longScore + (dayBullish                             ? 1 : 0)
longScore := longScore + (bullishSweep                           ? 1 : 0)
longScore := longScore + (recentBullBOS                          ? 1 : 0)
longScore := longScore + (bullOBRespected                        ? 1 : 0)
longScore := longScore + (bullFVGRespected                       ? 1 : 0)
longScore := longScore + (strongBullDiv ? 2 : hiddenBullDiv ? 1 : 0)
longScore := longScore + (macdBullish                            ? 1 : 0)
longScore := longScore + (rsiBullExtreme                         ? 1 : 0)

shortScore = 0
shortScore := shortScore + (bearishBias                          ? 1 : 0)
shortScore := shortScore + (monthBearAlign                       ? 1 : 0)
shortScore := shortScore + (weekBearish                          ? 1 : 0)
shortScore := shortScore + (dayBearish                           ? 1 : 0)
shortScore := shortScore + (bearishSweep                         ? 1 : 0)
shortScore := shortScore + (recentBearBOS                        ? 1 : 0)
shortScore := shortScore + (bearOBRespected                      ? 1 : 0)
shortScore := shortScore + (bearFVGRespected                     ? 1 : 0)
shortScore := shortScore + (strongBearDiv ? 2 : hiddenBearDiv ? 1 : 0)
shortScore := shortScore + (macdBearish                          ? 1 : 0)
shortScore := shortScore + (rsiBearExtreme                       ? 1 : 0)

// ═══════════════════════════════════════════
// CONFIDENCE LABEL
// ═══════════════════════════════════════════
confidenceLabel(score) =>
    score >= 11 ? "🔥 A+ PERFECT" :
     score >= 9  ? "💪 STRONG"     :
     score >= 8  ? "✅ GOOD TRADE" :
                   "⏳ WAIT"

// ═══════════════════════════════════════════
// RISK MANAGEMENT
// ═══════════════════════════════════════════
riskAmount = accountSize * (riskPercent / 100)
slBuffer   = longScore >= 9 or shortScore >= 9 ? atr * 0.2 : atr * 0.5

longSL    = not na(lastBullSweepLow)  ? lastBullSweepLow  - slBuffer : close - atr * 1.5
longRisk  = math.max(close - longSL,  0.0001)
longTP    = close + longRisk * 3
longSize  = riskAmount / longRisk

shortSL   = not na(lastBearSweepHigh) ? lastBearSweepHigh + slBuffer : close + atr * 1.5
shortRisk = math.max(shortSL - close, 0.0001)
shortTP   = close - shortRisk * 3
shortSize = riskAmount / shortRisk

// ═══════════════════════════════════════════
// ENTRY + EXIT ORDERS — Strategy Tester
// ═══════════════════════════════════════════
bullishEntry = longScore  >= minScore
bearishEntry = shortScore >= minScore

// Long entry
if bullishEntry and strategy.position_size == 0
    strategy.entry("BUY", strategy.long)
    strategy.exit("BUY Exit", from_entry="BUY", stop=longSL, limit=longTP)

// Short entry
if bearishEntry and strategy.position_size == 0
    strategy.entry("SELL", strategy.short)
    strategy.exit("SELL Exit", from_entry="SELL", stop=shortSL, limit=shortTP)

// ═══════════════════════════════════════════
// SIGNAL LABELS
// ═══════════════════════════════════════════
longConditions =
     "⬆ BUY — " + str.tostring(longScore) + "/12 — " + confidenceLabel(longScore) + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CORE CONDITIONS:\n" +
     (bullishBias      ? "✅ EMA 200: Bullish\n"          : "❌ EMA 200: Not Bullish\n")       +
     (weekBullish      ? "✅ Weekly Bias: Confirmed\n"     : "❌ Weekly Bias: Not Set\n")        +
     (dayBullish       ? "✅ Daily Bias: Confirmed\n"      : "❌ Daily Bias: Not Set\n")         +
     (recentBullBOS    ? "✅ Break of Structure: Yes\n"    : "❌ Break of Structure: No\n")      +
     (bullishSweep     ? "✅ Liquidity Sweep: Yes\n"       : "❌ Liquidity Sweep: No\n")         +
     (anyBullDiv       ? "✅ RSI Div: " + (strongBullDiv ? "Strong\n" : "Hidden\n") : "❌ RSI Div: None\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CONFLUENCE:\n" +
     (monthBullAlign   ? "✅ Monthly: Aligned\n"           : "➖ Monthly: Not Set\n")            +
     (bullOBRespected  ? "✅ Order Block: Respected\n"     : "➖ Order Block: Not Hit\n")        +
     (bullFVGRespected ? "✅ FVG: Respected\n"             : "➖ FVG: Not Respected\n")          +
     (macdBullish      ? "✅ MACD: Bullish\n"              : "➖ MACD: Not Bullish\n")           +
     (rsiBullExtreme   ? "✅ RSI Oversold: " + str.tostring(math.round(rsi, 1)) + "\n" : "➖ RSI: " + str.tostring(math.round(rsi, 1)) + "\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL:  $" + str.tostring(math.round(longSL,         2)) + "\n" +
     "🟢 TP:  $" + str.tostring(math.round(longTP,         2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmount,    2)) + "\n" +
     "🎯 Target: $" + str.tostring(math.round(riskAmount * 3, 2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(longSize,       4)) + " units"

shortConditions =
     "⬇ SELL — " + str.tostring(shortScore) + "/12 — " + confidenceLabel(shortScore) + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CORE CONDITIONS:\n" +
     (bearishBias      ? "✅ EMA 200: Bearish\n"           : "❌ EMA 200: Not Bearish\n")       +
     (weekBearish      ? "✅ Weekly Bias: Confirmed\n"      : "❌ Weekly Bias: Not Set\n")       +
     (dayBearish       ? "✅ Daily Bias: Confirmed\n"       : "❌ Daily Bias: Not Set\n")        +
     (recentBearBOS    ? "✅ Break of Structure: Yes\n"     : "❌ Break of Structure: No\n")     +
     (bearishSweep     ? "✅ Liquidity Sweep: Yes\n"        : "❌ Liquidity Sweep: No\n")        +
     (anyBearDiv       ? "✅ RSI Div: " + (strongBearDiv ? "Strong\n" : "Hidden\n") : "❌ RSI Div: None\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CONFLUENCE:\n" +
     (monthBearAlign   ? "✅ Monthly: Aligned\n"            : "➖ Monthly: Not Set\n")           +
     (bearOBRespected  ? "✅ Order Block: Respected\n"      : "➖ Order Block: Not Hit\n")       +
     (bearFVGRespected ? "✅ FVG: Respected\n"              : "➖ FVG: Not Respected\n")         +
     (macdBearish      ? "✅ MACD: Bearish\n"               : "➖ MACD: Not Bearish\n")          +
     (rsiBearExtreme   ? "✅ RSI Overbought: " + str.tostring(math.round(rsi, 1)) + "\n" : "➖ RSI: " + str.tostring(math.round(rsi, 1)) + "\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL:  $" + str.tostring(math.round(shortSL,        2)) + "\n" +
     "🟢 TP:  $" + str.tostring(math.round(shortTP,        2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmount,    2)) + "\n" +
     "🎯 Target: $" + str.tostring(math.round(riskAmount * 3, 2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(shortSize,      4)) + " units"

if showSignals and bullishEntry
    label.new(bar_index, low,  longConditions,  color=color.new(color.green, 10), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index, longSL,  bar_index + 40, longSL,  color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP,  bar_index + 40, longTP,  color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index + 40, longSL,  "🔴 SL", color=color.new(color.red,   0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index + 40, longTP,  "🟢 TP", color=color.new(color.green, 0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearishEntry
    label.new(bar_index, high, shortConditions, color=color.new(color.red,   10), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index + 40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index + 40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index + 40, shortSL, "🔴 SL", color=color.new(color.red,   0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index + 40, shortTP, "🟢 TP", color=color.new(color.green, 0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DASHBOARD
// ═══════════════════════════════════════════
var table dash = table.new(position.top_right, 2, 16, bgcolor=color.new(color.black, 60), border_color=color.new(color.gray, 50), border_width=1)

table.cell(dash, 0, 0,  "📊 CONDITION",       text_color=color.white, text_size=size.small, bgcolor=color.new(color.gray, 30))
table.cell(dash, 1, 0,  "STATUS",             text_color=color.white, text_size=size.small, bgcolor=color.new(color.gray, 30))
table.cell(dash, 0, 1,  "EMA 200 Trend",      text_color=color.white, text_size=size.small)
table.cell(dash, 1, 1,  bullishBias ? "🟢 BULL" : "🔴 BEAR",           text_color=bullishBias ? color.green : color.red,    text_size=size.small)
table.cell(dash, 0, 2,  "Monthly Bias",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 2,  monthBullAlign ? "🟢 BULL" : monthBearAlign ? "🔴 BEAR" : "⏳ WAIT", text_color=monthBullAlign ? color.green : monthBearAlign ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 0, 3,  "Weekly Bias",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 3,  weekBullish ? "🟢 BULL" : weekBearish ? "🔴 BEAR" : "⏳ WAIT",        text_color=weekBullish ? color.green : weekBearish ? color.red : color.yellow,       text_size=size.small)
table.cell(dash, 0, 4,  "Daily Bias",         text_color=color.white, text_size=size.small)
table.cell(dash, 1, 4,  dayBullish ? "🟢 BULL" : dayBearish ? "🔴 BEAR" : "⏳ WAIT",          text_color=dayBullish ? color.green : dayBearish ? color.red : color.yellow,         text_size=size.small)
table.cell(dash, 0, 5,  "Break of Structure", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 5,  recentBullBOS ? "🟢 BULL BOS" : recentBearBOS ? "🔴 BEAR BOS" : "⏳ NONE", text_color=recentBullBOS ? color.green : recentBearBOS ? color.red : color.gray, text_size=size.small)
table.cell(dash, 0, 6,  "Liquidity Sweep",    text_color=color.white, text_size=size.small)
table.cell(dash, 1, 6,  bullishSweep ? "🟢 BULL" : bearishSweep ? "🔴 BEAR" : "⏳ NONE",      text_color=bullishSweep or bearishSweep ? color.green : color.gray,                  text_size=size.small)
table.cell(dash, 0, 7,  "RSI Divergence",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 7,  strongBullDiv ? "🟢 Strong Bull" : hiddenBullDiv ? "🟢 Hidden Bull" : strongBearDiv ? "🔴 Strong Bear" : hiddenBearDiv ? "🔴 Hidden Bear" : "⏳ NONE",     text_color=anyBullDiv ? color.green : anyBearDiv ? color.red : color.gray, text_size=size.small)
table.cell(dash, 0, 8,  "── CONFLUENCE ──",   text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 8,  "──────────────",     text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 9,  "Order Block",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 9,  bullOBRespected ? "✅ BULL RESP" : bearOBRespected ? "✅ BEAR RESP" : bullOBAlive ? "🟦 BULL LIVE" : bearOBAlive ? "🟧 BEAR LIVE" : "⏳ NONE",             text_color=bullOBRespected or bearOBRespected ? color.green : bullOBAlive or bearOBAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 0, 10, "Fair Value Gap",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 10, bullFVGRespected ? "✅ BULL RESP" : bearFVGRespected ? "✅ BEAR RESP" : bullFVGAlive ? "📦 BULL LIVE" : bearFVGAlive ? "📦 BEAR LIVE" : "⏳ NONE",         text_color=bullFVGRespected or bearFVGRespected ? color.green : bullFVGAlive or bearFVGAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 0, 11, "MACD",              text_color=color.white, text_size=size.small)
table.cell(dash, 1, 11, macdBullish ? "🟢 BULL" : "🔴 BEAR",           text_color=macdBullish ? color.green : color.red,    text_size=size.small)
table.cell(dash, 0, 12, "RSI Value",         text_color=color.white, text_size=size.small)
table.cell(dash, 1, 12, str.tostring(math.round(rsi, 1)) + (rsiBullExtreme ? " 🟢 Oversold" : rsiBearExtreme ? " 🔴 Overbought" : " ⚪ Neutral"), text_color=rsiBullExtreme ? color.green : rsiBearExtreme ? color.red : color.white, text_size=size.small)
table.cell(dash, 0, 13, "── RISK ──",         text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 13, "──────────────",     text_color=color.gray,  text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 14, "Risk / Target",      text_color=color.white, text_size=size.small)
table.cell(dash, 1, 14, "💰 $" + str.tostring(math.round(riskAmount, 2)) + " → 🎯 $" + str.tostring(math.round(riskAmount * 3, 2)), text_color=color.yellow, text_size=size.small)
table.cell(dash, 0, 15, "🚦 SIGNAL",          text_color=color.white, text_size=size.small, bgcolor=color.new(color.gray, 30))
table.cell(dash, 1, 15,
     bullishEntry ? "⬆ BUY "  + str.tostring(longScore)  + "/12 " + confidenceLabel(longScore)  :
     bearishEntry ? "⬇ SELL " + str.tostring(shortScore) + "/12 " + confidenceLabel(shortScore) :
     "⏳ WAIT — " + str.tostring(math.max(longScore, shortScore)) + "/12",
     text_color=bullishEntry ? color.green : bearishEntry ? color.red : color.yellow,
     text_size=size.small)
```

```python
//@version=6
indicator("Manny's Liquidity Strategy V10", overlay=true, max_lines_count=500, max_labels_count=500, max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
showMonthly  = input.bool(true,  "📅 Show Monthly High/Low Lines",        group="📊 Chart Levels")
showWeekly   = input.bool(true,  "📅 Show Weekly High/Low Lines",         group="📊 Chart Levels")
showDaily    = input.bool(true,  "📅 Show Daily High/Low Lines (3 Days)", group="📊 Chart Levels")
showWeekBias = input.bool(true,  "📈 Show Weekly Bias (Mon 4AM/8AM GMT)", group="🧭 Bias Levels")
showDayBias  = input.bool(true,  "📈 Show Daily Bias (4AM GMT Candle)",   group="🧭 Bias Levels")
showEMA      = input.bool(true,  "📉 Show EMA 200 Trend Line",            group="🔧 Features")
showSweeps   = input.bool(true,  "💧 Show Liquidity Sweeps",              group="🔧 Features")
showFVG      = input.bool(true,  "📦 Show Fair Value Gaps (FVG)",         group="🔧 Features")
showDiv      = input.bool(true,  "📐 Show RSI Divergence",                group="🔧 Features")
showBOS      = input.bool(true,  "🔨 Show Break of Structure (BOS)",      group="🔧 Features")
showOB       = input.bool(true,  "🟦 Show Order Blocks",                  group="🔧 Features")
showSignals  = input.bool(true,  "🚦 Show Buy/Sell Signals",              group="🔧 Features")
showSession  = input.bool(true,  "🕐 Show Active Session",                group="🔧 Features")

// Signal Quality
minScore     = input.int(8,      "Minimum Score to Signal (out of 15)",   group="🎯 Signal Quality", minval=1, maxval=15, tooltip="8 = Good, 10 = Strong, 13+ = A+. Recommend 8 or above.")

// ADX Settings
adxLen       = input.int(14,     "ADX Period",                            group="📊 Market Filter", tooltip="ADX measures trend strength. Above 25 = trending market. Below 25 = ranging — no trades.")
adxThreshold = input.int(25,     "ADX Minimum (Trend Strength)",          group="📊 Market Filter", tooltip="Keep at 25. Lower = more signals but lower quality. Higher = fewer but stronger signals.")

// Risk Management
accountSize  = input.float(1000, "💰 Account Size ($)",                   group="💵 Risk Management", tooltip="Enter your total account balance")
riskPercent  = input.float(1.0,  "⚠️ Risk Per Trade (%)",                group="💵 Risk Management", tooltip="1% means you risk 1% of your account per trade. Recommended: 1-2%")
atrLength    = input.int(14,     "ATR Period (Default 14)",               group="💵 Risk Management")

// ═══════════════════════════════════════════
// TIME HELPERS (GMT)
// ═══════════════════════════════════════════
currentHour = hour(time, "GMT+0")
currentMin  = minute(time, "GMT+0")
currentDOW  = dayofweek(time, "GMT+0")
isMonday    = currentDOW == 2
is4AMCandle = currentHour == 4 and currentMin == 0
is8AMCandle = currentHour == 8 and currentMin == 0
isNewDay    = ta.change(time("D")) != 0
isNewWeek   = ta.change(time("W")) != 0
isNewMonth  = ta.change(time("M")) != 0

// ═══════════════════════════════════════════
// SESSION FILTER — London + NY only
// ═══════════════════════════════════════════
londonOpen  = currentHour >= 7  and currentHour < 12
nyOpen      = currentHour >= 13 and currentHour < 17
overlap     = currentHour >= 12 and currentHour < 13
activeSession = londonOpen or nyOpen or overlap

sessionName =
     overlap     ? "⚡ London/NY Overlap — BEST" :
     londonOpen  ? "🇬🇧 London Session — ACTIVE" :
     nyOpen      ? "🇺🇸 New York Session — ACTIVE" :
                   "😴 Asian Session — NO TRADES"

sessionColor =
     overlap    ? color.yellow :
     londonOpen ? color.green  :
     nyOpen     ? color.blue   :
                  color.gray

// ═══════════════════════════════════════════
// ATR + EMA 200 + EMA 50
// ═══════════════════════════════════════════
atr         = ta.atr(atrLength)
ema200      = ta.ema(close, 200)
ema50       = ta.ema(close, 50)
bullishBias = close > ema200
bearishBias = close < ema200

// Bear market protection
// Bull market  = EMA 50 above EMA 200
// Bear market  = EMA 50 below EMA 200 — sells only
// Extreme bear = price more than 20% below EMA 200 — no trades
pctBelowEMA200  = (ema200 - close) / ema200 * 100
extremeBear     = pctBelowEMA200 > 20
bullMarket      = ema50 > ema200
bearMarket      = ema50 < ema200
marketRegime    = extremeBear ? "🚨 EXTREME BEAR — NO TRADES" : bullMarket ? "🟢 BULL MARKET" : "🔴 BEAR MARKET"
allowLongs      = bullMarket and not extremeBear
allowShorts     = not extremeBear

plot(showEMA ? ema200 : na, "EMA 200", color=bullishBias ? color.new(color.green, 0) : color.new(color.red, 0), linewidth=2)
plot(showEMA ? ema50  : na, "EMA 50",  color=bullMarket  ? color.new(color.green, 50) : color.new(color.red, 50), linewidth=1)

// ═══════════════════════════════════════════
// ADX — TRENDING MARKET FILTER
// ═══════════════════════════════════════════
[diPlus, diMinus, adx] = ta.dmi(adxLen, adxLen)
isTrending  = adx >= adxThreshold
adxStatus   = isTrending ? "📈 TRENDING (" + str.tostring(math.round(adx, 1)) + ")" : "📉 RANGING (" + str.tostring(math.round(adx, 1)) + ") — WAIT"
adxColor    = isTrending ? color.green : color.red

// ═══════════════════════════════════════════
// HIGHER TIMEFRAME CONFIRMATION
// Automatically checks HTF trend
// ═══════════════════════════════════════════
// Daily trend
htfDailyClose  = request.security(syminfo.tickerid, "D",  close)
htfDailyEMA200 = request.security(syminfo.tickerid, "D",  ta.ema(close, 200))
htfDailyBull   = htfDailyClose > htfDailyEMA200
htfDailyBear   = htfDailyClose < htfDailyEMA200

// Weekly trend
htfWeeklyClose  = request.security(syminfo.tickerid, "W",  close)
htfWeeklyEMA200 = request.security(syminfo.tickerid, "W",  ta.ema(close, 200))
htfWeeklyBull   = htfWeeklyClose > htfWeeklyEMA200
htfWeeklyBear   = htfWeeklyClose < htfWeeklyEMA200

// 4H trend
htf4HClose  = request.security(syminfo.tickerid, "240", close)
htf4HEMA200 = request.security(syminfo.tickerid, "240", ta.ema(close, 200))
htf4HBull   = htf4HClose > htf4HEMA200
htf4HBear   = htf4HClose < htf4HEMA200

// 1H trend
htf1HClose  = request.security(syminfo.tickerid, "60",  close)
htf1HEMA200 = request.security(syminfo.tickerid, "60",  ta.ema(close, 200))
htf1HBull   = htf1HClose > htf1HEMA200
htf1HBear   = htf1HClose < htf1HEMA200

// HTF confirmation based on current timeframe
// 5M  needs 1H + 4H + Daily all bullish/bearish
// 1H  needs 4H + Daily all bullish/bearish
// 4H  needs Daily + Weekly all bullish/bearish
// Daily+ just uses Weekly
tfMins = timeframe.in_seconds() / 60

htfBullConfirm =
     tfMins <= 5   ? (htf1HBull and htf4HBull and htfDailyBull) :
     tfMins <= 60  ? (htf4HBull and htfDailyBull)               :
     tfMins <= 240 ? (htfDailyBull and htfWeeklyBull)           :
                      htfWeeklyBull

htfBearConfirm =
     tfMins <= 5   ? (htf1HBear and htf4HBear and htfDailyBear) :
     tfMins <= 60  ? (htf4HBear and htfDailyBear)               :
     tfMins <= 240 ? (htfDailyBear and htfWeeklyBear)           :
                      htfWeeklyBear

htfStatus =
     htfBullConfirm ? "🟢 HTF: ALL BULLISH ✓" :
     htfBearConfirm ? "🔴 HTF: ALL BEARISH ✓" :
                      "⏳ HTF: NOT ALIGNED"

// ═══════════════════════════════════════════
// MONTHLY LEVELS — Purple
// ═══════════════════════════════════════════
month1High = ta.valuewhen(isNewMonth, high[1], 0)
month1Low  = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, month1High, bar_index + 100, month1High, color=color.new(color.purple, 0), width=2, style=line.style_solid)
    mLL := line.new(bar_index, month1Low,  bar_index + 100, month1Low,  color=color.new(color.purple, 0), width=2, style=line.style_solid)
    label.new(bar_index, month1High, "📅 Monthly High", color=color.new(color.purple, 70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, month1Low,  "📅 Monthly Low",  color=color.new(color.purple, 70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)

monthBullAlign = close > month1Low  and bullishBias
monthBearAlign = close < month1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY LEVELS — Blue
// ═══════════════════════════════════════════
week1High = ta.valuewhen(isNewWeek, high[1], 0)
week1Low  = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, week1High, bar_index + 100, week1High, color=color.new(color.blue, 0), width=2, style=line.style_solid)
    wLL := line.new(bar_index, week1Low,  bar_index + 100, week1Low,  color=color.new(color.blue, 0), width=2, style=line.style_solid)
    label.new(bar_index, week1High, "📅 Weekly High", color=color.new(color.blue, 70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, week1Low,  "📅 Weekly Low",  color=color.new(color.blue, 70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

weekBullish = close > week1Low  and bullishBias
weekBearish = close < week1High and bearishBias

// ═══════════════════════════════════════════
// DAILY LEVELS — Black (3 days fading)
// ═══════════════════════════════════════════
day1High = ta.valuewhen(isNewDay, high[1], 0)
day1Low  = ta.valuewhen(isNewDay, low[1],  0)
day2High = ta.valuewhen(isNewDay, high[1], 1)
day2Low  = ta.valuewhen(isNewDay, low[1],  1)
day3High = ta.valuewhen(isNewDay, high[1], 2)
day3Low  = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, day1High, bar_index + 100, day1High, color=color.new(color.black, 0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, day1Low,  bar_index + 100, day1Low,  color=color.new(color.black, 0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, day2High, bar_index + 100, day2High, color=color.new(color.black, 40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, day2Low,  bar_index + 100, day2Low,  color=color.new(color.black, 40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, day3High, bar_index + 100, day3High, color=color.new(color.black, 70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, day3Low,  bar_index + 100, day3Low,  color=color.new(color.black, 70), width=1, style=line.style_dotted)
    label.new(bar_index, day1High, "D1H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day1Low,  "D1L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2High, "D2H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2Low,  "D2L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3High, "D3H", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3Low,  "D3L", color=color.new(color.gray, 80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

dayBullish = close > day1Low  and bullishBias
dayBearish = close < day1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY BIAS — Monday 4AM + 8AM (Gold)
// ═══════════════════════════════════════════
var float weekBiasHigh = na
var float weekBiasLow  = na
var line  wbHL         = na
var line  wbLL         = na

if isMonday and is4AMCandle and showWeekBias
    weekBiasHigh := high
    weekBiasLow  := low

if isMonday and is8AMCandle and showWeekBias
    weekBiasHigh := math.max(weekBiasHigh, high)
    weekBiasLow  := math.min(weekBiasLow,  low)
    line.delete(wbHL)
    line.delete(wbLL)
    wbHL := line.new(bar_index, weekBiasHigh, bar_index + 100, weekBiasHigh, color=color.new(color.yellow, 0), width=2, style=line.style_solid)
    wbLL := line.new(bar_index, weekBiasLow,  bar_index + 100, weekBiasLow,  color=color.new(color.yellow, 0), width=2, style=line.style_solid)
    label.new(bar_index, weekBiasHigh, "⚡ Weekly Bias High", color=color.new(color.yellow, 70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, weekBiasLow,  "⚡ Weekly Bias Low",  color=color.new(color.yellow, 70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY BIAS — 4AM Candle (Green)
// ═══════════════════════════════════════════
var float dayBiasHigh = na
var float dayBiasLow  = na
var line  dbHL        = na
var line  dbLL        = na

if is4AMCandle and showDayBias
    dayBiasHigh := high
    dayBiasLow  := low
    line.delete(dbHL)
    line.delete(dbLL)
    dbHL := line.new(bar_index, dayBiasHigh, bar_index + 100, dayBiasHigh, color=color.new(color.green, 0), width=1, style=line.style_dashed)
    dbLL := line.new(bar_index, dayBiasLow,  bar_index + 100, dayBiasLow,  color=color.new(color.green, 0), width=1, style=line.style_dashed)
    label.new(bar_index, dayBiasHigh, "🕓 4AM High", color=color.new(color.green, 70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, dayBiasLow,  "🕓 4AM Low",  color=color.new(color.green, 70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweepWick = day1Low - low
bearSweepWick = high - day1High
bullishSweep  = low < day1Low   and close > day1Low  and bullSweepWick >= atr * 0.3
bearishSweep  = high > day1High and close < day1High and bearSweepWick >= atr * 0.3

var float lastBullSweepLow  = na
var float lastBearSweepHigh = na
if bullishSweep
    lastBullSweepLow  := low
if bearishSweep
    lastBearSweepHigh := high

plotshape(showSweeps and bullishSweep, "💧 Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime, 0), size=size.small, text="Sweep")
plotshape(showSweeps and bearishSweep, "💧 Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,  0), size=size.small, text="Sweep")

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swingLen        = 5
swingHigh       = ta.highest(high, swingLen * 2 + 1)[swingLen]
swingLow        = ta.lowest(low,   swingLen * 2 + 1)[swingLen]
strongBOSCandle = math.abs(close - open) > atr * 0.7
bullBOS         = close > swingHigh and strongBOSCandle and bullishBias and close[1] <= swingHigh
bearBOS         = close < swingLow  and strongBOSCandle and bearishBias and close[1] >= swingLow

var bool recentBullBOS = false
var bool recentBearBOS = false
var int  bosTimer      = 0

if bullBOS
    recentBullBOS := true
    recentBearBOS := false
    bosTimer      := bar_index
if bearBOS
    recentBearBOS := true
    recentBullBOS := false
    bosTimer      := bar_index
if bar_index - bosTimer > 10
    recentBullBOS := false
    recentBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low  - atr * 0.5, "🔨 BOS ↑", color=color.new(color.green, 20), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index - 1, swingHigh, bar_index + 10, swingHigh, color=color.new(color.green, 40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr * 0.5, "🔨 BOS ↓", color=color.new(color.red,   20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index - 1, swingLow,  bar_index + 10, swingLow,  color=color.new(color.red,   40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
impulseLen   = 3
bullImpulse  = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImpulse  = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBCandle = bullImpulse and close[impulseLen] < open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5
bearOBCandle = bearImpulse and close[impulseLen] > open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5

var float bullOBTop       = na
var float bullOBBot       = na
var float bearOBTop       = na
var float bearOBBot       = na
var bool  bullOBAlive     = false
var bool  bearOBAlive     = false
var box   bullOBBox       = na
var box   bearOBBox       = na
var bool  bullOBRespected = false
var bool  bearOBRespected = false
var int   bullOBTouches   = 0
var int   bearOBTouches   = 0

if bullOBCandle and bullishBias and showOB
    bullOBTop       := high[impulseLen]
    bullOBBot       := low[impulseLen]
    bullOBAlive     := true
    bullOBRespected := false
    bullOBTouches   := 0
    box.delete(bullOBBox)
    bullOBBox := box.new(bar_index - impulseLen, bullOBTop, bar_index + 100, bullOBBot, bgcolor=color.new(color.blue, 80), border_color=color.new(color.blue, 30), border_width=1)
    label.new(bar_index - impulseLen, bullOBTop, "🟦 Bull OB", color=color.new(color.blue, 60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBCandle and bearishBias and showOB
    bearOBTop       := high[impulseLen]
    bearOBBot       := low[impulseLen]
    bearOBAlive     := true
    bearOBRespected := false
    bearOBTouches   := 0
    box.delete(bearOBBox)
    bearOBBox := box.new(bar_index - impulseLen, bearOBTop, bar_index + 100, bearOBBot, bgcolor=color.new(color.orange, 80), border_color=color.new(color.orange, 30), border_width=1)
    label.new(bar_index - impulseLen, bearOBBot, "🟧 Bear OB", color=color.new(color.orange, 60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bullOBAlive and bullOBTouches == 0
    if low <= bullOBTop and close >= bullOBBot
        bullOBRespected := true
        bullOBTouches   += 1
        box.set_bgcolor(bullOBBox, color.new(color.green, 70))
        box.set_border_color(bullOBBox, color.new(color.green, 20))
        label.new(bar_index, bullOBBot, "✅ OB Respected", color=color.new(color.green, 30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearOBAlive and bearOBTouches == 0
    if high >= bearOBBot and close <= bearOBTop
        bearOBRespected := true
        bearOBTouches   += 1
        box.set_bgcolor(bearOBBox, color.new(color.red, 70))
        box.set_border_color(bearOBBox, color.new(color.red, 20))
        label.new(bar_index, bearOBTop, "✅ OB Respected", color=color.new(color.red, 30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullOBAlive and close < bullOBBot
    bullOBAlive     := false
    bullOBRespected := false
    box.set_bgcolor(bullOBBox, color.new(color.gray, 90))

if bearOBAlive and close > bearOBTop
    bearOBAlive     := false
    bearOBRespected := false
    box.set_bgcolor(bearOBBox, color.new(color.gray, 90))

// ═══════════════════════════════════════════
// HIGH QUALITY FVG
// ═══════════════════════════════════════════
candleBody    = math.abs(close[1] - open[1])
strongCandle  = candleBody > atr * 1.0
bullFVGraw    = low > high[2] and strongCandle and close[1] > open[1] and bullishBias
bearFVGraw    = high < low[2] and strongCandle and close[1] < open[1] and bearishBias

var float bullFVGTop       = na
var float bullFVGBot       = na
var float bearFVGTop       = na
var float bearFVGBot       = na
var bool  bullFVGAlive     = false
var bool  bearFVGAlive     = false
var box   bullFVGBox       = na
var box   bearFVGBox       = na
var bool  bullFVGRespected = false
var bool  bearFVGRespected = false

if bullFVGraw and showFVG
    bullFVGTop       := low
    bullFVGBot       := high[2]
    bullFVGAlive     := true
    bullFVGRespected := false
    box.delete(bullFVGBox)
    bullFVGBox := box.new(bar_index - 2, bullFVGTop, bar_index + 100, bullFVGBot, bgcolor=color.new(color.green, 85), border_color=color.new(color.green, 40), border_width=1)
    label.new(bar_index, bullFVGBot, "📦 FVG", color=color.new(color.green, 50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGraw and showFVG
    bearFVGTop       := low[2]
    bearFVGBot       := high
    bearFVGAlive     := true
    bearFVGRespected := false
    box.delete(bearFVGBox)
    bearFVGBox := box.new(bar_index - 2, bearFVGTop, bar_index + 100, bearFVGBot, bgcolor=color.new(color.red, 85), border_color=color.new(color.red, 40), border_width=1)
    label.new(bar_index, bearFVGTop, "📦 FVG", color=color.new(color.red, 50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullFVGAlive and low <= bullFVGTop and close >= bullFVGTop
    bullFVGRespected := true
    box.set_bgcolor(bullFVGBox, color.new(color.green, 70))
    box.set_border_color(bullFVGBox, color.new(color.green, 20))
    label.new(bar_index, bullFVGBot, "✅ FVG Respected", color=color.new(color.green, 30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGAlive and high >= bearFVGBot and close <= bearFVGBot
    bearFVGRespected := true
    box.set_bgcolor(bearFVGBox, color.new(color.red, 70))
    box.set_border_color(bearFVGBox, color.new(color.red, 20))
    label.new(bar_index, bearFVGTop, "✅ FVG Respected", color=color.new(color.red, 30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullFVGAlive and close < bullFVGBot
    bullFVGAlive     := false
    bullFVGRespected := false
    box.set_bgcolor(bullFVGBox, color.new(color.gray, 90))

if bearFVGAlive and close > bearFVGTop
    bearFVGAlive     := false
    bearFVGRespected := false
    box.set_bgcolor(bearFVGBox, color.new(color.gray, 90))

// ═══════════════════════════════════════════
// RSI DIVERGENCE
// ═══════════════════════════════════════════
rsi = ta.rsi(close, 14)
lkb = 7

priceLL = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
priceHL = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
priceHH = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
priceLH = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)

rsiHL = ta.lowest(rsi,  lkb) > ta.lowest(rsi[lkb],  lkb)
rsiLL = ta.lowest(rsi,  lkb) < ta.lowest(rsi[lkb],  lkb)
rsiLH = ta.highest(rsi, lkb) < ta.highest(rsi[lkb], lkb)
rsiHH = ta.highest(rsi, lkb) > ta.highest(rsi[lkb], lkb)

rsiBullExtreme = rsi < 40
rsiBearExtreme = rsi > 60

strongBullDiv = priceLL and rsiHL
hiddenBullDiv = priceHL and rsiLL
strongBearDiv = priceHH and rsiLH
hiddenBearDiv = priceLH and rsiHH

anyBullDiv = strongBullDiv or hiddenBullDiv
anyBearDiv = strongBearDiv or hiddenBearDiv

if showDiv and strongBullDiv
    label.new(bar_index, low  - atr * 0.3, "◆ Strong Bull Div", color=color.new(color.green,  0), textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and hiddenBullDiv
    label.new(bar_index, low  - atr * 0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,   0), textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and strongBearDiv
    label.new(bar_index, high + atr * 0.3, "◆ Strong Bear Div", color=color.new(color.red,    0), textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hiddenBearDiv
    label.new(bar_index, high + atr * 0.3, "◇ Hidden Bear Div", color=color.new(color.orange, 0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdLine, signalLine, histLine] = ta.macd(close, 12, 26, 9)
macdBullish  = macdLine > signalLine
macdBearish  = macdLine < signalLine
macdMomentum = histLine > histLine[1]

// ═══════════════════════════════════════════
// CONFLUENCE SCORE — 15 POINTS
// ═══════════════════════════════════════════
longScore = 0
longScore := longScore + (bullishBias      and allowLongs   ? 1 : 0)  // 1.  EMA trend
longScore := longScore + (monthBullAlign                    ? 1 : 0)  // 2.  Monthly
longScore := longScore + (weekBullish                       ? 1 : 0)  // 3.  Weekly
longScore := longScore + (dayBullish                        ? 1 : 0)  // 4.  Daily
longScore := longScore + (bullishSweep                      ? 1 : 0)  // 5.  Sweep
longScore := longScore + (recentBullBOS                     ? 1 : 0)  // 6.  BOS
longScore := longScore + (bullOBRespected                   ? 1 : 0)  // 7.  OB
longScore := longScore + (bullFVGRespected                  ? 1 : 0)  // 8.  FVG
longScore := longScore + (strongBullDiv ? 2 : hiddenBullDiv ? 1 : 0) // 9-10. Div
longScore := longScore + (macdBullish                       ? 1 : 0)  // 11. MACD
longScore := longScore + (rsiBullExtreme                    ? 1 : 0)  // 12. RSI
longScore := longScore + (isTrending                        ? 1 : 0)  // 13. ADX
longScore := longScore + (htfBullConfirm                    ? 1 : 0)  // 14. HTF
longScore := longScore + (activeSession                     ? 1 : 0)  // 15. Session

shortScore = 0
shortScore := shortScore + (bearishBias   and allowShorts   ? 1 : 0)
shortScore := shortScore + (monthBearAlign                  ? 1 : 0)
shortScore := shortScore + (weekBearish                     ? 1 : 0)
shortScore := shortScore + (dayBearish                      ? 1 : 0)
shortScore := shortScore + (bearishSweep                    ? 1 : 0)
shortScore := shortScore + (recentBearBOS                   ? 1 : 0)
shortScore := shortScore + (bearOBRespected                 ? 1 : 0)
shortScore := shortScore + (bearFVGRespected                ? 1 : 0)
shortScore := shortScore + (strongBearDiv ? 2 : hiddenBearDiv ? 1 : 0)
shortScore := shortScore + (macdBearish                     ? 1 : 0)
shortScore := shortScore + (rsiBearExtreme                  ? 1 : 0)
shortScore := shortScore + (isTrending                      ? 1 : 0)
shortScore := shortScore + (htfBearConfirm                  ? 1 : 0)
shortScore := shortScore + (activeSession                   ? 1 : 0)

// ═══════════════════════════════════════════
// CONFIDENCE LABEL
// ═══════════════════════════════════════════
confidenceLabel(score) =>
    score >= 13 ? "🔥 A+ PERFECT"  :
     score >= 11 ? "💪 STRONG"      :
     score >= 9  ? "✅ GOOD TRADE"  :
                   "⏳ WAIT"

confidenceColor(score) =>
    score >= 13 ? color.new(color.yellow, 0) :
     score >= 11 ? color.new(color.green,  0) :
     score >= 9  ? color.new(color.teal,   0) :
                   color.new(color.gray,   0)

// ═══════════════════════════════════════════
// RISK MANAGEMENT
// ═══════════════════════════════════════════
riskAmount = accountSize * (riskPercent / 100)
slBuffer   = longScore >= 11 or shortScore >= 11 ? atr * 0.2 : atr * 0.5

longSL    = not na(lastBullSweepLow)  ? lastBullSweepLow  - slBuffer : close - atr * 1.5
longRisk  = math.max(close - longSL,  0.0001)
longTP    = close + longRisk * 3
longSize  = riskAmount / longRisk

shortSL   = not na(lastBearSweepHigh) ? lastBearSweepHigh + slBuffer : close + atr * 1.5
shortRisk = math.max(shortSL - close, 0.0001)
shortTP   = close - shortRisk * 3
shortSize = riskAmount / shortRisk

// ═══════════════════════════════════════════
// ENTRY SIGNALS
// Block trades in extreme bear or wrong session
// ═══════════════════════════════════════════
bullishEntry = longScore  >= minScore and allowLongs  and not extremeBear
bearishEntry = shortScore >= minScore and allowShorts and not extremeBear

longConditions =
     "⬆ BUY — " + str.tostring(longScore) + "/15 — " + confidenceLabel(longScore) + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CORE CONDITIONS:\n" +
     (bullishBias      ? "✅ EMA 200: Bullish\n"           : "❌ EMA 200: Not Bullish\n")       +
     (htfBullConfirm   ? "✅ HTF: All Bullish\n"           : "❌ HTF: Not Aligned\n")            +
     (weekBullish      ? "✅ Weekly Bias: Confirmed\n"      : "❌ Weekly: Not Confirmed\n")       +
     (dayBullish       ? "✅ Daily Bias: Confirmed\n"       : "❌ Daily: Not Confirmed\n")        +
     (recentBullBOS    ? "✅ Break of Structure: Yes\n"     : "❌ BOS: No\n")                     +
     (bullishSweep     ? "✅ Liquidity Sweep: Yes\n"        : "❌ Sweep: No\n")                   +
     (anyBullDiv       ? "✅ RSI Div: " + (strongBullDiv ? "Strong\n" : "Hidden\n") : "❌ RSI Div: None\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CONFLUENCE:\n" +
     (monthBullAlign   ? "✅ Monthly: Aligned\n"            : "➖ Monthly: Not Set\n")            +
     (bullOBRespected  ? "✅ Order Block: Respected\n"      : "➖ OB: Not Hit\n")                 +
     (bullFVGRespected ? "✅ FVG: Respected\n"              : "➖ FVG: Not Respected\n")          +
     (macdBullish      ? "✅ MACD: Bullish\n"               : "➖ MACD: Not Bullish\n")           +
     (rsiBullExtreme   ? "✅ RSI Oversold: " + str.tostring(math.round(rsi, 1)) + "\n" : "➖ RSI: " + str.tostring(math.round(rsi, 1)) + "\n") +
     (isTrending       ? "✅ ADX: Trending (" + str.tostring(math.round(adx, 1)) + ")\n" : "➖ ADX: Ranging\n") +
     (activeSession    ? "✅ Session: " + sessionName + "\n" : "➖ Session: Asian — Low Volume\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "MARKET: " + marketRegime + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL:     $" + str.tostring(math.round(longSL,         2)) + "\n" +
     "🟢 TP:     $" + str.tostring(math.round(longTP,         2)) + "\n" +
     "💰 Risk:   $" + str.tostring(math.round(riskAmount,     2)) + "\n" +
     "🎯 Target: $" + str.tostring(math.round(riskAmount * 3, 2)) + "\n" +
     "📊 Size:    " + str.tostring(math.round(longSize,       4)) + " units"

shortConditions =
     "⬇ SELL — " + str.tostring(shortScore) + "/15 — " + confidenceLabel(shortScore) + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CORE CONDITIONS:\n" +
     (bearishBias      ? "✅ EMA 200: Bearish\n"            : "❌ EMA 200: Not Bearish\n")       +
     (htfBearConfirm   ? "✅ HTF: All Bearish\n"            : "❌ HTF: Not Aligned\n")            +
     (weekBearish      ? "✅ Weekly Bias: Confirmed\n"       : "❌ Weekly: Not Confirmed\n")       +
     (dayBearish       ? "✅ Daily Bias: Confirmed\n"        : "❌ Daily: Not Confirmed\n")        +
     (recentBearBOS    ? "✅ Break of Structure: Yes\n"      : "❌ BOS: No\n")                     +
     (bearishSweep     ? "✅ Liquidity Sweep: Yes\n"         : "❌ Sweep: No\n")                   +
     (anyBearDiv       ? "✅ RSI Div: " + (strongBearDiv ? "Strong\n" : "Hidden\n") : "❌ RSI Div: None\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "CONFLUENCE:\n" +
     (monthBearAlign   ? "✅ Monthly: Aligned\n"             : "➖ Monthly: Not Set\n")            +
     (bearOBRespected  ? "✅ Order Block: Respected\n"       : "➖ OB: Not Hit\n")                 +
     (bearFVGRespected ? "✅ FVG: Respected\n"               : "➖ FVG: Not Respected\n")          +
     (macdBearish      ? "✅ MACD: Bearish\n"                : "➖ MACD: Not Bearish\n")           +
     (rsiBearExtreme   ? "✅ RSI Overbought: " + str.tostring(math.round(rsi, 1)) + "\n" : "➖ RSI: " + str.tostring(math.round(rsi, 1)) + "\n") +
     (isTrending       ? "✅ ADX: Trending (" + str.tostring(math.round(adx, 1)) + ")\n" : "➖ ADX: Ranging\n") +
     (activeSession    ? "✅ Session: " + sessionName + "\n" : "➖ Session: Asian — Low Volume\n") +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "MARKET: " + marketRegime + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL:     $" + str.tostring(math.round(shortSL,        2)) + "\n" +
     "🟢 TP:     $" + str.tostring(math.round(shortTP,        2)) + "\n" +
     "💰 Risk:   $" + str.tostring(math.round(riskAmount,     2)) + "\n" +
     "🎯 Target: $" + str.tostring(math.round(riskAmount * 3, 2)) + "\n" +
     "📊 Size:    " + str.tostring(math.round(shortSize,      4)) + " units"

if showSignals and bullishEntry
    label.new(bar_index, low,  longConditions,  color=color.new(color.green, 10), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index, longSL,  bar_index + 40, longSL,  color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP,  bar_index + 40, longTP,  color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index + 40, longSL,  "🔴 SL", color=color.new(color.red,   0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index + 40, longTP,  "🟢 TP", color=color.new(color.green, 0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearishEntry
    label.new(bar_index, high, shortConditions, color=color.new(color.red,   10), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index + 40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index + 40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index + 40, shortSL, "🔴 SL", color=color.new(color.red,   0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index + 40, shortTP, "🟢 TP", color=color.new(color.green, 0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DASHBOARD
// ═══════════════════════════════════════════
var table dash = table.new(position.top_right, 2, 22, bgcolor=color.new(color.black, 60), border_color=color.new(color.gray, 50), border_width=1)

table.cell(dash, 0, 0,  "📊 CONDITION",       text_color=color.white,  text_size=size.small, bgcolor=color.new(color.gray, 30))
table.cell(dash, 1, 0,  "STATUS",             text_color=color.white,  text_size=size.small, bgcolor=color.new(color.gray, 30))

// Market filters
table.cell(dash, 0, 1,  "── MARKET ──",       text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 1,  "──────────────",     text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 2,  "Market Regime",      text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 2,  marketRegime,         text_color=bullMarket and not extremeBear ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 3,  "ADX Trend Filter",   text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 3,  adxStatus,            text_color=adxColor,     text_size=size.small)
table.cell(dash, 0, 4,  "Session",            text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 4,  sessionName,          text_color=sessionColor, text_size=size.small)
table.cell(dash, 0, 5,  "HTF Alignment",      text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 5,  htfStatus,            text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.small)

// Core conditions
table.cell(dash, 0, 6,  "── CORE ──",         text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 6,  "──────────────",     text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 7,  "EMA 200 Trend",      text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 7,  bullishBias ? "🟢 BULL" : "🔴 BEAR",            text_color=bullishBias ? color.green : color.red,    text_size=size.small)
table.cell(dash, 0, 8,  "Monthly Bias",       text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 8,  monthBullAlign ? "🟢 BULL" : monthBearAlign ? "🔴 BEAR" : "⏳ WAIT",  text_color=monthBullAlign ? color.green : monthBearAlign ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 0, 9,  "Weekly Bias",        text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 9,  weekBullish ? "🟢 BULL" : weekBearish ? "🔴 BEAR" : "⏳ WAIT",         text_color=weekBullish ? color.green : weekBearish ? color.red : color.yellow,       text_size=size.small)
table.cell(dash, 0, 10, "Daily Bias",         text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 10, dayBullish ? "🟢 BULL" : dayBearish ? "🔴 BEAR" : "⏳ WAIT",           text_color=dayBullish ? color.green : dayBearish ? color.red : color.yellow,         text_size=size.small)
table.cell(dash, 0, 11, "Break of Structure", text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 11, recentBullBOS ? "🟢 BULL BOS" : recentBearBOS ? "🔴 BEAR BOS" : "⏳ NONE", text_color=recentBullBOS ? color.green : recentBearBOS ? color.red : color.gray, text_size=size.small)
table.cell(dash, 0, 12, "Liquidity Sweep",    text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 12, bullishSweep ? "🟢 BULL" : bearishSweep ? "🔴 BEAR" : "⏳ NONE",       text_color=bullishSweep or bearishSweep ? color.green : color.gray,                  text_size=size.small)
table.cell(dash, 0, 13, "RSI Divergence",     text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 13, strongBullDiv ? "🟢 Strong Bull" : hiddenBullDiv ? "🟢 Hidden Bull" : strongBearDiv ? "🔴 Strong Bear" : hiddenBearDiv ? "🔴 Hidden Bear" : "⏳ NONE",      text_color=anyBullDiv ? color.green : anyBearDiv ? color.red : color.gray, text_size=size.small)

// Confluence
table.cell(dash, 0, 14, "── CONFLUENCE ──",   text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 14, "──────────────",     text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 15, "Order Block",        text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 15, bullOBRespected ? "✅ BULL RESP" : bearOBRespected ? "✅ BEAR RESP" : bullOBAlive ? "🟦 LIVE" : bearOBAlive ? "🟧 LIVE" : "⏳ NONE",                        text_color=bullOBRespected or bearOBRespected ? color.green : bullOBAlive or bearOBAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 0, 16, "Fair Value Gap",     text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 16, bullFVGRespected ? "✅ BULL RESP" : bearFVGRespected ? "✅ BEAR RESP" : bullFVGAlive ? "📦 LIVE" : bearFVGAlive ? "📦 LIVE" : "⏳ NONE",                    text_color=bullFVGRespected or bearFVGRespected ? color.green : bullFVGAlive or bearFVGAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 0, 17, "MACD",              text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 17, macdBullish ? "🟢 BULL" : "🔴 BEAR",            text_color=macdBullish ? color.green : color.red,    text_size=size.small)
table.cell(dash, 0, 18, "RSI Value",         text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 18, str.tostring(math.round(rsi, 1)) + (rsiBullExtreme ? " 🟢 Oversold" : rsiBearExtreme ? " 🔴 Overbought" : " ⚪ Neutral"), text_color=rsiBullExtreme ? color.green : rsiBearExtreme ? color.red : color.white, text_size=size.small)

// Risk
table.cell(dash, 0, 19, "── RISK ──",         text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 1, 19, "──────────────",     text_color=color.gray,   text_size=size.tiny,  bgcolor=color.new(color.gray, 80))
table.cell(dash, 0, 20, "Risk / Target",      text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 20, "💰 $" + str.tostring(math.round(riskAmount, 2)) + " → 🎯 $" + str.tostring(math.round(riskAmount * 3, 2)), text_color=color.yellow, text_size=size.small)

// Signal
table.cell(dash, 0, 21, "🚦 SIGNAL",          text_color=color.white,  text_size=size.small, bgcolor=color.new(color.gray, 30))
table.cell(dash, 1, 21,
     bullishEntry ? "⬆ BUY "  + str.tostring(longScore)  + "/15 " + confidenceLabel(longScore)  :
     bearishEntry ? "⬇ SELL " + str.tostring(shortScore) + "/15 " + confidenceLabel(shortScore) :
     extremeBear  ? "🚨 EXTREME BEAR — NO TRADES"                                               :
     not isTrending ? "📉 MARKET RANGING — WAIT"                                                :
     not activeSession ? "😴 ASIAN SESSION — WAIT"                                              :
     "⏳ WAIT — " + str.tostring(math.max(longScore, shortScore)) + "/15",
     text_color=bullishEntry ? color.green : bearishEntry ? color.red : color.yellow,
     text_size=size.small)
```

```python
// ============================================================
// Manny's Gold Strategy V2
// IMPORTANT: Chart timezone must be set to UTC in TradingView
// for bias lines to work correctly.
// Auto detects timeframe and adjusts strategy accordingly.
// Works on ALL timeframes from 1M to Weekly.
// ============================================================
//@version=6
indicator("Manny's Gold Strategy V2", overlay=true, max_lines_count=500, max_labels_count=500, max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
showMonthly   = input.bool(true,  "📅 Monthly High/Low Lines",          group="📊 Chart Levels")
showWeekly    = input.bool(true,  "📅 Weekly High/Low Lines",           group="📊 Chart Levels")
showDaily     = input.bool(true,  "📅 Daily High/Low Lines (3 Days)",   group="📊 Chart Levels")
showEQLevels  = input.bool(true,  "📊 Equal Highs and Lows",            group="📊 Chart Levels")
showWeekBias  = input.bool(true,  "⚡ Weekly Bias Lines (Yellow)",      group="🧭 Bias Levels")
showDayBias   = input.bool(true,  "🟢 Daily Bias Lines (Green)",        group="🧭 Bias Levels")
showEMA       = input.bool(true,  "📉 EMA 200 + EMA 50",                group="🔧 Features")
showSweeps    = input.bool(true,  "💧 Liquidity Sweeps",                group="🔧 Features")
showFVG       = input.bool(true,  "📦 Fair Value Gaps",                 group="🔧 Features")
showDiv       = input.bool(true,  "📐 RSI Divergence",                  group="🔧 Features")
showBOS       = input.bool(true,  "🔨 Break of Structure",              group="🔧 Features")
showOB        = input.bool(true,  "🟦 Order Blocks",                    group="🔧 Features")
showSignals   = input.bool(true,  "🚦 Buy/Sell Signals",                group="🔧 Features")
showZone      = input.bool(true,  "📊 Premium/Discount Zone",           group="🔧 Features")
fvgMinSizeIn  = input.float(0.5,  "FVG Min Size (x ATR)",               group="🔧 Features", minval=0.1)
modeOverride  = input.string("Auto", "Trading Mode Override",
     options=["Auto", "Scalping", "Day Trade", "Swing Trade"],
     group="🎯 Signal Quality",
     tooltip="Auto = detects from chart timeframe. Override only if needed.")
minBonusOver  = input.int(0,      "Min Bonus Score Override (0=auto)",  group="🎯 Signal Quality", minval=0, maxval=7)
adxLen        = input.int(14,     "ADX Period",                         group="📊 Market Filter")
adxThreshold  = input.int(25,     "ADX Minimum",                        group="📊 Market Filter")
accountSize   = input.float(1000, "💰 Account Size ($)",                group="💵 Risk Management")
riskPercent   = input.float(1.0,  "⚠️ Risk Per Trade (%)",             group="💵 Risk Management")
atrLength     = input.int(14,     "ATR Period",                         group="💵 Risk Management")
goldBiasTime  = input.string("1AM / 5AM (Standard Time)", "🥇 Gold Bias Candle Time",
     options=["1AM / 5AM (Standard Time)", "2AM / 6AM (Daylight Saving)"],
     group="🥇 Gold Settings",
     tooltip="Standard=winter 1AM/5AM. Daylight Saving=summer 2AM/6AM. Switch manually when clocks change.")

// ═══════════════════════════════════════════
// AUTO TIMEFRAME DETECTION
// ═══════════════════════════════════════════
tfMins = timeframe.in_seconds() / 60

autoIsScalp  = tfMins <= 5
autoDayTrade = tfMins > 5  and tfMins <= 30
autoSwing    = tfMins > 30 and tfMins <= 240
autoLongTerm = tfMins > 240

autoModeStr = autoIsScalp ? "Scalping" : autoDayTrade ? "Day Trade" : autoSwing ? "Swing Trade" : "Long Term"

isScalp    = modeOverride == "Auto" ? autoIsScalp  : modeOverride == "Scalping"
isSwing    = modeOverride == "Auto" ? autoSwing    : modeOverride == "Swing Trade"
isDayTrade = modeOverride == "Auto" ? autoDayTrade : modeOverride == "Day Trade"

activeModeStr = modeOverride == "Auto" ? autoModeStr + " (Auto)" : modeOverride

tfLabel = tfMins == 1 ? "1M" : tfMins == 3 ? "3M" : tfMins == 5 ? "5M" : tfMins == 15 ? "15M" : tfMins == 30 ? "30M" : tfMins == 60 ? "1H" : tfMins == 240 ? "4H" : tfMins == 1440 ? "Daily" : tfMins == 10080 ? "Weekly" : str.tostring(math.round(tfMins)) + "M"

// Mode parameters
// FIX 5 — structure minimum adjusted per mode
// FIX 6 — bonus minimum adjusted per mode
minBonus   = minBonusOver > 0 ? minBonusOver : isScalp ? 2 : isSwing ? 4 : 3
bosReset   = isScalp ? 50 : isSwing ? 20 : 30
fvgMinSize = isScalp ? 0.3 : isSwing ? 0.7 : fvgMinSizeIn
minStr2    = isScalp ? 2 : isSwing ? 3 : 2

// FIX 4 — signal cooldown timeframe aware
sigCooldown = tfMins <= 1 ? 15 : tfMins <= 5 ? 12 : tfMins <= 15 ? 10 : tfMins <= 30 ? 8 : 5

// ═══════════════════════════════════════════
// GOLD BIAS TIME
// ═══════════════════════════════════════════
isStandard    = goldBiasTime == "1AM / 5AM (Standard Time)"
dailyBiasHour = isStandard ? 1 : 2
weekBiasHour1 = isStandard ? 1 : 2
weekBiasHour2 = isStandard ? 5 : 6
dailyBiasLbl  = isStandard ? "🕐 1AM 4H" : "🕑 2AM 4H"
weekBiasLbl   = isStandard ? "⚡ 1AM+5AM" : "⚡ 2AM+6AM"

// ═══════════════════════════════════════════
// TIME HELPERS — GMT+0
// ═══════════════════════════════════════════
gmtHour  = hour(time, "GMT+0")
gmtMin   = minute(time, "GMT+0")
gmtDow   = dayofweek(time, "GMT+0")
isMonday = gmtDow == 2
isNewDay   = ta.change(time("D")) != 0
isNewWeek  = ta.change(time("W")) != 0
isNewMonth = ta.change(time("M")) != 0

// ═══════════════════════════════════════════
// SESSION FILTER
// ═══════════════════════════════════════════
londonOpen    = gmtHour >= 7  and gmtHour < 12
nyOpen        = gmtHour >= 13 and gmtHour < 17
overlapSess   = gmtHour >= 12 and gmtHour < 13
activeSession = londonOpen or nyOpen or overlapSess

sessionColor = overlapSess ? color.yellow : londonOpen ? color.green : nyOpen ? color.blue : color.gray
sessionName  = overlapSess ? "⚡ Overlap — BEST" : londonOpen ? "🇬🇧 London — ACTIVE" : nyOpen ? "🇺🇸 New York — ACTIVE" : "😴 Asian — LOW VOLUME"

// ═══════════════════════════════════════════
// ATR + EMA
// ═══════════════════════════════════════════
atr         = ta.atr(atrLength)
ema200      = ta.ema(close, 200)
ema50       = ta.ema(close, 50)
bullishBias = close > ema200
bearishBias = close < ema200
bullMarket  = ema50 > ema200

pctBelowEMA = (ema200 - close) / ema200 * 100
extremeBear = pctBelowEMA > 20
allowLongs  = not extremeBear
allowShorts = not extremeBear

regimeStr = extremeBear ? "🚨 EXTREME BEAR" : bullMarket ? "🟢 BULL MARKET" : "🔴 BEAR MARKET"
regimeCol = extremeBear ? color.red : bullMarket ? color.green : color.red

ema200Col = bullishBias ? color.new(color.green, 0)  : color.new(color.red, 0)
ema50Col  = bullMarket  ? color.new(color.green, 50) : color.new(color.red, 50)
plot(showEMA ? ema200 : na, "EMA 200", color=ema200Col, linewidth=2)
plot(showEMA ? ema50  : na, "EMA 50",  color=ema50Col,  linewidth=1)

// ═══════════════════════════════════════════
// ADX
// ═══════════════════════════════════════════
[diPlus, diMinus, adx] = ta.dmi(adxLen, adxLen)
isTrending = adx >= adxThreshold
adxStr     = isTrending ? "📈 TRENDING (" + str.tostring(math.round(adx,1)) + ")" : "📉 RANGING (" + str.tostring(math.round(adx,1)) + ")"
adxCol     = isTrending ? color.green : color.red

// ═══════════════════════════════════════════
// HTF CONFIRMATION
// FIX 3 — adjusted per timeframe
// 4H chart only needs 4H EMA — weekly and daily
// bias gates already confirm higher TF direction
// ═══════════════════════════════════════════
htf1HClose   = request.security(syminfo.tickerid, "60",  close,              lookahead=barmerge.lookahead_off)
htf1HEMA200  = request.security(syminfo.tickerid, "60",  ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htf4HClose   = request.security(syminfo.tickerid, "240", close,              lookahead=barmerge.lookahead_off)
htf4HEMA200  = request.security(syminfo.tickerid, "240", ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htfDClose    = request.security(syminfo.tickerid, "D",   close,              lookahead=barmerge.lookahead_off)
htfDEMA200   = request.security(syminfo.tickerid, "D",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htfWClose    = request.security(syminfo.tickerid, "W",   close,              lookahead=barmerge.lookahead_off)
htfWEMA200   = request.security(syminfo.tickerid, "W",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)

htf1HBull  = htf1HClose > htf1HEMA200
htf4HBull  = htf4HClose > htf4HEMA200
htfDBull   = htfDClose  > htfDEMA200
htfWBull   = htfWClose  > htfWEMA200
htf1HBear  = htf1HClose < htf1HEMA200
htf4HBear  = htf4HClose < htf4HEMA200
htfDBear   = htfDClose  < htfDEMA200
htfWBear   = htfWClose  < htfWEMA200

// FIX 3 — HTF logic per timeframe
// Scalping 1M-5M  = 1H + 4H + Daily must agree
// Day Trade 15-30M = 1H + 4H must agree
// Swing 1H-4H     = 4H EMA only (bias gates cover higher TF)
// Long Term Daily+ = Daily + Weekly must agree
htfBullConfirm = autoIsScalp ? (htf1HBull and htf4HBull and htfDBull) : autoDayTrade ? (htf1HBull and htf4HBull) : autoSwing ? htf4HBull : (htfDBull and htfWBull)
htfBearConfirm = autoIsScalp ? (htf1HBear and htf4HBear and htfDBear) : autoDayTrade ? (htf1HBear and htf4HBear) : autoSwing ? htf4HBear : (htfDBear and htfWBear)

htfNeeded  = autoIsScalp ? "1H+4H+D" : autoDayTrade ? "1H+4H" : autoSwing ? "4H EMA" : "D+W"

htf1HE = htf1HBull ? "🟢" : "🔴"
htf4HE = htf4HBull ? "🟢" : "🔴"
htfDE  = htfDBull  ? "🟢" : "🔴"
htfWE  = htfWBull  ? "🟢" : "🔴"

htfStrBull = "🟢 " + htfNeeded + " BULL ✓"
htfStrBear = "🔴 " + htfNeeded + " BEAR ✓"
htfStrWait = "⏳ 1H:" + htf1HE + " 4H:" + htf4HE + " D:" + htfDE + " W:" + htfWE
htfStr     = htfBullConfirm ? htfStrBull : htfBearConfirm ? htfStrBear : htfStrWait

// ═══════════════════════════════════════════
// MONTHLY LEVELS
// ═══════════════════════════════════════════
month1High = ta.valuewhen(isNewMonth, high[1], 0)
month1Low  = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, month1High, bar_index+100, month1High, color=color.new(color.purple,0), width=2)
    mLL := line.new(bar_index, month1Low,  bar_index+100, month1Low,  color=color.new(color.purple,0), width=2)
    label.new(bar_index, month1High, "M High", color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, month1Low,  "M Low",  color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)

monthBullAlign = close > month1Low  and bullishBias
monthBearAlign = close < month1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY LEVELS
// ═══════════════════════════════════════════
week1High = ta.valuewhen(isNewWeek, high[1], 0)
week1Low  = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, week1High, bar_index+100, week1High, color=color.new(color.blue,0), width=2)
    wLL := line.new(bar_index, week1Low,  bar_index+100, week1Low,  color=color.new(color.blue,0), width=2)
    label.new(bar_index, week1High, "W High", color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, week1Low,  "W Low",  color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY LEVELS — 3 days fading
// ═══════════════════════════════════════════
day1High = ta.valuewhen(isNewDay, high[1], 0)
day1Low  = ta.valuewhen(isNewDay, low[1],  0)
day2High = ta.valuewhen(isNewDay, high[1], 1)
day2Low  = ta.valuewhen(isNewDay, low[1],  1)
day3High = ta.valuewhen(isNewDay, high[1], 2)
day3Low  = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, day1High, bar_index+100, day1High, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, day1Low,  bar_index+100, day1Low,  color=color.new(color.black,0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, day2High, bar_index+100, day2High, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, day2Low,  bar_index+100, day2Low,  color=color.new(color.black,40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, day3High, bar_index+100, day3High, color=color.new(color.black,70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, day3Low,  bar_index+100, day3Low,  color=color.new(color.black,70), width=1, style=line.style_dotted)
    label.new(bar_index, day1High, "D1H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day1Low,  "D1L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2High, "D2H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2Low,  "D2L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3High, "D3H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3Low,  "D3L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// EQUAL HIGHS AND LOWS
// ═══════════════════════════════════════════
swingH     = ta.pivothigh(high, 5, 5)
swingL     = ta.pivotlow(low,   5, 5)
prevSwingH = ta.valuewhen(not na(swingH), swingH, 1)
prevSwingL = ta.valuewhen(not na(swingL), swingL, 1)
isEQH      = not na(swingH) and not na(prevSwingH) and math.abs(swingH - prevSwingH) <= atr * 0.1
isEQL      = not na(swingL) and not na(prevSwingL) and math.abs(swingL - prevSwingL) <= atr * 0.1

if showEQLevels and isEQH
    line.new(bar_index-5, swingH, bar_index+20, swingH, color=color.new(color.red,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, swingH, "EQH — Liquidity Above", color=color.new(color.red,70), textcolor=color.red, style=label.style_label_right, size=size.tiny)
if showEQLevels and isEQL
    line.new(bar_index-5, swingL, bar_index+20, swingL, color=color.new(color.green,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, swingL, "EQL — Liquidity Below", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_right, size=size.tiny)

eqhSwept = not na(swingH) and high > swingH and close < swingH
eqlSwept = not na(swingL) and low  < swingL and close > swingL

// ═══════════════════════════════════════════
// GATE 1 — WEEKLY BIAS (Yellow Lines)
// FIX 2 — uses request.security to pull 4H
// candle levels so lines are identical on ALL
// timeframes — 1M chart shows same as 4H chart
// ═══════════════════════════════════════════
weekBias4HC1High = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour1 and minute(time, "GMT+0") == 0, high, 0), lookahead=barmerge.lookahead_off)
weekBias4HC1Low  = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour1 and minute(time, "GMT+0") == 0, low,  0), lookahead=barmerge.lookahead_off)
weekBias4HC2High = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour2 and minute(time, "GMT+0") == 0, high, 0), lookahead=barmerge.lookahead_off)
weekBias4HC2Low  = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour2 and minute(time, "GMT+0") == 0, low,  0), lookahead=barmerge.lookahead_off)

// Combine both candles into the weekly range
wbRangeHigh = not na(weekBias4HC1High) and not na(weekBias4HC2High) ? math.max(weekBias4HC1High, weekBias4HC2High) : not na(weekBias4HC1High) ? weekBias4HC1High : weekBias4HC2High
wbRangeLow  = not na(weekBias4HC1Low)  and not na(weekBias4HC2Low)  ? math.min(weekBias4HC1Low,  weekBias4HC2Low)  : not na(weekBias4HC1Low)  ? weekBias4HC1Low  : weekBias4HC2Low

var float weekBiasHigh = na
var float weekBiasLow  = na
var line  wbHL         = na
var line  wbLL         = na
var bool  weekBullish  = false
var bool  weekBearish  = false

if isNewWeek
    weekBullish  := false
    weekBearish  := false
    weekBiasHigh := na
    weekBiasLow  := na

// Update lines when the weekly range changes
if not na(wbRangeHigh) and not na(wbRangeLow)
    if wbRangeHigh != weekBiasHigh or wbRangeLow != weekBiasLow
        weekBiasHigh := wbRangeHigh
        weekBiasLow  := wbRangeLow
        if showWeekBias
            line.delete(wbHL)
            line.delete(wbLL)
            wbHL := line.new(bar_index, weekBiasHigh, bar_index+500, weekBiasHigh, color=color.new(color.yellow,0), width=2)
            wbLL := line.new(bar_index, weekBiasLow,  bar_index+500, weekBiasLow,  color=color.new(color.yellow,0), width=2)
            label.new(bar_index, weekBiasHigh, weekBiasLbl + " High", color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
            label.new(bar_index, weekBiasLow,  weekBiasLbl + " Low",  color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

if not na(weekBiasHigh) and not na(weekBiasLow)
    if close > weekBiasHigh
        weekBullish := true
        weekBearish := false
    else if close < weekBiasLow
        weekBearish := true
        weekBullish := false

// ═══════════════════════════════════════════
// GATE 2 — DAILY BIAS (Green Dashed Lines)
// FIX 1 — uses request.security to pull 4H
// candle levels so lines are identical on ALL
// timeframes — 1M chart shows same as 4H chart
// ═══════════════════════════════════════════
dailyBias4HHigh = request.security(syminfo.tickerid, "240", ta.valuewhen(hour(time, "GMT+0") == dailyBiasHour and minute(time, "GMT+0") == 0, high, 0), lookahead=barmerge.lookahead_off)
dailyBias4HLow  = request.security(syminfo.tickerid, "240", ta.valuewhen(hour(time, "GMT+0") == dailyBiasHour and minute(time, "GMT+0") == 0, low,  0), lookahead=barmerge.lookahead_off)

var float dayBiasHigh = na
var float dayBiasLow  = na
var line  dbHL        = na
var line  dbLL        = na
var bool  dayBullish  = false
var bool  dayBearish  = false

if isNewDay
    dayBullish := false
    dayBearish := false

// Update lines when the daily 4H level changes
if not na(dailyBias4HHigh) and not na(dailyBias4HLow)
    if dailyBias4HHigh != dayBiasHigh or dailyBias4HLow != dayBiasLow
        dayBiasHigh := dailyBias4HHigh
        dayBiasLow  := dailyBias4HLow
        if showDayBias
            line.delete(dbHL)
            line.delete(dbLL)
            dbHL := line.new(bar_index, dayBiasHigh, bar_index+100, dayBiasHigh, color=color.new(color.green,0), width=2, style=line.style_dashed)
            dbLL := line.new(bar_index, dayBiasLow,  bar_index+100, dayBiasLow,  color=color.new(color.green,0), width=2, style=line.style_dashed)
            label.new(bar_index, dayBiasHigh, dailyBiasLbl + " H", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
            label.new(bar_index, dayBiasLow,  dailyBiasLbl + " L", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

if not na(dayBiasHigh) and not na(dayBiasLow)
    if close > dayBiasHigh
        dayBullish := true
        dayBearish := false
    else if close < dayBiasLow
        dayBearish := true
        dayBullish := false

// ═══════════════════════════════════════════
// PREMIUM AND DISCOUNT ZONE
// ═══════════════════════════════════════════
weekRange  = week1High - week1Low
weekMid    = week1Low + weekRange / 2
inDiscount = close < weekMid and not na(weekMid)
inPremium  = close > weekMid and not na(weekMid)

if showZone and isNewWeek and not na(week1High) and not na(week1Low)
    line.new(bar_index, weekMid, bar_index+100, weekMid, color=color.new(color.gray,50), style=line.style_dashed, width=1)
    label.new(bar_index, weekMid, "Equilibrium", color=color.new(color.gray,70), textcolor=color.gray, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swingLen      = 5
swingHigh     = ta.highest(high, swingLen * 2 + 1)[swingLen]
swingLow      = ta.lowest(low,   swingLen * 2 + 1)[swingLen]
strongBOSBody = math.abs(close - open) > atr * 0.7
bullBOS       = close > swingHigh and strongBOSBody and bullishBias and close[1] <= swingHigh
bearBOS       = close < swingLow  and strongBOSBody and bearishBias and close[1] >= swingLow

var bool recentBullBOS = false
var bool recentBearBOS = false
var int  bosTimer      = 0

if bullBOS
    recentBullBOS := true
    recentBearBOS := false
    bosTimer      := bar_index
if bearBOS
    recentBearBOS := true
    recentBullBOS := false
    bosTimer      := bar_index
if bar_index - bosTimer > bosReset
    recentBullBOS := false
    recentBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low - atr*0.5, "🔨 BOS ↑", color=color.new(color.green,20), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index-1, swingHigh, bar_index+10, swingHigh, color=color.new(color.green,40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr*0.5, "🔨 BOS ↓", color=color.new(color.red,20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index-1, swingLow, bar_index+10, swingLow, color=color.new(color.red,40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweepWick = day1Low - low
bearSweepWick = high - day1High
bullishSweep  = low < day1Low   and close > day1Low  and bullSweepWick >= atr * 0.3
bearishSweep  = high > day1High and close < day1High and bearSweepWick >= atr * 0.3

var float lastBullSweepLow  = na
var float lastBearSweepHigh = na
if bullishSweep
    lastBullSweepLow  := low
if bearishSweep
    lastBearSweepHigh := high

bullSweepRej = bullishSweep and (close - low)  > atr * 1.5
bearSweepRej = bearishSweep and (high - close) > atr * 1.5

plotshape(showSweeps and bullishSweep, "💧 Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime,0), size=size.small, text="Sweep")
plotshape(showSweeps and bearishSweep, "💧 Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,0),  size=size.small, text="Sweep")

intBullSweep      = low  < ta.lowest(low[1],   3)[1] and close > ta.lowest(low[1],   3)[1]
intBearSweep      = high > ta.highest(high[1], 3)[1] and close < ta.highest(high[1], 3)[1]
internalBullSweep = intBullSweep and bullishBias
internalBearSweep = intBearSweep and bearishBias

// ═══════════════════════════════════════════
// ORDER BLOCKS — TRIGGER 2
// ═══════════════════════════════════════════
impulseLen   = 3
bullImpulse  = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImpulse  = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBCandle = bullImpulse and close[impulseLen] < open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5
bearOBCandle = bearImpulse and close[impulseLen] > open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5

var float bullOBTop       = na
var float bullOBBot       = na
var float bearOBTop       = na
var float bearOBBot       = na
var bool  bullOBAlive     = false
var bool  bearOBAlive     = false
var box   bullOBBox       = na
var box   bearOBBox       = na
var bool  bullOBRespected = false
var bool  bearOBRespected = false
var int   bullOBTouches   = 0
var int   bearOBTouches   = 0

if bullOBCandle and bullishBias and showOB
    bullOBTop       := high[impulseLen]
    bullOBBot       := low[impulseLen]
    bullOBAlive     := true
    bullOBRespected := false
    bullOBTouches   := 0
    box.delete(bullOBBox)
    bullOBBox := box.new(bar_index-impulseLen, bullOBTop, bar_index+100, bullOBBot, bgcolor=color.new(color.blue,80), border_color=color.new(color.blue,30), border_width=1)
    label.new(bar_index-impulseLen, bullOBTop, "🟦 Bull OB", color=color.new(color.blue,60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBCandle and bearishBias and showOB
    bearOBTop       := high[impulseLen]
    bearOBBot       := low[impulseLen]
    bearOBAlive     := true
    bearOBRespected := false
    bearOBTouches   := 0
    box.delete(bearOBBox)
    bearOBBox := box.new(bar_index-impulseLen, bearOBTop, bar_index+100, bearOBBot, bgcolor=color.new(color.orange,80), border_color=color.new(color.orange,30), border_width=1)
    label.new(bar_index-impulseLen, bearOBBot, "🟧 Bear OB", color=color.new(color.orange,60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bullOBAlive and bullOBTouches == 0
    if low <= bullOBTop and close >= bullOBBot
        bullOBRespected := true
        bullOBTouches   += 1
        box.set_bgcolor(bullOBBox, color.new(color.green,70))
        box.set_border_color(bullOBBox, color.new(color.green,20))
        label.new(bar_index, bullOBBot, "✅ OB Entry Zone", color=color.new(color.green,30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearOBAlive and bearOBTouches == 0
    if high >= bearOBBot and close <= bearOBTop
        bearOBRespected := true
        bearOBTouches   += 1
        box.set_bgcolor(bearOBBox, color.new(color.red,70))
        box.set_border_color(bearOBBox, color.new(color.red,20))
        label.new(bar_index, bearOBTop, "✅ OB Entry Zone", color=color.new(color.red,30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullOBAlive and close < bullOBBot
    bullOBAlive     := false
    bullOBRespected := false
    box.set_bgcolor(bullOBBox, color.new(color.gray,90))

if bearOBAlive and close > bearOBTop
    bearOBAlive     := false
    bearOBRespected := false
    box.set_bgcolor(bearOBBox, color.new(color.gray,90))

// ═══════════════════════════════════════════
// FVG WITH RETRACEMENT ENTRY — TRIGGER 1
// ═══════════════════════════════════════════
candleBody   = math.abs(close[1] - open[1])
strongCandle = candleBody > atr * 1.0
dispCandle   = math.abs(close - open) > atr * 1.5
bullFVGSize  = low - high[2]
bearFVGSize  = low[2] - high
bullFVGraw   = low > high[2] and strongCandle and close[1] > open[1] and bullishBias and bullFVGSize >= atr * fvgMinSize
bearFVGraw   = high < low[2] and strongCandle and close[1] < open[1] and bearishBias and bearFVGSize >= atr * fvgMinSize

var float bullFVGTop       = na
var float bullFVGBot       = na
var float bearFVGTop       = na
var float bearFVGBot       = na
var bool  bullFVGAlive     = false
var bool  bearFVGAlive     = false
var box   bullFVGBox       = na
var box   bearFVGBox       = na
var bool  bullFVGRespected = false
var bool  bearFVGRespected = false
var bool  bullEntryZone    = false
var bool  bearEntryZone    = false
var box   bullEZBox        = na
var box   bearEZBox        = na

if bullFVGraw and showFVG
    bullFVGTop       := low
    bullFVGBot       := high[2]
    bullFVGAlive     := true
    bullFVGRespected := false
    bullEntryZone    := false
    box.delete(bullFVGBox)
    bullFVGBox := box.new(bar_index-2, bullFVGTop, bar_index+100, bullFVGBot, bgcolor=color.new(color.green,85), border_color=color.new(color.green,40), border_width=1)
    label.new(bar_index, bullFVGBot, "📦 Bull FVG", color=color.new(color.green,50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGraw and showFVG
    bearFVGTop       := low[2]
    bearFVGBot       := high
    bearFVGAlive     := true
    bearFVGRespected := false
    bearEntryZone    := false
    box.delete(bearFVGBox)
    bearFVGBox := box.new(bar_index-2, bearFVGTop, bar_index+100, bearFVGBot, bgcolor=color.new(color.red,85), border_color=color.new(color.red,40), border_width=1)
    label.new(bar_index, bearFVGTop, "📦 Bear FVG", color=color.new(color.red,50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullFVGAlive and low <= bullFVGTop and close >= bullFVGBot
    bullFVGRespected := true
    box.set_bgcolor(bullFVGBox, color.new(color.green,70))
    if dispCandle and bullishBias and weekBullish and dayBullish and not bullEntryZone
        bullEntryZone := true
        box.delete(bullEZBox)
        bullEZBox := box.new(bar_index, bullFVGTop, bar_index+30, bullFVGBot, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, bullFVGBot, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_up, size=size.small)

if bearFVGAlive and high >= bearFVGBot and close <= bearFVGTop
    bearFVGRespected := true
    box.set_bgcolor(bearFVGBox, color.new(color.red,70))
    if dispCandle and bearishBias and weekBearish and dayBearish and not bearEntryZone
        bearEntryZone := true
        box.delete(bearEZBox)
        bearEZBox := box.new(bar_index, bearFVGTop, bar_index+30, bearFVGBot, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, bearFVGTop, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_down, size=size.small)

if bullFVGAlive and close < bullFVGBot
    bullFVGAlive     := false
    bullFVGRespected := false
    bullEntryZone    := false
    box.set_bgcolor(bullFVGBox, color.new(color.gray,90))

if bearFVGAlive and close > bearFVGTop
    bearFVGAlive     := false
    bearFVGRespected := false
    bearEntryZone    := false
    box.set_bgcolor(bearFVGBox, color.new(color.gray,90))

// ═══════════════════════════════════════════
// RSI DIVERGENCE
// ═══════════════════════════════════════════
rsi = ta.rsi(close, 14)
lkb = 7
priceLL = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
priceHL = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
priceHH = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
priceLH = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)
rsiHL   = ta.lowest(rsi,  lkb) > ta.lowest(rsi[lkb],  lkb)
rsiLL   = ta.lowest(rsi,  lkb) < ta.lowest(rsi[lkb],  lkb)
rsiLH   = ta.highest(rsi, lkb) < ta.highest(rsi[lkb], lkb)
rsiHH   = ta.highest(rsi, lkb) > ta.highest(rsi[lkb], lkb)

rsiBullExt    = rsi < 40
rsiBearExt    = rsi > 60
strongBullDiv = priceLL and rsiHL
hiddenBullDiv = priceHL and rsiLL
strongBearDiv = priceHH and rsiLH
hiddenBearDiv = priceLH and rsiHH
anyBullDiv    = strongBullDiv or hiddenBullDiv
anyBearDiv    = strongBearDiv or hiddenBearDiv

if showDiv and strongBullDiv
    label.new(bar_index, low - atr*0.3, "◆ Strong Bull Div", color=color.new(color.green,0), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showDiv and hiddenBullDiv
    label.new(bar_index, low - atr*0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,0), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showDiv and strongBearDiv
    label.new(bar_index, high + atr*0.3, "◆ Strong Bear Div", color=color.new(color.red,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hiddenBearDiv
    label.new(bar_index, high + atr*0.3, "◇ Hidden Bear Div", color=color.new(color.orange,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdLine, signalLine, histLine] = ta.macd(close, 12, 26, 9)
macdBullish = macdLine > signalLine
macdBearish = macdLine < signalLine

// ═══════════════════════════════════════════
// TIER 2 — STRUCTURE SCORE
// ═══════════════════════════════════════════
str2Long = 0
str2Long := str2Long + (bullishBias    ? 1 : 0)
str2Long := str2Long + (recentBullBOS  ? 1 : 0)
str2Long := str2Long + (bullishSweep   ? 1 : 0)
str2Long := str2Long + (inDiscount     ? 1 : 0)
str2Long := str2Long + (monthBullAlign ? 1 : 0)

str2Short = 0
str2Short := str2Short + (bearishBias    ? 1 : 0)
str2Short := str2Short + (recentBearBOS  ? 1 : 0)
str2Short := str2Short + (bearishSweep   ? 1 : 0)
str2Short := str2Short + (inPremium      ? 1 : 0)
str2Short := str2Short + (monthBearAlign ? 1 : 0)

// ═══════════════════════════════════════════
// TIER 3 — ENTRY TRIGGERS
// ═══════════════════════════════════════════
trigger1Bull   = bullEntryZone
trigger2Bull   = bullOBRespected
trigger3Bull   = bullSweepRej and weekBullish and dayBullish
anyTriggerBull = trigger1Bull or trigger2Bull or trigger3Bull

trigger1Bear   = bearEntryZone
trigger2Bear   = bearOBRespected
trigger3Bear   = bearSweepRej and weekBearish and dayBearish
anyTriggerBear = trigger1Bear or trigger2Bear or trigger3Bear

trigStrBull = trigger1Bull ? "🎯 FVG Retracement" : trigger2Bull ? "🟦 OB Respected" : trigger3Bull ? "💧 Sweep Rejection" : "⏳ None"
trigStrBear = trigger1Bear ? "🎯 FVG Retracement" : trigger2Bear ? "🟧 OB Respected" : trigger3Bear ? "💧 Sweep Rejection" : "⏳ None"

// ═══════════════════════════════════════════
// TIER 4 — BONUS SCORE
// ═══════════════════════════════════════════
bonusLong = 0
bonusLong := bonusLong + (strongBullDiv     ? 2 : hiddenBullDiv ? 1 : 0)
bonusLong := bonusLong + (macdBullish       ? 1 : 0)
bonusLong := bonusLong + (rsiBullExt        ? 1 : 0)
bonusLong := bonusLong + (isTrending        ? 1 : 0)
bonusLong := bonusLong + (internalBullSweep ? 1 : 0)
bonusLong := bonusLong + (eqlSwept          ? 1 : 0)

bonusShort = 0
bonusShort := bonusShort + (strongBearDiv     ? 2 : hiddenBearDiv ? 1 : 0)
bonusShort := bonusShort + (macdBearish       ? 1 : 0)
bonusShort := bonusShort + (rsiBearExt        ? 1 : 0)
bonusShort := bonusShort + (isTrending        ? 1 : 0)
bonusShort := bonusShort + (internalBearSweep ? 1 : 0)
bonusShort := bonusShort + (eqhSwept          ? 1 : 0)

// ═══════════════════════════════════════════
// SIGNAL FIRING
// ═══════════════════════════════════════════
gate1L = weekBullish
gate2L = dayBullish
gate3L = htfBullConfirm
gate4L = activeSession
gate5L = allowLongs
allGatesLong = gate1L and gate2L and gate3L and gate4L and gate5L

gate1S = weekBearish
gate2S = dayBearish
gate3S = htfBearConfirm
gate4S = activeSession
gate5S = allowShorts
allGatesShort = gate1S and gate2S and gate3S and gate4S and gate5S

var int lastBullBar = 0
var int lastBearBar = 0

rawBullEntry = allGatesLong  and str2Long  >= minStr2 and anyTriggerBull and bonusLong  >= minBonus
rawBearEntry = allGatesShort and str2Short >= minStr2 and anyTriggerBear and bonusShort >= minBonus

// FIX 4 — timeframe aware cooldown
bullishEntry = rawBullEntry and (bar_index - lastBullBar) > sigCooldown
bearishEntry = rawBearEntry and (bar_index - lastBearBar) > sigCooldown

if bullishEntry
    lastBullBar := bar_index
if bearishEntry
    lastBearBar := bar_index

// ═══════════════════════════════════════════
// SIGNAL QUALITY
// ═══════════════════════════════════════════
qualityLong  = bonusLong  >= 7 ? "🔥 A+ HIGH CONFIDENCE" : bonusLong  >= 5 ? "💪 STRONG" : "✅ VALID"
qualityShort = bonusShort >= 7 ? "🔥 A+ HIGH CONFIDENCE" : bonusShort >= 5 ? "💪 STRONG" : "✅ VALID"

// ═══════════════════════════════════════════
// RISK MANAGEMENT — 1:3 RR always
// ═══════════════════════════════════════════
riskAmount = accountSize * (riskPercent / 100)
slBufferL  = bonusLong  >= 6 ? atr * 0.2 : atr * 0.5
slBufferS  = bonusShort >= 6 ? atr * 0.2 : atr * 0.5

longSL    = not na(lastBullSweepLow)  ? lastBullSweepLow  - slBufferL : close - atr * 1.5
longRisk  = math.max(close - longSL,  0.0001)
longTP    = close + longRisk * 3
longSize  = riskAmount / longRisk

shortSL   = not na(lastBearSweepHigh) ? lastBearSweepHigh + slBufferS : close + atr * 1.5
shortRisk = math.max(shortSL - close, 0.0001)
shortTP   = close - shortRisk * 3
shortSize = riskAmount / shortRisk

// ═══════════════════════════════════════════
// SIGNAL LABELS
// ═══════════════════════════════════════════
longLbl =
     "⬆ BUY — " + tfLabel + " — " + qualityLong + "\n" +
     "Mode: " + activeModeStr + " | HTF: " + htfNeeded + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "✅ ALL GATES PASSED\n" +
     "Weekly: 🟢 Bullish | Daily: 🟢 Bullish\n" +
     "HTF: " + htfStr + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "STRUCTURE (" + str.tostring(str2Long) + "/5 need " + str.tostring(minStr2) + "):\n" +
     (bullishBias    ? "✅" : "❌") + " EMA Bull  " + (recentBullBOS ? "✅" : "❌") + " BOS  " + (bullishSweep ? "✅" : "❌") + " Sweep\n" +
     (inDiscount     ? "✅" : "❌") + " Discount Zone  " + (monthBullAlign ? "✅" : "❌") + " Monthly\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRIGGER: " + trigStrBull + "\n" +
     "BONUS: " + str.tostring(bonusLong) + "/7\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL:  $" + str.tostring(math.round(longSL,    2)) + "\n" +
     "🟢 TP:  $" + str.tostring(math.round(longTP,    2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmount,2)) + "  🎯 Target: $" + str.tostring(math.round(riskAmount*3,2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(longSize,  4)) + " units  |  RR: 1:3"

shortLbl =
     "⬇ SELL — " + tfLabel + " — " + qualityShort + "\n" +
     "Mode: " + activeModeStr + " | HTF: " + htfNeeded + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "✅ ALL GATES PASSED\n" +
     "Weekly: 🔴 Bearish | Daily: 🔴 Bearish\n" +
     "HTF: " + htfStr + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "STRUCTURE (" + str.tostring(str2Short) + "/5 need " + str.tostring(minStr2) + "):\n" +
     (bearishBias    ? "✅" : "❌") + " EMA Bear  " + (recentBearBOS ? "✅" : "❌") + " BOS  " + (bearishSweep ? "✅" : "❌") + " Sweep\n" +
     (inPremium      ? "✅" : "❌") + " Premium Zone  " + (monthBearAlign ? "✅" : "❌") + " Monthly\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRIGGER: " + trigStrBear + "\n" +
     "BONUS: " + str.tostring(bonusShort) + "/7\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL:  $" + str.tostring(math.round(shortSL,   2)) + "\n" +
     "🟢 TP:  $" + str.tostring(math.round(shortTP,   2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmount,2)) + "  🎯 Target: $" + str.tostring(math.round(riskAmount*3,2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(shortSize,  4)) + " units  |  RR: 1:3"

if showSignals and bullishEntry
    label.new(bar_index, low,  longLbl,  color=color.new(color.green,10), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index, longSL,  bar_index+40, longSL,  color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP,  bar_index+40, longTP,  color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, longSL,  "🔴 SL $" + str.tostring(math.round(longSL,2)),  color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, longTP,  "🟢 TP $" + str.tostring(math.round(longTP,2)),  color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearishEntry
    label.new(bar_index, high, shortLbl, color=color.new(color.red,10),   textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index+40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index+40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, shortSL, "🔴 SL $" + str.tostring(math.round(shortSL,2)), color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, shortTP, "🟢 TP $" + str.tostring(math.round(shortTP,2)), color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// ALERTS
// ═══════════════════════════════════════════
alertcondition(bullishEntry, title="🟢 GOLD BUY — High Confidence",  message="⬆ BUY SIGNAL — XAUUSD — Check chart for full details")
alertcondition(bearishEntry, title="🔴 GOLD SELL — High Confidence", message="⬇ SELL SIGNAL — XAUUSD — Check chart for full details")
alertcondition(bullishEntry or bearishEntry, title="🚦 GOLD Signal",  message="Gold signal detected on XAUUSD — Check chart")

// ═══════════════════════════════════════════
// WAIT REASON — tells trader exactly what to watch for
// ═══════════════════════════════════════════
wrl1 = not gate1L           ? "⏳ WAIT — Weekly Bias Not Confirmed"              : ""
wrl2 = not gate2L           ? "⏳ WAIT — Daily Bias Not Confirmed"                : ""
wrl3 = not gate3L           ? "⏳ WAIT — HTF (" + htfNeeded + ") Not Aligned"    : ""
wrl4 = not gate4L           ? "⏳ WAIT — Asian Session — Low Volume"              : ""
wrl5 = not gate5L           ? "⏳ WAIT — Extreme Bear Market"                     : ""
wrl6 = str2Long < minStr2   ? "⏳ WAIT — Need " + str.tostring(minStr2 - str2Long) + " More Structure"    : ""
wrl7 = not anyTriggerBull   ? "⏳ WAIT — Need Entry Trigger (FVG/OB/Sweep)"      : ""
wrl8 = bonusLong < minBonus ? "⏳ WAIT — Need " + str.tostring(minBonus - bonusLong) + " More Bonus Pts" : ""
waitReasonLong = not gate1L ? wrl1 : not gate2L ? wrl2 : not gate3L ? wrl3 : not gate4L ? wrl4 : not gate5L ? wrl5 : str2Long < minStr2 ? wrl6 : not anyTriggerBull ? wrl7 : bonusLong < minBonus ? wrl8 : "✅ LONG CONDITIONS MET"

wrs1 = not gate1S            ? "⏳ WAIT — Weekly Bias Not Confirmed"              : ""
wrs2 = not gate2S            ? "⏳ WAIT — Daily Bias Not Confirmed"                : ""
wrs3 = not gate3S            ? "⏳ WAIT — HTF (" + htfNeeded + ") Not Aligned"    : ""
wrs4 = not gate4S            ? "⏳ WAIT — Asian Session — Low Volume"              : ""
wrs5 = not gate5S            ? "⏳ WAIT — Extreme Bear Market"                     : ""
wrs6 = str2Short < minStr2   ? "⏳ WAIT — Need " + str.tostring(minStr2 - str2Short) + " More Structure"   : ""
wrs7 = not anyTriggerBear    ? "⏳ WAIT — Need Entry Trigger (FVG/OB/Sweep)"      : ""
wrs8 = bonusShort < minBonus ? "⏳ WAIT — Need " + str.tostring(minBonus - bonusShort) + " More Bonus Pts" : ""
waitReasonShort = not gate1S ? wrs1 : not gate2S ? wrs2 : not gate3S ? wrs3 : not gate4S ? wrs4 : not gate5S ? wrs5 : str2Short < minStr2 ? wrs6 : not anyTriggerBear ? wrs7 : bonusShort < minBonus ? wrs8 : "✅ SHORT CONDITIONS MET"

// ═══════════════════════════════════════════
// DASHBOARD
// ═══════════════════════════════════════════
var table dash = table.new(position.top_right, 3, 35, bgcolor=color.new(color.black,50), border_color=color.new(color.gray,30), border_width=1)
hdrBG  = color.new(color.blue, 40)
secBG  = color.new(color.gray, 60)
secCol = color.yellow

table.cell(dash, 0, 0, "📊 CONDITION",   text_color=color.white,  text_size=size.small, bgcolor=hdrBG)
table.cell(dash, 1, 0, "STATUS",         text_color=color.white,  text_size=size.small, bgcolor=hdrBG)
table.cell(dash, 2, 0, "🥇 " + tfLabel,  text_color=color.yellow, text_size=size.small, bgcolor=hdrBG)

table.cell(dash, 0, 1, "── MARKET ──",   text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 1, "────────────",   text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 1, "──────",         text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 2, "Regime",         text_color=color.white, text_size=size.small)
table.cell(dash, 1, 2, regimeStr,        text_color=regimeCol,   text_size=size.small)
table.cell(dash, 2, 2, extremeBear ? "🚫" : "✅", text_color=extremeBear ? color.red : color.green, text_size=size.small)

table.cell(dash, 0, 3, "ADX",            text_color=color.white, text_size=size.small)
table.cell(dash, 1, 3, adxStr,           text_color=adxCol,      text_size=size.small)
table.cell(dash, 2, 3, isTrending ? "✅" : "❌", text_color=isTrending ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 4, "Session",        text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 4, sessionName,      text_color=sessionColor, text_size=size.small)
table.cell(dash, 2, 4, activeSession ? "✅" : "❌", text_color=activeSession ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 5, "TF Mode",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 5, activeModeStr,    text_color=color.aqua,  text_size=size.small)
table.cell(dash, 2, 5, "B≥" + str.tostring(minBonus) + " S≥" + str.tostring(minStr2), text_color=color.aqua, text_size=size.tiny)

table.cell(dash, 0, 6, "HTF (" + htfNeeded + ")", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 6, htfStr, text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.tiny)
table.cell(dash, 2, 6, htfBullConfirm ? "✅🟢" : htfBearConfirm ? "✅🔴" : "❌", text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.small)

table.cell(dash, 0, 7, "── TIER 1 GATES ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 7, "────────────────",   text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 7, "──────",             text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 8, "Weekly Bias ⚡",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 8, weekBullish ? "🟢 BULL CONFIRMED" : weekBearish ? "🔴 BEAR CONFIRMED" : "⏳ WAITING", text_color=weekBullish ? color.green : weekBearish ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 8, weekBullish or weekBearish ? "✅" : "❌", text_color=weekBullish or weekBearish ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 9, "Daily Bias 🟢",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 9, dayBullish ? "🟢 BULL CONFIRMED" : dayBearish ? "🔴 BEAR CONFIRMED" : "⏳ WAITING", text_color=dayBullish ? color.green : dayBearish ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 9, dayBullish or dayBearish ? "✅" : "❌", text_color=dayBullish or dayBearish ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 10, "HTF Aligned",    text_color=color.white, text_size=size.small)
table.cell(dash, 1, 10, htfStr,           text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.tiny)
table.cell(dash, 2, 10, htfBullConfirm or htfBearConfirm ? "✅" : "❌", text_color=htfBullConfirm or htfBearConfirm ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 11, "Session Gate",   text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 11, sessionName,      text_color=sessionColor, text_size=size.small)
table.cell(dash, 2, 11, activeSession ? "✅" : "❌", text_color=activeSession ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 12, "Extreme Bear",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 12, extremeBear ? "🚨 BLOCKED" : "✅ SAFE",   text_color=extremeBear ? color.red : color.green, text_size=size.small)
table.cell(dash, 2, 12, extremeBear ? "❌" : "✅", text_color=extremeBear ? color.red : color.green, text_size=size.small)

table.cell(dash, 0, 13, "── TIER 2 STRUCTURE ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 13, "────────────────────",   text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 13, "──────",                 text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 14, "EMA 200",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 14, bullishBias ? "🟢 BULLISH" : "🔴 BEARISH", text_color=bullishBias ? color.green : color.red, text_size=size.small)
table.cell(dash, 2, 14, "✅", text_color=color.green, text_size=size.small)

table.cell(dash, 0, 15, "BOS",            text_color=color.white, text_size=size.small)
table.cell(dash, 1, 15, recentBullBOS ? "🟢 BULL" : recentBearBOS ? "🔴 BEAR" : "⏳ NONE", text_color=recentBullBOS ? color.green : recentBearBOS ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 15, recentBullBOS or recentBearBOS ? "✅" : "❌", text_color=recentBullBOS or recentBearBOS ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 16, "Sweep",          text_color=color.white, text_size=size.small)
table.cell(dash, 1, 16, bullishSweep ? "🟢 BULL" : bearishSweep ? "🔴 BEAR" : "⏳ NONE", text_color=bullishSweep or bearishSweep ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 16, bullishSweep or bearishSweep ? "✅" : "❌", text_color=bullishSweep or bearishSweep ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 17, "Zone",           text_color=color.white, text_size=size.small)
table.cell(dash, 1, 17, inDiscount ? "🟢 DISCOUNT" : inPremium ? "🔴 PREMIUM" : "⚪ EQUILIBRIUM", text_color=inDiscount ? color.green : inPremium ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 17, inDiscount or inPremium ? "✅" : "➖", text_color=inDiscount or inPremium ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 18, "Monthly",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 18, monthBullAlign ? "🟢 BULL" : monthBearAlign ? "🔴 BEAR" : "⏳ WAIT", text_color=monthBullAlign ? color.green : monthBearAlign ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 18, monthBullAlign or monthBearAlign ? "✅" : "❌", text_color=monthBullAlign or monthBearAlign ? color.green : color.red, text_size=size.small)

str2Display = bearishBias ? str2Short : str2Long
table.cell(dash, 0, 19, "Structure Score", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 19, str.tostring(str2Display) + "/5  (need " + str.tostring(minStr2) + ")", text_color=str2Display >= minStr2 ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 19, str2Display >= minStr2 ? "✅" : "❌", text_color=str2Display >= minStr2 ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 20, "── TIER 3 TRIGGER ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 20, "────────────────────", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 20, "──────",               text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 21, "FVG Entry Zone",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 21, bullEntryZone or bearEntryZone ? "🎯 ACTIVE" : "⏳ NONE", text_color=bullEntryZone or bearEntryZone ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 21, bullEntryZone or bearEntryZone ? "✅" : "➖", text_color=bullEntryZone or bearEntryZone ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 22, "Order Block",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 22, bullOBRespected ? "✅ BULL" : bearOBRespected ? "✅ BEAR" : bullOBAlive ? "🟦 LIVE" : bearOBAlive ? "🟧 LIVE" : "⏳ NONE", text_color=bullOBRespected or bearOBRespected ? color.green : bullOBAlive or bearOBAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 22, bullOBRespected or bearOBRespected ? "✅" : bullOBAlive or bearOBAlive ? "⏳" : "➖", text_color=bullOBRespected or bearOBRespected ? color.green : color.yellow, text_size=size.small)

table.cell(dash, 0, 23, "Sweep Rejection", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 23, bullSweepRej ? "🟢 BULL" : bearSweepRej ? "🔴 BEAR" : "⏳ NONE", text_color=bullSweepRej or bearSweepRej ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 23, bullSweepRej or bearSweepRej ? "✅" : "➖", text_color=bullSweepRej or bearSweepRej ? color.green : color.gray, text_size=size.small)

trigActive = anyTriggerBull or anyTriggerBear
table.cell(dash, 0, 24, "Trigger Status",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 24, trigActive ? "🟢 TRIGGER ACTIVE" : "⏳ WAITING", text_color=trigActive ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 24, trigActive ? "✅" : "❌", text_color=trigActive ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 25, "── TIER 4 BONUS ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 25, "──────────────────", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 25, "──────",             text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 26, "RSI Div",         text_color=color.white, text_size=size.small)
table.cell(dash, 1, 26, strongBullDiv ? "🟢 Strong(2)" : hiddenBullDiv ? "🟢 Hidden(1)" : strongBearDiv ? "🔴 Strong(2)" : hiddenBearDiv ? "🔴 Hidden(1)" : "⏳ NONE", text_color=anyBullDiv ? color.green : anyBearDiv ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 26, anyBullDiv or anyBearDiv ? "✅" : "➖", text_color=anyBullDiv or anyBearDiv ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 27, "MACD",            text_color=color.white, text_size=size.small)
table.cell(dash, 1, 27, macdBullish ? "🟢 BULLISH" : "🔴 BEARISH", text_color=macdBullish ? color.green : color.red, text_size=size.small)
table.cell(dash, 2, 27, "✅", text_color=color.green, text_size=size.small)

table.cell(dash, 0, 28, "RSI",             text_color=color.white, text_size=size.small)
table.cell(dash, 1, 28, str.tostring(math.round(rsi,1)) + (rsiBullExt ? " 🟢 OVERSOLD" : rsiBearExt ? " 🔴 OVERBOUGHT" : " ⚪ NEUTRAL"), text_color=rsiBullExt ? color.green : rsiBearExt ? color.red : color.white, text_size=size.small)
table.cell(dash, 2, 28, rsiBullExt or rsiBearExt ? "✅" : "➖", text_color=rsiBullExt or rsiBearExt ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 29, "Int Sweep",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 29, internalBullSweep ? "🟢 BULL" : internalBearSweep ? "🔴 BEAR" : "⏳ NONE", text_color=internalBullSweep or internalBearSweep ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 29, internalBullSweep or internalBearSweep ? "✅" : "➖", text_color=internalBullSweep or internalBearSweep ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 30, "EQH/EQL Swept",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 30, eqhSwept ? "🔴 EQH SWEPT" : eqlSwept ? "🟢 EQL SWEPT" : "⏳ NONE", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 30, eqhSwept or eqlSwept ? "✅" : "➖", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)

bonusDisplay = bearishBias ? bonusShort : bonusLong
table.cell(dash, 0, 31, "Bonus Score",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 31, str.tostring(bonusDisplay) + "/7  (need " + str.tostring(minBonus) + ")", text_color=bonusDisplay >= minBonus ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 31, bonusDisplay >= minBonus ? "✅" : "❌", text_color=bonusDisplay >= minBonus ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 32, "── RISK ──",      text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 32, "────────────",    text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 32, "──────",          text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 33, "Risk/Target",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 33, "💰 $" + str.tostring(math.round(riskAmount,2)) + " → 🎯 $" + str.tostring(math.round(riskAmount*3,2)) + " (1:3)", text_color=color.yellow, text_size=size.small)
table.cell(dash, 2, 33, "RR 1:3",          text_color=color.yellow, text_size=size.small)

waitStr = bearishBias ? waitReasonShort : waitReasonLong
sigStr  = bullishEntry ? "⬆ BUY " + tfLabel + " — " + qualityLong : bearishEntry ? "⬇ SELL " + tfLabel + " — " + qualityShort : waitStr
sigCol  = bullishEntry ? color.green : bearishEntry ? color.red : color.yellow
sig3    = bullishEntry ? "🟢 BUY" : bearishEntry ? "🔴 SELL" : "⏳ WAIT"
sig3Col = bullishEntry ? color.green : bearishEntry ? color.red : color.yellow

table.cell(dash, 0, 34, "🚦 SIGNAL",       text_color=color.white, text_size=size.normal, bgcolor=color.new(color.gray,30))
table.cell(dash, 1, 34, sigStr,            text_color=sigCol,      text_size=size.small)
table.cell(dash, 2, 34, sig3,              text_color=sig3Col,     text_size=size.normal)
```

```python
// ============================================================
// Manny's Gold Strategy V2 — BACKTEST VERSION
// IMPORTANT: Chart timezone must be set to UTC
// ============================================================
//@version=6
strategy("Manny's Gold Strategy V2 — Backtest",
     overlay=true,
     max_lines_count=500,
     max_labels_count=500,
     max_boxes_count=500,
     initial_capital=1000,
     default_qty_type=strategy.fixed,
     default_qty_value=1,
     commission_type=strategy.commission.percent,
     commission_value=0.01,
     slippage=2)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
showMonthly   = input.bool(true,  "📅 Monthly High/Low Lines",          group="📊 Chart Levels")
showWeekly    = input.bool(true,  "📅 Weekly High/Low Lines",           group="📊 Chart Levels")
showDaily     = input.bool(true,  "📅 Daily High/Low Lines (3 Days)",   group="📊 Chart Levels")
showEQLevels  = input.bool(true,  "📊 Equal Highs and Lows",            group="📊 Chart Levels")
showWeekBias  = input.bool(true,  "⚡ Weekly Bias Lines (Yellow)",      group="🧭 Bias Levels")
showDayBias   = input.bool(true,  "🟢 Daily Bias Lines (Green)",        group="🧭 Bias Levels")
showEMA       = input.bool(true,  "📉 EMA 200 + EMA 50",                group="🔧 Features")
showSweeps    = input.bool(true,  "💧 Liquidity Sweeps",                group="🔧 Features")
showFVG       = input.bool(true,  "📦 Fair Value Gaps",                 group="🔧 Features")
showDiv       = input.bool(true,  "📐 RSI Divergence",                  group="🔧 Features")
showBOS       = input.bool(true,  "🔨 Break of Structure",              group="🔧 Features")
showOB        = input.bool(true,  "🟦 Order Blocks",                    group="🔧 Features")
showZone      = input.bool(true,  "📊 Premium/Discount Zone",           group="🔧 Features")
fvgMinSizeIn  = input.float(0.5,  "FVG Min Size (x ATR)",               group="🔧 Features", minval=0.1)
modeOverride  = input.string("Auto", "Trading Mode Override",
     options=["Auto", "Scalping", "Day Trade", "Swing Trade"],
     group="🎯 Signal Quality")
minBonusOver  = input.int(0,      "Min Bonus Score Override (0=auto)",  group="🎯 Signal Quality", minval=0, maxval=7)
adxLen        = input.int(14,     "ADX Period",                         group="📊 Market Filter")
adxThreshold  = input.int(25,     "ADX Minimum",                        group="📊 Market Filter")
accountSize   = input.float(1000, "💰 Account Size ($)",                group="💵 Risk Management")
riskPercent   = input.float(1.0,  "⚠️ Risk Per Trade (%)",             group="💵 Risk Management")
atrLength     = input.int(14,     "ATR Period",                         group="💵 Risk Management")
goldBiasTime  = input.string("1AM / 5AM (Standard Time)", "🥇 Gold Bias Candle Time",
     options=["1AM / 5AM (Standard Time)", "2AM / 6AM (Daylight Saving)"],
     group="🥇 Gold Settings")

// ═══════════════════════════════════════════
// AUTO TIMEFRAME DETECTION
// ═══════════════════════════════════════════
tfMins = timeframe.in_seconds() / 60

autoIsScalp  = tfMins <= 5
autoDayTrade = tfMins > 5  and tfMins <= 30
autoSwing    = tfMins > 30 and tfMins <= 240
autoLongTerm = tfMins > 240

autoModeStr = autoIsScalp ? "Scalping" : autoDayTrade ? "Day Trade" : autoSwing ? "Swing Trade" : "Long Term"

isScalp    = modeOverride == "Auto" ? autoIsScalp  : modeOverride == "Scalping"
isSwing    = modeOverride == "Auto" ? autoSwing    : modeOverride == "Swing Trade"
isDayTrade = modeOverride == "Auto" ? autoDayTrade : modeOverride == "Day Trade"

activeModeStr = modeOverride == "Auto" ? autoModeStr + " (Auto)" : modeOverride
tfLabel = tfMins == 1 ? "1M" : tfMins == 3 ? "3M" : tfMins == 5 ? "5M" : tfMins == 15 ? "15M" : tfMins == 30 ? "30M" : tfMins == 60 ? "1H" : tfMins == 240 ? "4H" : tfMins == 1440 ? "Daily" : tfMins == 10080 ? "Weekly" : str.tostring(math.round(tfMins)) + "M"

minBonus   = minBonusOver > 0 ? minBonusOver : isScalp ? 2 : isSwing ? 4 : 3
bosReset   = isScalp ? 50 : isSwing ? 20 : 30
fvgMinSize = isScalp ? 0.3 : isSwing ? 0.7 : fvgMinSizeIn
minStr2    = isScalp ? 2 : isSwing ? 3 : 2
sigCooldown = tfMins <= 1 ? 15 : tfMins <= 5 ? 12 : tfMins <= 15 ? 10 : tfMins <= 30 ? 8 : 5

// ═══════════════════════════════════════════
// GOLD BIAS TIME
// ═══════════════════════════════════════════
isStandard    = goldBiasTime == "1AM / 5AM (Standard Time)"
dailyBiasHour = isStandard ? 1 : 2
weekBiasHour1 = isStandard ? 1 : 2
weekBiasHour2 = isStandard ? 5 : 6
dailyBiasLbl  = isStandard ? "🕐 1AM 4H" : "🕑 2AM 4H"
weekBiasLbl   = isStandard ? "⚡ 1AM+5AM" : "⚡ 2AM+6AM"

// ═══════════════════════════════════════════
// TIME HELPERS
// ═══════════════════════════════════════════
gmtHour  = hour(time, "GMT+0")
gmtMin   = minute(time, "GMT+0")
gmtDow   = dayofweek(time, "GMT+0")
isMonday = gmtDow == 2
isNewDay   = ta.change(time("D")) != 0
isNewWeek  = ta.change(time("W")) != 0
isNewMonth = ta.change(time("M")) != 0

// ═══════════════════════════════════════════
// SESSION FILTER
// ═══════════════════════════════════════════
londonOpen    = gmtHour >= 7  and gmtHour < 12
nyOpen        = gmtHour >= 13 and gmtHour < 17
overlapSess   = gmtHour >= 12 and gmtHour < 13
activeSession = londonOpen or nyOpen or overlapSess

sessionColor = overlapSess ? color.yellow : londonOpen ? color.green : nyOpen ? color.blue : color.gray
sessionName  = overlapSess ? "⚡ Overlap — BEST" : londonOpen ? "🇬🇧 London — ACTIVE" : nyOpen ? "🇺🇸 New York — ACTIVE" : "😴 Asian — LOW VOLUME"

// ═══════════════════════════════════════════
// ATR + EMA
// ═══════════════════════════════════════════
atr         = ta.atr(atrLength)
ema200      = ta.ema(close, 200)
ema50       = ta.ema(close, 50)
bullishBias = close > ema200
bearishBias = close < ema200
bullMarket  = ema50 > ema200

pctBelowEMA = (ema200 - close) / ema200 * 100
extremeBear = pctBelowEMA > 20
allowLongs  = not extremeBear
allowShorts = not extremeBear

regimeStr = extremeBear ? "🚨 EXTREME BEAR" : bullMarket ? "🟢 BULL MARKET" : "🔴 BEAR MARKET"
regimeCol = extremeBear ? color.red : bullMarket ? color.green : color.red

ema200Col = bullishBias ? color.new(color.green, 0)  : color.new(color.red, 0)
ema50Col  = bullMarket  ? color.new(color.green, 50) : color.new(color.red, 50)
plot(showEMA ? ema200 : na, "EMA 200", color=ema200Col, linewidth=2)
plot(showEMA ? ema50  : na, "EMA 50",  color=ema50Col,  linewidth=1)

// ═══════════════════════════════════════════
// ADX
// ═══════════════════════════════════════════
[diPlus, diMinus, adx] = ta.dmi(adxLen, adxLen)
isTrending = adx >= adxThreshold
adxStr     = isTrending ? "📈 TRENDING (" + str.tostring(math.round(adx,1)) + ")" : "📉 RANGING (" + str.tostring(math.round(adx,1)) + ")"
adxCol     = isTrending ? color.green : color.red

// ═══════════════════════════════════════════
// HTF CONFIRMATION
// ═══════════════════════════════════════════
htf1HClose   = request.security(syminfo.tickerid, "60",  close,              lookahead=barmerge.lookahead_off)
htf1HEMA200  = request.security(syminfo.tickerid, "60",  ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htf4HClose   = request.security(syminfo.tickerid, "240", close,              lookahead=barmerge.lookahead_off)
htf4HEMA200  = request.security(syminfo.tickerid, "240", ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htfDClose    = request.security(syminfo.tickerid, "D",   close,              lookahead=barmerge.lookahead_off)
htfDEMA200   = request.security(syminfo.tickerid, "D",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htfWClose    = request.security(syminfo.tickerid, "W",   close,              lookahead=barmerge.lookahead_off)
htfWEMA200   = request.security(syminfo.tickerid, "W",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)

htf1HBull  = htf1HClose > htf1HEMA200
htf4HBull  = htf4HClose > htf4HEMA200
htfDBull   = htfDClose  > htfDEMA200
htfWBull   = htfWClose  > htfWEMA200
htf1HBear  = htf1HClose < htf1HEMA200
htf4HBear  = htf4HClose < htf4HEMA200
htfDBear   = htfDClose  < htfDEMA200
htfWBear   = htfWClose  < htfWEMA200

htfBullConfirm = autoIsScalp ? (htf1HBull and htf4HBull and htfDBull) : autoDayTrade ? (htf1HBull and htf4HBull) : autoSwing ? htf4HBull : (htfDBull and htfWBull)
htfBearConfirm = autoIsScalp ? (htf1HBear and htf4HBear and htfDBear) : autoDayTrade ? (htf1HBear and htf4HBear) : autoSwing ? htf4HBear : (htfDBear and htfWBear)

htfNeeded  = autoIsScalp ? "1H+4H+D" : autoDayTrade ? "1H+4H" : autoSwing ? "4H EMA" : "D+W"
htf1HE = htf1HBull ? "🟢" : "🔴"
htf4HE = htf4HBull ? "🟢" : "🔴"
htfDE  = htfDBull  ? "🟢" : "🔴"
htfWE  = htfWBull  ? "🟢" : "🔴"
htfStrBull = "🟢 " + htfNeeded + " BULL ✓"
htfStrBear = "🔴 " + htfNeeded + " BEAR ✓"
htfStrWait = "⏳ 1H:" + htf1HE + " 4H:" + htf4HE + " D:" + htfDE + " W:" + htfWE
htfStr     = htfBullConfirm ? htfStrBull : htfBearConfirm ? htfStrBear : htfStrWait

// ═══════════════════════════════════════════
// MONTHLY LEVELS
// ═══════════════════════════════════════════
month1High = ta.valuewhen(isNewMonth, high[1], 0)
month1Low  = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, month1High, bar_index+100, month1High, color=color.new(color.purple,0), width=2)
    mLL := line.new(bar_index, month1Low,  bar_index+100, month1Low,  color=color.new(color.purple,0), width=2)
    label.new(bar_index, month1High, "M High", color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, month1Low,  "M Low",  color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)

monthBullAlign = close > month1Low  and bullishBias
monthBearAlign = close < month1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY LEVELS
// ═══════════════════════════════════════════
week1High = ta.valuewhen(isNewWeek, high[1], 0)
week1Low  = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, week1High, bar_index+100, week1High, color=color.new(color.blue,0), width=2)
    wLL := line.new(bar_index, week1Low,  bar_index+100, week1Low,  color=color.new(color.blue,0), width=2)
    label.new(bar_index, week1High, "W High", color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, week1Low,  "W Low",  color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY LEVELS
// ═══════════════════════════════════════════
day1High = ta.valuewhen(isNewDay, high[1], 0)
day1Low  = ta.valuewhen(isNewDay, low[1],  0)
day2High = ta.valuewhen(isNewDay, high[1], 1)
day2Low  = ta.valuewhen(isNewDay, low[1],  1)
day3High = ta.valuewhen(isNewDay, high[1], 2)
day3Low  = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, day1High, bar_index+100, day1High, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, day1Low,  bar_index+100, day1Low,  color=color.new(color.black,0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, day2High, bar_index+100, day2High, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, day2Low,  bar_index+100, day2Low,  color=color.new(color.black,40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, day3High, bar_index+100, day3High, color=color.new(color.black,70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, day3Low,  bar_index+100, day3Low,  color=color.new(color.black,70), width=1, style=line.style_dotted)
    label.new(bar_index, day1High, "D1H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day1Low,  "D1L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// EQUAL HIGHS AND LOWS
// ═══════════════════════════════════════════
swingH     = ta.pivothigh(high, 5, 5)
swingL     = ta.pivotlow(low,   5, 5)
prevSwingH = ta.valuewhen(not na(swingH), swingH, 1)
prevSwingL = ta.valuewhen(not na(swingL), swingL, 1)
isEQH      = not na(swingH) and not na(prevSwingH) and math.abs(swingH - prevSwingH) <= atr * 0.1
isEQL      = not na(swingL) and not na(prevSwingL) and math.abs(swingL - prevSwingL) <= atr * 0.1

if showEQLevels and isEQH
    line.new(bar_index-5, swingH, bar_index+20, swingH, color=color.new(color.red,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, swingH, "EQH", color=color.new(color.red,70), textcolor=color.red, style=label.style_label_right, size=size.tiny)
if showEQLevels and isEQL
    line.new(bar_index-5, swingL, bar_index+20, swingL, color=color.new(color.green,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, swingL, "EQL", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_right, size=size.tiny)

eqhSwept = not na(swingH) and high > swingH and close < swingH
eqlSwept = not na(swingL) and low  < swingL and close > swingL

// ═══════════════════════════════════════════
// GATE 1 — WEEKLY BIAS
// ═══════════════════════════════════════════
weekBias4HC1High = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour1 and minute(time, "GMT+0") == 0, high, 0), lookahead=barmerge.lookahead_off)
weekBias4HC1Low  = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour1 and minute(time, "GMT+0") == 0, low,  0), lookahead=barmerge.lookahead_off)
weekBias4HC2High = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour2 and minute(time, "GMT+0") == 0, high, 0), lookahead=barmerge.lookahead_off)
weekBias4HC2Low  = request.security(syminfo.tickerid, "240", ta.valuewhen(dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == weekBiasHour2 and minute(time, "GMT+0") == 0, low,  0), lookahead=barmerge.lookahead_off)

wbRangeHigh = not na(weekBias4HC1High) and not na(weekBias4HC2High) ? math.max(weekBias4HC1High, weekBias4HC2High) : not na(weekBias4HC1High) ? weekBias4HC1High : weekBias4HC2High
wbRangeLow  = not na(weekBias4HC1Low)  and not na(weekBias4HC2Low)  ? math.min(weekBias4HC1Low,  weekBias4HC2Low)  : not na(weekBias4HC1Low)  ? weekBias4HC1Low  : weekBias4HC2Low

var float weekBiasHigh = na
var float weekBiasLow  = na
var line  wbHL         = na
var line  wbLL         = na
var bool  weekBullish  = false
var bool  weekBearish  = false

if isNewWeek
    weekBullish  := false
    weekBearish  := false
    weekBiasHigh := na
    weekBiasLow  := na

if not na(wbRangeHigh) and not na(wbRangeLow)
    if wbRangeHigh != weekBiasHigh or wbRangeLow != weekBiasLow
        weekBiasHigh := wbRangeHigh
        weekBiasLow  := wbRangeLow
        if showWeekBias
            line.delete(wbHL)
            line.delete(wbLL)
            wbHL := line.new(bar_index, weekBiasHigh, bar_index+500, weekBiasHigh, color=color.new(color.yellow,0), width=2)
            wbLL := line.new(bar_index, weekBiasLow,  bar_index+500, weekBiasLow,  color=color.new(color.yellow,0), width=2)
            label.new(bar_index, weekBiasHigh, weekBiasLbl + " High", color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
            label.new(bar_index, weekBiasLow,  weekBiasLbl + " Low",  color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

if not na(weekBiasHigh) and not na(weekBiasLow)
    if close > weekBiasHigh
        weekBullish := true
        weekBearish := false
    else if close < weekBiasLow
        weekBearish := true
        weekBullish := false

// ═══════════════════════════════════════════
// GATE 2 — DAILY BIAS
// ═══════════════════════════════════════════
dailyBias4HHigh = request.security(syminfo.tickerid, "240", ta.valuewhen(hour(time, "GMT+0") == dailyBiasHour and minute(time, "GMT+0") == 0, high, 0), lookahead=barmerge.lookahead_off)
dailyBias4HLow  = request.security(syminfo.tickerid, "240", ta.valuewhen(hour(time, "GMT+0") == dailyBiasHour and minute(time, "GMT+0") == 0, low,  0), lookahead=barmerge.lookahead_off)

var float dayBiasHigh = na
var float dayBiasLow  = na
var line  dbHL        = na
var line  dbLL        = na
var bool  dayBullish  = false
var bool  dayBearish  = false

if isNewDay
    dayBullish := false
    dayBearish := false

if not na(dailyBias4HHigh) and not na(dailyBias4HLow)
    if dailyBias4HHigh != dayBiasHigh or dailyBias4HLow != dayBiasLow
        dayBiasHigh := dailyBias4HHigh
        dayBiasLow  := dailyBias4HLow
        if showDayBias
            line.delete(dbHL)
            line.delete(dbLL)
            dbHL := line.new(bar_index, dayBiasHigh, bar_index+100, dayBiasHigh, color=color.new(color.green,0), width=2, style=line.style_dashed)
            dbLL := line.new(bar_index, dayBiasLow,  bar_index+100, dayBiasLow,  color=color.new(color.green,0), width=2, style=line.style_dashed)
            label.new(bar_index, dayBiasHigh, dailyBiasLbl + " H", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
            label.new(bar_index, dayBiasLow,  dailyBiasLbl + " L", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

if not na(dayBiasHigh) and not na(dayBiasLow)
    if close > dayBiasHigh
        dayBullish := true
        dayBearish := false
    else if close < dayBiasLow
        dayBearish := true
        dayBullish := false

// ═══════════════════════════════════════════
// PREMIUM AND DISCOUNT ZONE
// ═══════════════════════════════════════════
weekRange  = week1High - week1Low
weekMid    = week1Low + weekRange / 2
inDiscount = close < weekMid and not na(weekMid)
inPremium  = close > weekMid and not na(weekMid)

if showZone and isNewWeek and not na(week1High) and not na(week1Low)
    line.new(bar_index, weekMid, bar_index+100, weekMid, color=color.new(color.gray,50), style=line.style_dashed, width=1)
    label.new(bar_index, weekMid, "Equilibrium", color=color.new(color.gray,70), textcolor=color.gray, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swingLen      = 5
swingHigh     = ta.highest(high, swingLen * 2 + 1)[swingLen]
swingLow      = ta.lowest(low,   swingLen * 2 + 1)[swingLen]
strongBOSBody = math.abs(close - open) > atr * 0.7
bullBOS       = close > swingHigh and strongBOSBody and bullishBias and close[1] <= swingHigh
bearBOS       = close < swingLow  and strongBOSBody and bearishBias and close[1] >= swingLow

var bool recentBullBOS = false
var bool recentBearBOS = false
var int  bosTimer      = 0

if bullBOS
    recentBullBOS := true
    recentBearBOS := false
    bosTimer      := bar_index
if bearBOS
    recentBearBOS := true
    recentBullBOS := false
    bosTimer      := bar_index
if bar_index - bosTimer > bosReset
    recentBullBOS := false
    recentBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low - atr*0.5, "🔨 BOS ↑", color=color.new(color.green,20), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index-1, swingHigh, bar_index+10, swingHigh, color=color.new(color.green,40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr*0.5, "🔨 BOS ↓", color=color.new(color.red,20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index-1, swingLow, bar_index+10, swingLow, color=color.new(color.red,40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweepWick = day1Low - low
bearSweepWick = high - day1High
bullishSweep  = low < day1Low   and close > day1Low  and bullSweepWick >= atr * 0.3
bearishSweep  = high > day1High and close < day1High and bearSweepWick >= atr * 0.3

var float lastBullSweepLow  = na
var float lastBearSweepHigh = na
if bullishSweep
    lastBullSweepLow  := low
if bearishSweep
    lastBearSweepHigh := high

bullSweepRej = bullishSweep and (close - low)  > atr * 1.5
bearSweepRej = bearishSweep and (high - close) > atr * 1.5

plotshape(showSweeps and bullishSweep, "💧 Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime,0), size=size.small, text="Sweep")
plotshape(showSweeps and bearishSweep, "💧 Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,0),  size=size.small, text="Sweep")

intBullSweep      = low  < ta.lowest(low[1],   3)[1] and close > ta.lowest(low[1],   3)[1]
intBearSweep      = high > ta.highest(high[1], 3)[1] and close < ta.highest(high[1], 3)[1]
internalBullSweep = intBullSweep and bullishBias
internalBearSweep = intBearSweep and bearishBias

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
impulseLen   = 3
bullImpulse  = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImpulse  = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBCandle = bullImpulse and close[impulseLen] < open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5
bearOBCandle = bearImpulse and close[impulseLen] > open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5

var float bullOBTop       = na
var float bullOBBot       = na
var float bearOBTop       = na
var float bearOBBot       = na
var bool  bullOBAlive     = false
var bool  bearOBAlive     = false
var box   bullOBBox       = na
var box   bearOBBox       = na
var bool  bullOBRespected = false
var bool  bearOBRespected = false
var int   bullOBTouches   = 0
var int   bearOBTouches   = 0

if bullOBCandle and bullishBias and showOB
    bullOBTop       := high[impulseLen]
    bullOBBot       := low[impulseLen]
    bullOBAlive     := true
    bullOBRespected := false
    bullOBTouches   := 0
    box.delete(bullOBBox)
    bullOBBox := box.new(bar_index-impulseLen, bullOBTop, bar_index+100, bullOBBot, bgcolor=color.new(color.blue,80), border_color=color.new(color.blue,30), border_width=1)
    label.new(bar_index-impulseLen, bullOBTop, "🟦 Bull OB", color=color.new(color.blue,60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBCandle and bearishBias and showOB
    bearOBTop       := high[impulseLen]
    bearOBBot       := low[impulseLen]
    bearOBAlive     := true
    bearOBRespected := false
    bearOBTouches   := 0
    box.delete(bearOBBox)
    bearOBBox := box.new(bar_index-impulseLen, bearOBTop, bar_index+100, bearOBBot, bgcolor=color.new(color.orange,80), border_color=color.new(color.orange,30), border_width=1)
    label.new(bar_index-impulseLen, bearOBBot, "🟧 Bear OB", color=color.new(color.orange,60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bullOBAlive and bullOBTouches == 0
    if low <= bullOBTop and close >= bullOBBot
        bullOBRespected := true
        bullOBTouches   += 1
        box.set_bgcolor(bullOBBox, color.new(color.green,70))
        box.set_border_color(bullOBBox, color.new(color.green,20))

if bearOBAlive and bearOBTouches == 0
    if high >= bearOBBot and close <= bearOBTop
        bearOBRespected := true
        bearOBTouches   += 1
        box.set_bgcolor(bearOBBox, color.new(color.red,70))
        box.set_border_color(bearOBBox, color.new(color.red,20))

if bullOBAlive and close < bullOBBot
    bullOBAlive     := false
    bullOBRespected := false
    box.set_bgcolor(bullOBBox, color.new(color.gray,90))

if bearOBAlive and close > bearOBTop
    bearOBAlive     := false
    bearOBRespected := false
    box.set_bgcolor(bearOBBox, color.new(color.gray,90))

// ═══════════════════════════════════════════
// FVG
// ═══════════════════════════════════════════
candleBody   = math.abs(close[1] - open[1])
strongCandle = candleBody > atr * 1.0
dispCandle   = math.abs(close - open) > atr * 1.5
bullFVGSize  = low - high[2]
bearFVGSize  = low[2] - high
bullFVGraw   = low > high[2] and strongCandle and close[1] > open[1] and bullishBias and bullFVGSize >= atr * fvgMinSize
bearFVGraw   = high < low[2] and strongCandle and close[1] < open[1] and bearishBias and bearFVGSize >= atr * fvgMinSize

var float bullFVGTop       = na
var float bullFVGBot       = na
var float bearFVGTop       = na
var float bearFVGBot       = na
var bool  bullFVGAlive     = false
var bool  bearFVGAlive     = false
var box   bullFVGBox       = na
var box   bearFVGBox       = na
var bool  bullEntryZone    = false
var bool  bearEntryZone    = false

if bullFVGraw and showFVG
    bullFVGTop    := low
    bullFVGBot    := high[2]
    bullFVGAlive  := true
    bullEntryZone := false
    box.delete(bullFVGBox)
    bullFVGBox := box.new(bar_index-2, bullFVGTop, bar_index+100, bullFVGBot, bgcolor=color.new(color.green,85), border_color=color.new(color.green,40), border_width=1)

if bearFVGraw and showFVG
    bearFVGTop    := low[2]
    bearFVGBot    := high
    bearFVGAlive  := true
    bearEntryZone := false
    box.delete(bearFVGBox)
    bearFVGBox := box.new(bar_index-2, bearFVGTop, bar_index+100, bearFVGBot, bgcolor=color.new(color.red,85), border_color=color.new(color.red,40), border_width=1)

if bullFVGAlive and low <= bullFVGTop and close >= bullFVGBot
    box.set_bgcolor(bullFVGBox, color.new(color.green,70))
    if dispCandle and bullishBias and weekBullish and dayBullish and not bullEntryZone
        bullEntryZone := true

if bearFVGAlive and high >= bearFVGBot and close <= bearFVGTop
    box.set_bgcolor(bearFVGBox, color.new(color.red,70))
    if dispCandle and bearishBias and weekBearish and dayBearish and not bearEntryZone
        bearEntryZone := true

if bullFVGAlive and close < bullFVGBot
    bullFVGAlive  := false
    bullEntryZone := false
    box.set_bgcolor(bullFVGBox, color.new(color.gray,90))

if bearFVGAlive and close > bearFVGTop
    bearFVGAlive  := false
    bearEntryZone := false
    box.set_bgcolor(bearFVGBox, color.new(color.gray,90))

// ═══════════════════════════════════════════
// RSI DIVERGENCE
// ═══════════════════════════════════════════
rsi = ta.rsi(close, 14)
lkb = 7
priceLL = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
priceHL = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
priceHH = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
priceLH = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)
rsiHL   = ta.lowest(rsi,  lkb) > ta.lowest(rsi[lkb],  lkb)
rsiLL   = ta.lowest(rsi,  lkb) < ta.lowest(rsi[lkb],  lkb)
rsiLH   = ta.highest(rsi, lkb) < ta.highest(rsi[lkb], lkb)
rsiHH   = ta.highest(rsi, lkb) > ta.highest(rsi[lkb], lkb)

rsiBullExt    = rsi < 40
rsiBearExt    = rsi > 60
strongBullDiv = priceLL and rsiHL
hiddenBullDiv = priceHL and rsiLL
strongBearDiv = priceHH and rsiLH
hiddenBearDiv = priceLH and rsiHH
anyBullDiv    = strongBullDiv or hiddenBullDiv
anyBearDiv    = strongBearDiv or hiddenBearDiv

if showDiv and strongBullDiv
    label.new(bar_index, low - atr*0.3, "◆ Strong Bull Div", color=color.new(color.green,0), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showDiv and hiddenBullDiv
    label.new(bar_index, low - atr*0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,0), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showDiv and strongBearDiv
    label.new(bar_index, high + atr*0.3, "◆ Strong Bear Div", color=color.new(color.red,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hiddenBearDiv
    label.new(bar_index, high + atr*0.3, "◇ Hidden Bear Div", color=color.new(color.orange,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdLine, signalLine, histLine] = ta.macd(close, 12, 26, 9)
macdBullish = macdLine > signalLine
macdBearish = macdLine < signalLine

// ═══════════════════════════════════════════
// TIER 2 STRUCTURE
// ═══════════════════════════════════════════
str2Long = 0
str2Long := str2Long + (bullishBias    ? 1 : 0)
str2Long := str2Long + (recentBullBOS  ? 1 : 0)
str2Long := str2Long + (bullishSweep   ? 1 : 0)
str2Long := str2Long + (inDiscount     ? 1 : 0)
str2Long := str2Long + (monthBullAlign ? 1 : 0)

str2Short = 0
str2Short := str2Short + (bearishBias    ? 1 : 0)
str2Short := str2Short + (recentBearBOS  ? 1 : 0)
str2Short := str2Short + (bearishSweep   ? 1 : 0)
str2Short := str2Short + (inPremium      ? 1 : 0)
str2Short := str2Short + (monthBearAlign ? 1 : 0)

// ═══════════════════════════════════════════
// TIER 3 TRIGGERS
// ═══════════════════════════════════════════
trigger1Bull   = bullEntryZone
trigger2Bull   = bullOBRespected
trigger3Bull   = bullSweepRej and weekBullish and dayBullish
anyTriggerBull = trigger1Bull or trigger2Bull or trigger3Bull

trigger1Bear   = bearEntryZone
trigger2Bear   = bearOBRespected
trigger3Bear   = bearSweepRej and weekBearish and dayBearish
anyTriggerBear = trigger1Bear or trigger2Bear or trigger3Bear

// ═══════════════════════════════════════════
// TIER 4 BONUS
// ═══════════════════════════════════════════
bonusLong = 0
bonusLong := bonusLong + (strongBullDiv     ? 2 : hiddenBullDiv ? 1 : 0)
bonusLong := bonusLong + (macdBullish       ? 1 : 0)
bonusLong := bonusLong + (rsiBullExt        ? 1 : 0)
bonusLong := bonusLong + (isTrending        ? 1 : 0)
bonusLong := bonusLong + (internalBullSweep ? 1 : 0)
bonusLong := bonusLong + (eqlSwept          ? 1 : 0)

bonusShort = 0
bonusShort := bonusShort + (strongBearDiv     ? 2 : hiddenBearDiv ? 1 : 0)
bonusShort := bonusShort + (macdBearish       ? 1 : 0)
bonusShort := bonusShort + (rsiBearExt        ? 1 : 0)
bonusShort := bonusShort + (isTrending        ? 1 : 0)
bonusShort := bonusShort + (internalBearSweep ? 1 : 0)
bonusShort := bonusShort + (eqhSwept          ? 1 : 0)

// ═══════════════════════════════════════════
// SIGNAL LOGIC
// ═══════════════════════════════════════════
gate1L = weekBullish
gate2L = dayBullish
gate3L = htfBullConfirm
gate4L = activeSession
gate5L = allowLongs
allGatesLong = gate1L and gate2L and gate3L and gate4L and gate5L

gate1S = weekBearish
gate2S = dayBearish
gate3S = htfBearConfirm
gate4S = activeSession
gate5S = allowShorts
allGatesShort = gate1S and gate2S and gate3S and gate4S and gate5S

var int lastBullBar = 0
var int lastBearBar = 0

rawBullEntry = allGatesLong  and str2Long  >= minStr2 and anyTriggerBull and bonusLong  >= minBonus
rawBearEntry = allGatesShort and str2Short >= minStr2 and anyTriggerBear and bonusShort >= minBonus

bullishEntry = rawBullEntry and (bar_index - lastBullBar) > sigCooldown and strategy.position_size == 0
bearishEntry = rawBearEntry and (bar_index - lastBearBar) > sigCooldown and strategy.position_size == 0

if bullishEntry
    lastBullBar := bar_index
if bearishEntry
    lastBearBar := bar_index

// ═══════════════════════════════════════════
// RISK MANAGEMENT
// ═══════════════════════════════════════════
riskAmount = accountSize * (riskPercent / 100)
slBufferL  = bonusLong  >= 6 ? atr * 0.2 : atr * 0.5
slBufferS  = bonusShort >= 6 ? atr * 0.2 : atr * 0.5

longSL    = not na(lastBullSweepLow)  ? lastBullSweepLow  - slBufferL : close - atr * 1.5
longRisk  = math.max(close - longSL,  0.0001)
longTP    = close + longRisk * 3
longSize  = riskAmount / longRisk

shortSL   = not na(lastBearSweepHigh) ? lastBearSweepHigh + slBufferS : close + atr * 1.5
shortRisk = math.max(shortSL - close, 0.0001)
shortTP   = close - shortRisk * 3
shortSize = riskAmount / shortRisk

// ═══════════════════════════════════════════
// STRATEGY ENTRIES AND EXITS
// This is what makes it a backtest —
// entries and exits are tracked automatically
// ═══════════════════════════════════════════
if bullishEntry
    strategy.entry("BUY " + tfLabel, strategy.long, qty=longSize, comment="⬆ BUY " + tfLabel)
    strategy.exit("BUY Exit", "BUY " + tfLabel, stop=longSL, limit=longTP)

if bearishEntry
    strategy.entry("SELL " + tfLabel, strategy.short, qty=shortSize, comment="⬇ SELL " + tfLabel)
    strategy.exit("SELL Exit", "SELL " + tfLabel, stop=shortSL, limit=shortTP)

// Visual entry markers
if bullishEntry
    label.new(bar_index, low - atr, "⬆ BUY\n" + tfLabel + "\nSL: " + str.tostring(math.round(longSL,2)) + "\nTP: " + str.tostring(math.round(longTP,2)), color=color.new(color.green,10), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index, longSL, bar_index+40, longSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP, bar_index+40, longTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, longSL,  "🔴 SL", color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, longTP,  "🟢 TP", color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if bearishEntry
    label.new(bar_index, high + atr, "⬇ SELL\n" + tfLabel + "\nSL: " + str.tostring(math.round(shortSL,2)) + "\nTP: " + str.tostring(math.round(shortTP,2)), color=color.new(color.red,10), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index+40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index+40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, shortSL, "🔴 SL", color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, shortTP, "🟢 TP", color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)
```

```python
// ============================================================
// Manny's Gold Strategy V2
// Chart timezone must be set to UTC
// ============================================================
//@version=6
indicator("Manny's Gold Strategy V2", overlay=true, max_lines_count=500, max_labels_count=500, max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
showMonthly   = input.bool(true,  "📅 Monthly High/Low Lines",         group="📊 Chart Levels")
showWeekly    = input.bool(true,  "📅 Weekly High/Low Lines",          group="📊 Chart Levels")
showDaily     = input.bool(true,  "📅 Daily High/Low Lines (3 Days)",  group="📊 Chart Levels")
showEQLevels  = input.bool(true,  "📊 Equal Highs and Lows",           group="📊 Chart Levels")
showWeekBias  = input.bool(true,  "⚡ Weekly Bias Lines (Yellow)",     group="🧭 Bias Levels")
showDayBias   = input.bool(true,  "🟢 Daily Bias Lines (Green)",       group="🧭 Bias Levels")
requireBias   = input.bool(false, "🔒 Require Bias Confirmation",      group="🧭 Bias Levels", tooltip="OFF = signals fire without waiting for bias lines. ON = bias must be confirmed first.")
showEMA       = input.bool(true,  "📉 EMA 200 + EMA 50",               group="🔧 Features")
showSweeps    = input.bool(true,  "💧 Liquidity Sweeps",               group="🔧 Features")
showFVG       = input.bool(true,  "📦 Fair Value Gaps",                group="🔧 Features")
showDiv       = input.bool(true,  "📐 RSI Divergence",                 group="🔧 Features")
showBOS       = input.bool(true,  "🔨 Break of Structure",             group="🔧 Features")
showOB        = input.bool(true,  "🟦 Order Blocks",                   group="🔧 Features")
showSignals   = input.bool(true,  "🚦 Buy/Sell Signals",               group="🔧 Features")
showZone      = input.bool(true,  "📊 Premium/Discount Zone",          group="🔧 Features")
fvgMinSizeIn  = input.float(0.5,  "FVG Min Size (x ATR)",              group="🔧 Features", minval=0.1)
modeOverride  = input.string("Auto", "Trading Mode Override",
     options=["Auto", "Scalping", "Day Trade", "Swing Trade"],
     group="🎯 Signal Quality")
minBonusOver  = input.int(0,      "Min Bonus Score Override (0=auto)", group="🎯 Signal Quality", minval=0, maxval=7)
adxLen        = input.int(14,     "ADX Period",                        group="📊 Market Filter")
adxThreshold  = input.int(25,     "ADX Minimum",                       group="📊 Market Filter")
accountSize   = input.float(1000, "💰 Account Size ($)",               group="💵 Risk Management")
riskPercent   = input.float(1.0,  "⚠️ Risk Per Trade (%)",            group="💵 Risk Management")
atrLength     = input.int(14,     "ATR Period",                        group="💵 Risk Management")
goldBiasTime  = input.string("1AM / 5AM (Standard Time)", "🥇 Gold Bias Candle Time",
     options=["1AM / 5AM (Standard Time)", "2AM / 6AM (Daylight Saving)"],
     group="🥇 Gold Settings")

// ═══════════════════════════════════════════
// AUTO TIMEFRAME DETECTION
// ═══════════════════════════════════════════
tfMins        = timeframe.in_seconds() / 60
autoIsScalp   = tfMins <= 5
autoDayTrade  = tfMins > 5  and tfMins <= 30
autoSwing     = tfMins > 30 and tfMins <= 240
autoLongTerm  = tfMins > 240
autoModeStr   = autoIsScalp ? "Scalping" : autoDayTrade ? "Day Trade" : autoSwing ? "Swing Trade" : "Long Term"
isScalp       = modeOverride == "Auto" ? autoIsScalp  : modeOverride == "Scalping"
isSwing       = modeOverride == "Auto" ? autoSwing    : modeOverride == "Swing Trade"
isDayTrade    = modeOverride == "Auto" ? autoDayTrade : modeOverride == "Day Trade"
activeModeStr = modeOverride == "Auto" ? autoModeStr + " (Auto)" : modeOverride
tfLabel       = tfMins == 1 ? "1M" : tfMins == 3 ? "3M" : tfMins == 5 ? "5M" : tfMins == 15 ? "15M" : tfMins == 30 ? "30M" : tfMins == 60 ? "1H" : tfMins == 240 ? "4H" : tfMins == 1440 ? "Daily" : tfMins == 10080 ? "Weekly" : str.tostring(math.round(tfMins)) + "M"
minBonus      = minBonusOver > 0 ? minBonusOver : isScalp ? 2 : isSwing ? 4 : 3
bosReset      = isScalp ? 50 : isSwing ? 20 : 30
fvgMinSize    = isScalp ? 0.3 : isSwing ? 0.7 : fvgMinSizeIn
minStr2       = isScalp ? 2 : isSwing ? 3 : 2
sigCooldown   = tfMins <= 1 ? 15 : tfMins <= 5 ? 12 : tfMins <= 15 ? 10 : tfMins <= 30 ? 8 : 5

// ═══════════════════════════════════════════
// GOLD BIAS TIME
// ═══════════════════════════════════════════
isStandard    = goldBiasTime == "1AM / 5AM (Standard Time)"
dailyBiasHour = isStandard ? 1 : 2
weekBiasHour1 = isStandard ? 1 : 2
weekBiasHour2 = isStandard ? 5 : 6
dailyBiasLbl  = isStandard ? "🕐 1AM 4H" : "🕑 2AM 4H"
weekBiasLbl   = isStandard ? "⚡ 1AM+5AM" : "⚡ 2AM+6AM"

// ═══════════════════════════════════════════
// TIME HELPERS
// ═══════════════════════════════════════════
gmtHour    = hour(time, "GMT+0")
gmtMin     = minute(time, "GMT+0")
gmtDow     = dayofweek(time, "GMT+0")
isNewDay   = ta.change(time("D")) != 0
isNewWeek  = ta.change(time("W")) != 0
isNewMonth = ta.change(time("M")) != 0

// ═══════════════════════════════════════════
// SESSION FILTER
// ═══════════════════════════════════════════
londonOpen    = gmtHour >= 7  and gmtHour < 12
nyOpen        = gmtHour >= 13 and gmtHour < 17
overlapSess   = gmtHour >= 12 and gmtHour < 13
activeSession = londonOpen or nyOpen or overlapSess
sessionColor  = overlapSess ? color.yellow : londonOpen ? color.green : nyOpen ? color.blue : color.gray
sessionName   = overlapSess ? "⚡ Overlap — BEST" : londonOpen ? "🇬🇧 London — ACTIVE" : nyOpen ? "🇺🇸 New York — ACTIVE" : "😴 Asian — LOW VOLUME"

// ═══════════════════════════════════════════
// ATR + EMA
// ═══════════════════════════════════════════
atr         = ta.atr(atrLength)
ema200      = ta.ema(close, 200)
ema50       = ta.ema(close, 50)
bullishBias = close > ema200
bearishBias = close < ema200
bullMarket  = ema50 > ema200
pctBelowEMA = (ema200 - close) / ema200 * 100
extremeBear = pctBelowEMA > 20
allowLongs  = not extremeBear
allowShorts = not extremeBear
regimeStr   = extremeBear ? "🚨 EXTREME BEAR" : bullMarket ? "🟢 BULL MARKET" : "🔴 BEAR MARKET"
regimeCol   = extremeBear ? color.red : bullMarket ? color.green : color.red
ema200Col   = bullishBias ? color.new(color.green, 0) : color.new(color.red, 0)
ema50Col    = bullMarket  ? color.new(color.green, 50) : color.new(color.red, 50)
plot(showEMA ? ema200 : na, "EMA 200", color=ema200Col, linewidth=2)
plot(showEMA ? ema50  : na, "EMA 50",  color=ema50Col,  linewidth=1)

// ═══════════════════════════════════════════
// ADX
// ═══════════════════════════════════════════
[diPlus, diMinus, adx] = ta.dmi(adxLen, adxLen)
isTrending = adx >= adxThreshold
adxStr     = isTrending ? "📈 TRENDING (" + str.tostring(math.round(adx,1)) + ")" : "📉 RANGING (" + str.tostring(math.round(adx,1)) + ")"
adxCol     = isTrending ? color.green : color.red

// ═══════════════════════════════════════════
// HTF CONFIRMATION
// ═══════════════════════════════════════════
htf1HClose  = request.security(syminfo.tickerid, "60",  close,              lookahead=barmerge.lookahead_off)
htf1HEMA200 = request.security(syminfo.tickerid, "60",  ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htf4HClose  = request.security(syminfo.tickerid, "240", close,              lookahead=barmerge.lookahead_off)
htf4HEMA200 = request.security(syminfo.tickerid, "240", ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htfDClose   = request.security(syminfo.tickerid, "D",   close,              lookahead=barmerge.lookahead_off)
htfDEMA200  = request.security(syminfo.tickerid, "D",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)
htfWClose   = request.security(syminfo.tickerid, "W",   close,              lookahead=barmerge.lookahead_off)
htfWEMA200  = request.security(syminfo.tickerid, "W",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)

htf1HBull = htf1HClose > htf1HEMA200
htf4HBull = htf4HClose > htf4HEMA200
htfDBull  = htfDClose  > htfDEMA200
htfWBull  = htfWClose  > htfWEMA200
htf1HBear = htf1HClose < htf1HEMA200
htf4HBear = htf4HClose < htf4HEMA200
htfDBear  = htfDClose  < htfDEMA200
htfWBear  = htfWClose  < htfWEMA200

htfBullConfirm = autoIsScalp ? (htf1HBull and htf4HBull and htfDBull) : autoDayTrade ? (htf1HBull and htf4HBull) : autoSwing ? htf4HBull : (htfDBull and htfWBull)
htfBearConfirm = autoIsScalp ? (htf1HBear and htf4HBear and htfDBear) : autoDayTrade ? (htf1HBear and htf4HBear) : autoSwing ? htf4HBear : (htfDBear and htfWBear)
htfNeeded      = autoIsScalp ? "1H+4H+D" : autoDayTrade ? "1H+4H" : autoSwing ? "4H EMA" : "D+W"
htf1HE = htf1HBull ? "🟢" : "🔴"
htf4HE = htf4HBull ? "🟢" : "🔴"
htfDE  = htfDBull  ? "🟢" : "🔴"
htfWE  = htfWBull  ? "🟢" : "🔴"
htfStrBull = "🟢 " + htfNeeded + " BULL ✓"
htfStrBear = "🔴 " + htfNeeded + " BEAR ✓"
htfStrWait = "⏳ 1H:" + htf1HE + " 4H:" + htf4HE + " D:" + htfDE + " W:" + htfWE
htfStr     = htfBullConfirm ? htfStrBull : htfBearConfirm ? htfStrBear : htfStrWait

// ═══════════════════════════════════════════
// MONTHLY LEVELS
// ═══════════════════════════════════════════
month1High = ta.valuewhen(isNewMonth, high[1], 0)
month1Low  = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, month1High, bar_index+100, month1High, color=color.new(color.purple,0), width=2)
    mLL := line.new(bar_index, month1Low,  bar_index+100, month1Low,  color=color.new(color.purple,0), width=2)
    label.new(bar_index, month1High, "M High", color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, month1Low,  "M Low",  color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
monthBullAlign = close > month1Low  and bullishBias
monthBearAlign = close < month1High and bearishBias

// ═══════════════════════════════════════════
// WEEKLY LEVELS
// ═══════════════════════════════════════════
week1High = ta.valuewhen(isNewWeek, high[1], 0)
week1Low  = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, week1High, bar_index+100, week1High, color=color.new(color.blue,0), width=2)
    wLL := line.new(bar_index, week1Low,  bar_index+100, week1Low,  color=color.new(color.blue,0), width=2)
    label.new(bar_index, week1High, "W High", color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, week1Low,  "W Low",  color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY LEVELS
// ═══════════════════════════════════════════
day1High = ta.valuewhen(isNewDay, high[1], 0)
day1Low  = ta.valuewhen(isNewDay, low[1],  0)
day2High = ta.valuewhen(isNewDay, high[1], 1)
day2Low  = ta.valuewhen(isNewDay, low[1],  1)
day3High = ta.valuewhen(isNewDay, high[1], 2)
day3Low  = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, day1High, bar_index+100, day1High, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, day1Low,  bar_index+100, day1Low,  color=color.new(color.black,0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, day2High, bar_index+100, day2High, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, day2Low,  bar_index+100, day2Low,  color=color.new(color.black,40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, day3High, bar_index+100, day3High, color=color.new(color.black,70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, day3Low,  bar_index+100, day3Low,  color=color.new(color.black,70), width=1, style=line.style_dotted)
    label.new(bar_index, day1High, "D1H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day1Low,  "D1L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2High, "D2H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day2Low,  "D2L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3High, "D3H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, day3Low,  "D3L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// EQUAL HIGHS AND LOWS
// ═══════════════════════════════════════════
swingH     = ta.pivothigh(high, 5, 5)
swingL     = ta.pivotlow(low,   5, 5)
prevSwingH = ta.valuewhen(not na(swingH), swingH, 1)
prevSwingL = ta.valuewhen(not na(swingL), swingL, 1)
isEQH      = not na(swingH) and not na(prevSwingH) and math.abs(swingH - prevSwingH) <= atr * 0.1
isEQL      = not na(swingL) and not na(prevSwingL) and math.abs(swingL - prevSwingL) <= atr * 0.1
if showEQLevels and isEQH
    line.new(bar_index-5, swingH, bar_index+20, swingH, color=color.new(color.red,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, swingH, "EQH — Liquidity Above", color=color.new(color.red,70), textcolor=color.red, style=label.style_label_right, size=size.tiny)
if showEQLevels and isEQL
    line.new(bar_index-5, swingL, bar_index+20, swingL, color=color.new(color.green,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, swingL, "EQL — Liquidity Below", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_right, size=size.tiny)
eqhSwept = not na(swingH) and high > swingH and close < swingH
eqlSwept = not na(swingL) and low  < swingL and close > swingL

// ═══════════════════════════════════════════
// WEEKLY BIAS LINES (Yellow) — Visual only
// Detection uses GMT hour window directly
// ═══════════════════════════════════════════
var float weekBiasHigh = na
var float weekBiasLow  = na
var float wbTempH1     = na
var float wbTempL1     = na
var float wbTempH2     = na
var float wbTempL2     = na
var line  wbHL         = na
var line  wbLL         = na
var bool  weekBullish  = false
var bool  weekBearish  = false
var bool  wbC1Set      = false
var bool  wbC2Set      = false
var int   wbWeekNum    = weekofyear(time, "GMT+0")

isNewISOWeek = ta.change(weekofyear(time, "GMT+0")) != 0

if isNewISOWeek
    weekBullish  := false
    weekBearish  := false
    weekBiasHigh := na
    weekBiasLow  := na
    wbTempH1     := na
    wbTempL1     := na
    wbTempH2     := na
    wbTempL2     := na
    wbC1Set      := false
    wbC2Set      := false
    wbWeekNum    := weekofyear(time, "GMT+0")

isC1Bar = dayofweek(time, "GMT+0") == 2 and gmtHour >= weekBiasHour1 and gmtHour < weekBiasHour1 + 4
isC2Bar = dayofweek(time, "GMT+0") == 2 and gmtHour >= weekBiasHour2 and gmtHour < weekBiasHour2 + 4

if isC1Bar and weekofyear(time, "GMT+0") == wbWeekNum
    wbTempH1 := na(wbTempH1) ? high : math.max(wbTempH1, high)
    wbTempL1 := na(wbTempL1) ? low  : math.min(wbTempL1, low)
    wbC1Set  := true

if isC2Bar and weekofyear(time, "GMT+0") == wbWeekNum
    wbTempH2 := na(wbTempH2) ? high : math.max(wbTempH2, high)
    wbTempL2 := na(wbTempL2) ? low  : math.min(wbTempL2, low)
    wbC2Set  := true

wbNewHigh = wbC1Set and wbC2Set ? math.max(wbTempH1, wbTempH2) : wbC1Set ? wbTempH1 : na
wbNewLow  = wbC1Set and wbC2Set ? math.min(wbTempL1, wbTempL2) : wbC1Set ? wbTempL1 : na

if not na(wbNewHigh) and not na(wbNewLow)
    if wbNewHigh != weekBiasHigh or wbNewLow != weekBiasLow
        weekBiasHigh := wbNewHigh
        weekBiasLow  := wbNewLow
        if showWeekBias
            line.delete(wbHL)
            line.delete(wbLL)
            wbHL := line.new(bar_index, weekBiasHigh, bar_index+500, weekBiasHigh, color=color.new(color.yellow,0), width=2)
            wbLL := line.new(bar_index, weekBiasLow,  bar_index+500, weekBiasLow,  color=color.new(color.yellow,0), width=2)
            label.new(bar_index, weekBiasHigh, weekBiasLbl + " High", color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
            label.new(bar_index, weekBiasLow,  weekBiasLbl + " Low",  color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

if not na(weekBiasHigh) and not na(weekBiasLow)
    if close > weekBiasHigh
        weekBullish := true
        weekBearish := false
    else if close < weekBiasLow
        weekBearish := true
        weekBullish := false
    else
        weekBullish := false
        weekBearish := false

// ═══════════════════════════════════════════
// DAILY BIAS LINES (Green) — Visual only
// ═══════════════════════════════════════════
var float dayBiasHigh = na
var float dayBiasLow  = na
var float dbTempH     = na
var float dbTempL     = na
var line  dbHL        = na
var line  dbLL        = na
var bool  dayBullish  = false
var bool  dayBearish  = false
var bool  dbSet       = false
var int   dbDayNum    = dayofmonth(time, "GMT+0")

isNewGMTDay = ta.change(dayofmonth(time, "GMT+0")) != 0

if isNewGMTDay
    dayBullish  := false
    dayBearish  := false
    dayBiasHigh := na
    dayBiasLow  := na
    dbTempH     := na
    dbTempL     := na
    dbSet       := false
    dbDayNum    := dayofmonth(time, "GMT+0")

isDailyBiasBar = gmtHour >= dailyBiasHour and gmtHour < dailyBiasHour + 4

if isDailyBiasBar and dayofmonth(time, "GMT+0") == dbDayNum
    dbTempH := na(dbTempH) ? high : math.max(dbTempH, high)
    dbTempL := na(dbTempL) ? low  : math.min(dbTempL, low)
    dbSet   := true

if dbSet and not na(dbTempH) and not na(dbTempL)
    if dbTempH != dayBiasHigh or dbTempL != dayBiasLow
        dayBiasHigh := dbTempH
        dayBiasLow  := dbTempL
        if showDayBias
            line.delete(dbHL)
            line.delete(dbLL)
            dbHL := line.new(bar_index, dayBiasHigh, bar_index+100, dayBiasHigh, color=color.new(color.green,0), width=2, style=line.style_dashed)
            dbLL := line.new(bar_index, dayBiasLow,  bar_index+100, dayBiasLow,  color=color.new(color.green,0), width=2, style=line.style_dashed)
            label.new(bar_index, dayBiasHigh, dailyBiasLbl + " H", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
            label.new(bar_index, dayBiasLow,  dailyBiasLbl + " L", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

if not na(dayBiasHigh) and not na(dayBiasLow)
    if close > dayBiasHigh
        dayBullish := true
        dayBearish := false
    else if close < dayBiasLow
        dayBearish := true
        dayBullish := false
    else
        dayBullish := false
        dayBearish := false

// ═══════════════════════════════════════════
// PREMIUM AND DISCOUNT ZONE
// ═══════════════════════════════════════════
weekRange  = week1High - week1Low
weekMid    = week1Low + weekRange / 2
inDiscount = close < weekMid and not na(weekMid)
inPremium  = close > weekMid and not na(weekMid)
if showZone and isNewWeek and not na(week1High) and not na(week1Low)
    line.new(bar_index, weekMid, bar_index+100, weekMid, color=color.new(color.gray,50), style=line.style_dashed, width=1)
    label.new(bar_index, weekMid, "Equilibrium", color=color.new(color.gray,70), textcolor=color.gray, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swingLen      = 5
swingHigh     = ta.highest(high, swingLen * 2 + 1)[swingLen]
swingLow      = ta.lowest(low,   swingLen * 2 + 1)[swingLen]
strongBOSBody = math.abs(close - open) > atr * 0.7
bullBOS       = close > swingHigh and strongBOSBody and bullishBias and close[1] <= swingHigh
bearBOS       = close < swingLow  and strongBOSBody and bearishBias and close[1] >= swingLow

var bool recentBullBOS = false
var bool recentBearBOS = false
var int  bosTimer      = 0

if bullBOS
    recentBullBOS := true
    recentBearBOS := false
    bosTimer      := bar_index
if bearBOS
    recentBearBOS := true
    recentBullBOS := false
    bosTimer      := bar_index
if bar_index - bosTimer > bosReset
    recentBullBOS := false
    recentBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low - atr*0.5, "🔨 BOS ↑", color=color.new(color.green,20), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index-1, swingHigh, bar_index+10, swingHigh, color=color.new(color.green,40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr*0.5, "🔨 BOS ↓", color=color.new(color.red,20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index-1, swingLow, bar_index+10, swingLow, color=color.new(color.red,40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweepWick = day1Low - low
bearSweepWick = high - day1High
bullishSweep  = low < day1Low   and close > day1Low  and bullSweepWick >= atr * 0.3
bearishSweep  = high > day1High and close < day1High and bearSweepWick >= atr * 0.3

var float lastBullSweepLow  = na
var float lastBearSweepHigh = na
if bullishSweep
    lastBullSweepLow  := low
if bearishSweep
    lastBearSweepHigh := high

bullSweepRej = bullishSweep and (close - low)  > atr * 1.5
bearSweepRej = bearishSweep and (high - close) > atr * 1.5

plotshape(showSweeps and bullishSweep, "💧 Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime,0), size=size.small, text="Sweep")
plotshape(showSweeps and bearishSweep, "💧 Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,0),  size=size.small, text="Sweep")

intBullSweep      = low  < ta.lowest(low[1],   3)[1] and close > ta.lowest(low[1],   3)[1]
intBearSweep      = high > ta.highest(high[1], 3)[1] and close < ta.highest(high[1], 3)[1]
internalBullSweep = intBullSweep and bullishBias
internalBearSweep = intBearSweep and bearishBias

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
impulseLen   = 3
bullImpulse  = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImpulse  = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBCandle = bullImpulse and close[impulseLen] < open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5
bearOBCandle = bearImpulse and close[impulseLen] > open[impulseLen] and math.abs(close[impulseLen] - open[impulseLen]) > atr * 0.5

var float bullOBTop       = na
var float bullOBBot       = na
var float bearOBTop       = na
var float bearOBBot       = na
var bool  bullOBAlive     = false
var bool  bearOBAlive     = false
var box   bullOBBox       = na
var box   bearOBBox       = na
var bool  bullOBRespected = false
var bool  bearOBRespected = false
var int   bullOBTouches   = 0
var int   bearOBTouches   = 0

if bullOBCandle and bullishBias and showOB
    bullOBTop       := high[impulseLen]
    bullOBBot       := low[impulseLen]
    bullOBAlive     := true
    bullOBRespected := false
    bullOBTouches   := 0
    box.delete(bullOBBox)
    bullOBBox := box.new(bar_index-impulseLen, bullOBTop, bar_index+100, bullOBBot, bgcolor=color.new(color.blue,80), border_color=color.new(color.blue,30), border_width=1)
    label.new(bar_index-impulseLen, bullOBTop, "🟦 Bull OB", color=color.new(color.blue,60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBCandle and bearishBias and showOB
    bearOBTop       := high[impulseLen]
    bearOBBot       := low[impulseLen]
    bearOBAlive     := true
    bearOBRespected := false
    bearOBTouches   := 0
    box.delete(bearOBBox)
    bearOBBox := box.new(bar_index-impulseLen, bearOBTop, bar_index+100, bearOBBot, bgcolor=color.new(color.orange,80), border_color=color.new(color.orange,30), border_width=1)
    label.new(bar_index-impulseLen, bearOBBot, "🟧 Bear OB", color=color.new(color.orange,60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bullOBAlive and bullOBTouches == 0
    if low <= bullOBTop and close >= bullOBBot
        bullOBRespected := true
        bullOBTouches   += 1
        box.set_bgcolor(bullOBBox, color.new(color.green,70))
        box.set_border_color(bullOBBox, color.new(color.green,20))
        label.new(bar_index, bullOBBot, "✅ OB Entry Zone", color=color.new(color.green,30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearOBAlive and bearOBTouches == 0
    if high >= bearOBBot and close <= bearOBTop
        bearOBRespected := true
        bearOBTouches   += 1
        box.set_bgcolor(bearOBBox, color.new(color.red,70))
        box.set_border_color(bearOBBox, color.new(color.red,20))
        label.new(bar_index, bearOBTop, "✅ OB Entry Zone", color=color.new(color.red,30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullOBAlive and close < bullOBBot
    bullOBAlive     := false
    bullOBRespected := false
    box.set_bgcolor(bullOBBox, color.new(color.gray,90))
if bearOBAlive and close > bearOBTop
    bearOBAlive     := false
    bearOBRespected := false
    box.set_bgcolor(bearOBBox, color.new(color.gray,90))

// ═══════════════════════════════════════════
// FVG WITH RETRACEMENT ENTRY
// ═══════════════════════════════════════════
candleBody   = math.abs(close[1] - open[1])
strongCandle = candleBody > atr * 1.0
dispCandle   = math.abs(close - open) > atr * 1.5
bullFVGSize  = low - high[2]
bearFVGSize  = low[2] - high
bullFVGraw   = low > high[2] and strongCandle and close[1] > open[1] and bullishBias and bullFVGSize >= atr * fvgMinSize
bearFVGraw   = high < low[2] and strongCandle and close[1] < open[1] and bearishBias and bearFVGSize >= atr * fvgMinSize

var float bullFVGTop       = na
var float bullFVGBot       = na
var float bearFVGTop       = na
var float bearFVGBot       = na
var bool  bullFVGAlive     = false
var bool  bearFVGAlive     = false
var box   bullFVGBox       = na
var box   bearFVGBox       = na
var bool  bullFVGRespected = false
var bool  bearFVGRespected = false
var bool  bullEntryZone    = false
var bool  bearEntryZone    = false
var box   bullEZBox        = na
var box   bearEZBox        = na

if bullFVGraw and showFVG
    bullFVGTop       := low
    bullFVGBot       := high[2]
    bullFVGAlive     := true
    bullFVGRespected := false
    bullEntryZone    := false
    box.delete(bullFVGBox)
    bullFVGBox := box.new(bar_index-2, bullFVGTop, bar_index+100, bullFVGBot, bgcolor=color.new(color.green,85), border_color=color.new(color.green,40), border_width=1)
    label.new(bar_index, bullFVGBot, "📦 Bull FVG", color=color.new(color.green,50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bearFVGraw and showFVG
    bearFVGTop       := low[2]
    bearFVGBot       := high
    bearFVGAlive     := true
    bearFVGRespected := false
    bearEntryZone    := false
    box.delete(bearFVGBox)
    bearFVGBox := box.new(bar_index-2, bearFVGTop, bar_index+100, bearFVGBot, bgcolor=color.new(color.red,85), border_color=color.new(color.red,40), border_width=1)
    label.new(bar_index, bearFVGTop, "📦 Bear FVG", color=color.new(color.red,50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bullFVGAlive and low <= bullFVGTop and close >= bullFVGBot
    bullFVGRespected := true
    box.set_bgcolor(bullFVGBox, color.new(color.green,70))
    if dispCandle and bullishBias and not bullEntryZone
        bullEntryZone := true
        box.delete(bullEZBox)
        bullEZBox := box.new(bar_index, bullFVGTop, bar_index+30, bullFVGBot, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, bullFVGBot, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_up, size=size.small)

if bearFVGAlive and high >= bearFVGBot and close <= bearFVGTop
    bearFVGRespected := true
    box.set_bgcolor(bearFVGBox, color.new(color.red,70))
    if dispCandle and bearishBias and not bearEntryZone
        bearEntryZone := true
        box.delete(bearEZBox)
        bearEZBox := box.new(bar_index, bearFVGTop, bar_index+30, bearFVGBot, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, bearFVGTop, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_down, size=size.small)

if bullFVGAlive and close < bullFVGBot
    bullFVGAlive     := false
    bullFVGRespected := false
    bullEntryZone    := false
    box.set_bgcolor(bullFVGBox, color.new(color.gray,90))
if bearFVGAlive and close > bearFVGTop
    bearFVGAlive     := false
    bearFVGRespected := false
    bearEntryZone    := false
    box.set_bgcolor(bearFVGBox, color.new(color.gray,90))

// ═══════════════════════════════════════════
// RSI DIVERGENCE
// ═══════════════════════════════════════════
rsi = ta.rsi(close, 14)
lkb = 7
priceLL = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
priceHL = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
priceHH = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
priceLH = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)
rsiHL   = ta.lowest(rsi,  lkb) > ta.lowest(rsi[lkb],  lkb)
rsiLL   = ta.lowest(rsi,  lkb) < ta.lowest(rsi[lkb],  lkb)
rsiLH   = ta.highest(rsi, lkb) < ta.highest(rsi[lkb], lkb)
rsiHH   = ta.highest(rsi, lkb) > ta.highest(rsi[lkb], lkb)

rsiBullExt    = rsi < 40
rsiBearExt    = rsi > 60
strongBullDiv = priceLL and rsiHL
hiddenBullDiv = priceHL and rsiLL
strongBearDiv = priceHH and rsiLH
hiddenBearDiv = priceLH and rsiHH
anyBullDiv    = strongBullDiv or hiddenBullDiv
anyBearDiv    = strongBearDiv or hiddenBearDiv

if showDiv and strongBullDiv
    label.new(bar_index, low - atr*0.3, "◆ Strong Bull Div", color=color.new(color.green,0), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showDiv and hiddenBullDiv
    label.new(bar_index, low - atr*0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,0), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showDiv and strongBearDiv
    label.new(bar_index, high + atr*0.3, "◆ Strong Bear Div", color=color.new(color.red,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hiddenBearDiv
    label.new(bar_index, high + atr*0.3, "◇ Hidden Bear Div", color=color.new(color.orange,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdLine, signalLine, histLine] = ta.macd(close, 12, 26, 9)
macdBullish = macdLine > signalLine
macdBearish = macdLine < signalLine

// ═══════════════════════════════════════════
// TIER 2 STRUCTURE
// ═══════════════════════════════════════════
str2Long = 0
str2Long := str2Long + (bullishBias    ? 1 : 0)
str2Long := str2Long + (recentBullBOS  ? 1 : 0)
str2Long := str2Long + (bullishSweep   ? 1 : 0)
str2Long := str2Long + (inDiscount     ? 1 : 0)
str2Long := str2Long + (monthBullAlign ? 1 : 0)

str2Short = 0
str2Short := str2Short + (bearishBias    ? 1 : 0)
str2Short := str2Short + (recentBearBOS  ? 1 : 0)
str2Short := str2Short + (bearishSweep   ? 1 : 0)
str2Short := str2Short + (inPremium      ? 1 : 0)
str2Short := str2Short + (monthBearAlign ? 1 : 0)

// ═══════════════════════════════════════════
// TIER 3 TRIGGERS
// ═══════════════════════════════════════════
trigger1Bull   = bullEntryZone
trigger2Bull   = bullOBRespected
trigger3Bull   = bullSweepRej
anyTriggerBull = trigger1Bull or trigger2Bull or trigger3Bull

trigger1Bear   = bearEntryZone
trigger2Bear   = bearOBRespected
trigger3Bear   = bearSweepRej
anyTriggerBear = trigger1Bear or trigger2Bear or trigger3Bear

trigStrBull = trigger1Bull ? "🎯 FVG Retracement" : trigger2Bull ? "🟦 OB Respected" : trigger3Bull ? "💧 Sweep Rejection" : "⏳ None"
trigStrBear = trigger1Bear ? "🎯 FVG Retracement" : trigger2Bear ? "🟧 OB Respected" : trigger3Bear ? "💧 Sweep Rejection" : "⏳ None"

// ═══════════════════════════════════════════
// TIER 4 BONUS
// ═══════════════════════════════════════════
bonusLong = 0
bonusLong := bonusLong + (strongBullDiv     ? 2 : hiddenBullDiv ? 1 : 0)
bonusLong := bonusLong + (macdBullish       ? 1 : 0)
bonusLong := bonusLong + (rsiBullExt        ? 1 : 0)
bonusLong := bonusLong + (isTrending        ? 1 : 0)
bonusLong := bonusLong + (internalBullSweep ? 1 : 0)
bonusLong := bonusLong + (eqlSwept          ? 1 : 0)
bonusLong := bonusLong + (weekBullish       ? 1 : 0)
bonusLong := bonusLong + (dayBullish        ? 1 : 0)

bonusShort = 0
bonusShort := bonusShort + (strongBearDiv     ? 2 : hiddenBearDiv ? 1 : 0)
bonusShort := bonusShort + (macdBearish       ? 1 : 0)
bonusShort := bonusShort + (rsiBearExt        ? 1 : 0)
bonusShort := bonusShort + (isTrending        ? 1 : 0)
bonusShort := bonusShort + (internalBearSweep ? 1 : 0)
bonusShort := bonusShort + (eqhSwept          ? 1 : 0)
bonusShort := bonusShort + (weekBearish       ? 1 : 0)
bonusShort := bonusShort + (dayBearish        ? 1 : 0)

// ═══════════════════════════════════════════
// SIGNAL FIRING
// KEY CHANGE — bias is now optional
// requireBias toggle controls whether bias
// gates are mandatory or bonus only
// ═══════════════════════════════════════════
gate1L = requireBias ? weekBullish : true
gate2L = requireBias ? dayBullish  : true
gate3L = htfBullConfirm
gate4L = activeSession
gate5L = allowLongs
allGatesLong  = gate1L and gate2L and gate3L and gate4L and gate5L

gate1S = requireBias ? weekBearish : true
gate2S = requireBias ? dayBearish  : true
gate3S = htfBearConfirm
gate4S = activeSession
gate5S = allowShorts
allGatesShort = gate1S and gate2S and gate3S and gate4S and gate5S

var int lastBullBar = 0
var int lastBearBar = 0

rawBullEntry = allGatesLong  and str2Long  >= minStr2 and anyTriggerBull and bonusLong  >= minBonus
rawBearEntry = allGatesShort and str2Short >= minStr2 and anyTriggerBear and bonusShort >= minBonus

bullishEntry = rawBullEntry and (bar_index - lastBullBar) > sigCooldown
bearishEntry = rawBearEntry and (bar_index - lastBearBar) > sigCooldown

if bullishEntry
    lastBullBar := bar_index
if bearishEntry
    lastBearBar := bar_index

// ═══════════════════════════════════════════
// SIGNAL QUALITY
// ═══════════════════════════════════════════
qualityLong  = bonusLong  >= 7 ? "🔥 A+ HIGH CONFIDENCE" : bonusLong  >= 5 ? "💪 STRONG" : "✅ VALID"
qualityShort = bonusShort >= 7 ? "🔥 A+ HIGH CONFIDENCE" : bonusShort >= 5 ? "💪 STRONG" : "✅ VALID"

// ═══════════════════════════════════════════
// RISK MANAGEMENT
// ═══════════════════════════════════════════
riskAmount = accountSize * (riskPercent / 100)
slBufferL  = bonusLong  >= 6 ? atr * 0.2 : atr * 0.5
slBufferS  = bonusShort >= 6 ? atr * 0.2 : atr * 0.5

longSL    = not na(lastBullSweepLow)  ? lastBullSweepLow  - slBufferL : close - atr * 1.5
longRisk  = math.max(close - longSL,  0.0001)
longTP    = close + longRisk * 3
longSize  = riskAmount / longRisk

shortSL   = not na(lastBearSweepHigh) ? lastBearSweepHigh + slBufferS : close + atr * 1.5
shortRisk = math.max(shortSL - close, 0.0001)
shortTP   = close - shortRisk * 3
shortSize = riskAmount / shortRisk

// ═══════════════════════════════════════════
// SIGNAL LABELS
// ═══════════════════════════════════════════
biasBullStr = weekBullish and dayBullish ? "🟢 BOTH CONFIRMED" : weekBullish ? "⚡ Week✓ Day?" : dayBullish ? "🟢 Day✓ Week?" : "⏳ Bias Pending"
biasBearStr = weekBearish and dayBearish ? "🔴 BOTH CONFIRMED" : weekBearish ? "⚡ Week✓ Day?" : dayBearish ? "🔴 Day✓ Week?" : "⏳ Bias Pending"

longLbl =
     "⬆ BUY — " + tfLabel + " — " + qualityLong + "\n" +
     "Mode: " + activeModeStr + " | HTF: " + htfNeeded + "\n" +
     "Bias: " + biasBullStr + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "STRUCTURE (" + str.tostring(str2Long) + "/5 need " + str.tostring(minStr2) + "):\n" +
     (bullishBias    ? "✅" : "❌") + " EMA  " + (recentBullBOS ? "✅" : "❌") + " BOS  " + (bullishSweep ? "✅" : "❌") + " Sweep\n" +
     (inDiscount     ? "✅" : "❌") + " Discount  " + (monthBullAlign ? "✅" : "❌") + " Monthly\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRIGGER: " + trigStrBull + "\n" +
     "BONUS: " + str.tostring(bonusLong) + "/9\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL: $" + str.tostring(math.round(longSL,2)) + "\n" +
     "🟢 TP: $" + str.tostring(math.round(longTP,2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmount,2)) + "  🎯 Target: $" + str.tostring(math.round(riskAmount*3,2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(longSize,4)) + " units  |  RR: 1:3"

shortLbl =
     "⬇ SELL — " + tfLabel + " — " + qualityShort + "\n" +
     "Mode: " + activeModeStr + " | HTF: " + htfNeeded + "\n" +
     "Bias: " + biasBearStr + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "STRUCTURE (" + str.tostring(str2Short) + "/5 need " + str.tostring(minStr2) + "):\n" +
     (bearishBias    ? "✅" : "❌") + " EMA  " + (recentBearBOS ? "✅" : "❌") + " BOS  " + (bearishSweep ? "✅" : "❌") + " Sweep\n" +
     (inPremium      ? "✅" : "❌") + " Premium  " + (monthBearAlign ? "✅" : "❌") + " Monthly\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRIGGER: " + trigStrBear + "\n" +
     "BONUS: " + str.tostring(bonusShort) + "/9\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL: $" + str.tostring(math.round(shortSL,2)) + "\n" +
     "🟢 TP: $" + str.tostring(math.round(shortTP,2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmount,2)) + "  🎯 Target: $" + str.tostring(math.round(riskAmount*3,2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(shortSize,4)) + " units  |  RR: 1:3"

if showSignals and bullishEntry
    label.new(bar_index, low,  longLbl,  color=color.new(color.green,10), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index, longSL,  bar_index+40, longSL,  color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP,  bar_index+40, longTP,  color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, longSL, "🔴 SL $" + str.tostring(math.round(longSL,2)),  color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, longTP, "🟢 TP $" + str.tostring(math.round(longTP,2)),  color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearishEntry
    label.new(bar_index, high, shortLbl, color=color.new(color.red,10),   textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index+40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index+40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, shortSL, "🔴 SL $" + str.tostring(math.round(shortSL,2)), color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, shortTP, "🟢 TP $" + str.tostring(math.round(shortTP,2)), color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// ALERTS
// ═══════════════════════════════════════════
alertcondition(bullishEntry, title="🟢 GOLD BUY — High Confidence",  message="⬆ BUY SIGNAL — XAUUSD — Check chart for full details")
alertcondition(bearishEntry, title="🔴 GOLD SELL — High Confidence", message="⬇ SELL SIGNAL — XAUUSD — Check chart for full details")
alertcondition(bullishEntry or bearishEntry, title="🚦 GOLD Signal",  message="Gold signal detected on XAUUSD — Check chart")

// ═══════════════════════════════════════════
// WAIT REASON
// ═══════════════════════════════════════════
wrl3 = not gate3L           ? "⏳ WAIT — HTF (" + htfNeeded + ") Not Aligned"   : ""
wrl4 = not gate4L           ? "⏳ WAIT — Asian Session"                          : ""
wrl5 = not gate5L           ? "⏳ WAIT — Extreme Bear Market"                    : ""
wrl6 = str2Long < minStr2   ? "⏳ WAIT — Need " + str.tostring(minStr2 - str2Long) + " More Structure"    : ""
wrl7 = not anyTriggerBull   ? "⏳ WAIT — Need Entry Trigger (FVG/OB/Sweep)"     : ""
wrl8 = bonusLong < minBonus ? "⏳ WAIT — Need " + str.tostring(minBonus - bonusLong) + " More Bonus Pts" : ""
waitReasonLong = not gate3L ? wrl3 : not gate4L ? wrl4 : not gate5L ? wrl5 : str2Long < minStr2 ? wrl6 : not anyTriggerBull ? wrl7 : bonusLong < minBonus ? wrl8 : "✅ LONG CONDITIONS MET"

wrs3 = not gate3S            ? "⏳ WAIT — HTF (" + htfNeeded + ") Not Aligned"   : ""
wrs4 = not gate4S            ? "⏳ WAIT — Asian Session"                          : ""
wrs5 = not gate5S            ? "⏳ WAIT — Extreme Bear Market"                    : ""
wrs6 = str2Short < minStr2   ? "⏳ WAIT — Need " + str.tostring(minStr2 - str2Short) + " More Structure"   : ""
wrs7 = not anyTriggerBear    ? "⏳ WAIT — Need Entry Trigger (FVG/OB/Sweep)"     : ""
wrs8 = bonusShort < minBonus ? "⏳ WAIT — Need " + str.tostring(minBonus - bonusShort) + " More Bonus Pts" : ""
waitReasonShort = not gate3S ? wrs3 : not gate4S ? wrs4 : not gate5S ? wrs5 : str2Short < minStr2 ? wrs6 : not anyTriggerBear ? wrs7 : bonusShort < minBonus ? wrs8 : "✅ SHORT CONDITIONS MET"

// ═══════════════════════════════════════════
// DASHBOARD
// ═══════════════════════════════════════════
var table dash = table.new(position.top_right, 3, 35, bgcolor=color.new(color.black,50), border_color=color.new(color.gray,30), border_width=1)
hdrBG  = color.new(color.blue, 40)
secBG  = color.new(color.gray, 60)
secCol = color.yellow

table.cell(dash, 0, 0, "📊 CONDITION",  text_color=color.white,  text_size=size.small, bgcolor=hdrBG)
table.cell(dash, 1, 0, "STATUS",        text_color=color.white,  text_size=size.small, bgcolor=hdrBG)
table.cell(dash, 2, 0, "🥇 " + tfLabel, text_color=color.yellow, text_size=size.small, bgcolor=hdrBG)

table.cell(dash, 0, 1, "── MARKET ──",  text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 1, "────────────",  text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 1, "──────",        text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 2, "Regime",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 2, regimeStr,       text_color=regimeCol,   text_size=size.small)
table.cell(dash, 2, 2, extremeBear ? "🚫" : "✅", text_color=extremeBear ? color.red : color.green, text_size=size.small)

table.cell(dash, 0, 3, "ADX",           text_color=color.white, text_size=size.small)
table.cell(dash, 1, 3, adxStr,          text_color=adxCol,      text_size=size.small)
table.cell(dash, 2, 3, isTrending ? "✅" : "❌", text_color=isTrending ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 4, "Session",       text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 4, sessionName,     text_color=sessionColor, text_size=size.small)
table.cell(dash, 2, 4, activeSession ? "✅" : "❌", text_color=activeSession ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 5, "TF Mode",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 5, activeModeStr,   text_color=color.aqua,  text_size=size.small)
table.cell(dash, 2, 5, requireBias ? "🔒 Bias ON" : "🔓 Bias OFF", text_color=requireBias ? color.orange : color.green, text_size=size.tiny)

table.cell(dash, 0, 6, "HTF (" + htfNeeded + ")", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 6, htfStr, text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.tiny)
table.cell(dash, 2, 6, htfBullConfirm ? "✅🟢" : htfBearConfirm ? "✅🔴" : "❌", text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.small)

table.cell(dash, 0, 7, "── BIAS (BONUS) ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 7, requireBias ? "── MANDATORY ──" : "── OPTIONAL +1pt each ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 7, "──────",             text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 8, "Weekly Bias ⚡", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 8, weekBullish ? "🟢 BULL — Above Yellow ✓" : weekBearish ? "🔴 BEAR — Below Yellow ✓" : not na(weekBiasHigh) ? "⏳ RANGING Inside Lines" : "⏳ Lines Not Set Yet", text_color=weekBullish ? color.green : weekBearish ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 8, weekBullish or weekBearish ? "✅" : "➖", text_color=weekBullish or weekBearish ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 9, "Daily Bias 🟢",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 9, dayBullish ? "🟢 BULL — Above Green ✓" : dayBearish ? "🔴 BEAR — Below Green ✓" : not na(dayBiasHigh) ? "⏳ RANGING Inside Lines" : "⏳ Lines Not Set Yet", text_color=dayBullish ? color.green : dayBearish ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 9, dayBullish or dayBearish ? "✅" : "➖", text_color=dayBullish or dayBearish ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 10, "HTF Aligned",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 10, htfStr,          text_color=htfBullConfirm ? color.green : htfBearConfirm ? color.red : color.yellow, text_size=size.tiny)
table.cell(dash, 2, 10, htfBullConfirm or htfBearConfirm ? "✅" : "❌", text_color=htfBullConfirm or htfBearConfirm ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 11, "Session Gate",  text_color=color.white,  text_size=size.small)
table.cell(dash, 1, 11, sessionName,     text_color=sessionColor, text_size=size.small)
table.cell(dash, 2, 11, activeSession ? "✅" : "❌", text_color=activeSession ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 12, "Extreme Bear",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 12, extremeBear ? "🚨 BLOCKED" : "✅ SAFE",  text_color=extremeBear ? color.red : color.green, text_size=size.small)
table.cell(dash, 2, 12, extremeBear ? "❌" : "✅", text_color=extremeBear ? color.red : color.green, text_size=size.small)

table.cell(dash, 0, 13, "── TIER 2 STRUCTURE ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 13, "────────────────────",   text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 13, "──────",                 text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 14, "EMA 200",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 14, bullishBias ? "🟢 BULLISH" : "🔴 BEARISH", text_color=bullishBias ? color.green : color.red, text_size=size.small)
table.cell(dash, 2, 14, "✅", text_color=color.green, text_size=size.small)

table.cell(dash, 0, 15, "BOS",           text_color=color.white, text_size=size.small)
table.cell(dash, 1, 15, recentBullBOS ? "🟢 BULL" : recentBearBOS ? "🔴 BEAR" : "⏳ NONE", text_color=recentBullBOS ? color.green : recentBearBOS ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 15, recentBullBOS or recentBearBOS ? "✅" : "❌", text_color=recentBullBOS or recentBearBOS ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 16, "Sweep",         text_color=color.white, text_size=size.small)
table.cell(dash, 1, 16, bullishSweep ? "🟢 BULL" : bearishSweep ? "🔴 BEAR" : "⏳ NONE", text_color=bullishSweep or bearishSweep ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 16, bullishSweep or bearishSweep ? "✅" : "❌", text_color=bullishSweep or bearishSweep ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 17, "Zone",          text_color=color.white, text_size=size.small)
table.cell(dash, 1, 17, inDiscount ? "🟢 DISCOUNT" : inPremium ? "🔴 PREMIUM" : "⚪ EQUILIBRIUM", text_color=inDiscount ? color.green : inPremium ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 17, inDiscount or inPremium ? "✅" : "➖", text_color=inDiscount or inPremium ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 18, "Monthly",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 18, monthBullAlign ? "🟢 BULL" : monthBearAlign ? "🔴 BEAR" : "⏳ WAIT", text_color=monthBullAlign ? color.green : monthBearAlign ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 18, monthBullAlign or monthBearAlign ? "✅" : "❌", text_color=monthBullAlign or monthBearAlign ? color.green : color.red, text_size=size.small)

str2Display = bearishBias ? str2Short : str2Long
table.cell(dash, 0, 19, "Structure Score", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 19, str.tostring(str2Display) + "/5  (need " + str.tostring(minStr2) + ")", text_color=str2Display >= minStr2 ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 19, str2Display >= minStr2 ? "✅" : "❌", text_color=str2Display >= minStr2 ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 20, "── TIER 3 TRIGGER ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 20, "────────────────────", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 20, "──────",               text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 21, "FVG Entry Zone", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 21, bullEntryZone or bearEntryZone ? "🎯 ACTIVE" : "⏳ NONE", text_color=bullEntryZone or bearEntryZone ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 21, bullEntryZone or bearEntryZone ? "✅" : "➖", text_color=bullEntryZone or bearEntryZone ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 22, "Order Block",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 22, bullOBRespected ? "✅ BULL" : bearOBRespected ? "✅ BEAR" : bullOBAlive ? "🟦 LIVE" : bearOBAlive ? "🟧 LIVE" : "⏳ NONE", text_color=bullOBRespected or bearOBRespected ? color.green : bullOBAlive or bearOBAlive ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 22, bullOBRespected or bearOBRespected ? "✅" : bullOBAlive or bearOBAlive ? "⏳" : "➖", text_color=bullOBRespected or bearOBRespected ? color.green : color.yellow, text_size=size.small)

table.cell(dash, 0, 23, "Sweep Rejection", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 23, bullSweepRej ? "🟢 BULL" : bearSweepRej ? "🔴 BEAR" : "⏳ NONE", text_color=bullSweepRej or bearSweepRej ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 23, bullSweepRej or bearSweepRej ? "✅" : "➖", text_color=bullSweepRej or bearSweepRej ? color.green : color.gray, text_size=size.small)

trigActive = anyTriggerBull or anyTriggerBear
table.cell(dash, 0, 24, "Trigger Status", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 24, trigActive ? "🟢 TRIGGER ACTIVE" : "⏳ WAITING", text_color=trigActive ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 24, trigActive ? "✅" : "❌", text_color=trigActive ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 25, "── TIER 4 BONUS ──", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 25, "──────────────────", text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 25, "──────",             text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 26, "RSI Div",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 26, strongBullDiv ? "🟢 Strong(2)" : hiddenBullDiv ? "🟢 Hidden(1)" : strongBearDiv ? "🔴 Strong(2)" : hiddenBearDiv ? "🔴 Hidden(1)" : "⏳ NONE", text_color=anyBullDiv ? color.green : anyBearDiv ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 26, anyBullDiv or anyBearDiv ? "✅" : "➖", text_color=anyBullDiv or anyBearDiv ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 27, "MACD",          text_color=color.white, text_size=size.small)
table.cell(dash, 1, 27, macdBullish ? "🟢 BULLISH" : "🔴 BEARISH", text_color=macdBullish ? color.green : color.red, text_size=size.small)
table.cell(dash, 2, 27, "✅", text_color=color.green, text_size=size.small)

table.cell(dash, 0, 28, "RSI",           text_color=color.white, text_size=size.small)
table.cell(dash, 1, 28, str.tostring(math.round(rsi,1)) + (rsiBullExt ? " 🟢 OS" : rsiBearExt ? " 🔴 OB" : " ⚪ N"), text_color=rsiBullExt ? color.green : rsiBearExt ? color.red : color.white, text_size=size.small)
table.cell(dash, 2, 28, rsiBullExt or rsiBearExt ? "✅" : "➖", text_color=rsiBullExt or rsiBearExt ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 29, "Int Sweep",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 29, internalBullSweep ? "🟢 BULL" : internalBearSweep ? "🔴 BEAR" : "⏳ NONE", text_color=internalBullSweep or internalBearSweep ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 29, internalBullSweep or internalBearSweep ? "✅" : "➖", text_color=internalBullSweep or internalBearSweep ? color.green : color.gray, text_size=size.small)

table.cell(dash, 0, 30, "EQH/EQL Swept", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 30, eqhSwept ? "🔴 EQH SWEPT" : eqlSwept ? "🟢 EQL SWEPT" : "⏳ NONE", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 30, eqhSwept or eqlSwept ? "✅" : "➖", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)

bonusDisplay = bearishBias ? bonusShort : bonusLong
table.cell(dash, 0, 31, "Bonus Score",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 31, str.tostring(bonusDisplay) + "/9  (need " + str.tostring(minBonus) + ")", text_color=bonusDisplay >= minBonus ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 31, bonusDisplay >= minBonus ? "✅" : "❌", text_color=bonusDisplay >= minBonus ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 32, "── RISK ──",    text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 1, 32, "────────────",  text_color=secCol, text_size=size.tiny, bgcolor=secBG)
table.cell(dash, 2, 32, "──────",        text_color=secCol, text_size=size.tiny, bgcolor=secBG)

table.cell(dash, 0, 33, "Risk/Target",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 33, "💰 $" + str.tostring(math.round(riskAmount,2)) + " → 🎯 $" + str.tostring(math.round(riskAmount*3,2)) + " (1:3)", text_color=color.yellow, text_size=size.small)
table.cell(dash, 2, 33, "RR 1:3",        text_color=color.yellow, text_size=size.small)

waitStr = bearishBias ? waitReasonShort : waitReasonLong
sigStr  = bullishEntry ? "⬆ BUY " + tfLabel + " — " + qualityLong : bearishEntry ? "⬇ SELL " + tfLabel + " — " + qualityShort : waitStr
sigCol  = bullishEntry ? color.green : bearishEntry ? color.red : color.yellow
sig3    = bullishEntry ? "🟢 BUY" : bearishEntry ? "🔴 SELL" : "⏳ WAIT"
sig3Col = bullishEntry ? color.green : bearishEntry ? color.red : color.yellow

table.cell(dash, 0, 34, "🚦 SIGNAL",     text_color=color.white, text_size=size.normal, bgcolor=color.new(color.gray,30))
table.cell(dash, 1, 34, sigStr,          text_color=sigCol,      text_size=size.small)
table.cell(dash, 2, 34, sig3,            text_color=sig3Col,     text_size=size.normal)
```

```python
// ============================================================
// Manny's Gold Strategy V2 — STEVE BUILD FINAL
// Pine Script v6 — XAUUSD Only
// Chart timezone must be set to UTC
// ============================================================
//@version=6
indicator("Manny's Gold Strategy V2", overlay=true, max_lines_count=500, max_labels_count=500, max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
showMonthly  = input.bool(true,  "📅 Monthly High/Low",           group="📊 Chart Levels")
showWeekly   = input.bool(true,  "📅 Weekly High/Low",            group="📊 Chart Levels")
showDaily    = input.bool(true,  "📅 Daily High/Low (3 Days)",    group="📊 Chart Levels")
showEQL      = input.bool(true,  "📊 Equal Highs and Lows",       group="📊 Chart Levels")
showZone     = input.bool(true,  "📊 Premium/Discount Zone",      group="📊 Chart Levels")
showYellow   = input.bool(true,  "⚡ Monday Bias Lines (Yellow)", group="🧭 Bias Visuals")
showGreen    = input.bool(true,  "🟢 Daily 1AM Lines (Green)",    group="🧭 Bias Visuals")
goldBiasTime = input.string("1AM / 5AM (Standard Time)", "🥇 Bias Candle Time",
     options=["1AM / 5AM (Standard Time)", "2AM / 6AM (Daylight Saving)"],
     group="🧭 Bias Visuals",
     tooltip="Standard = winter. Daylight Saving = summer. Switch manually when clocks change.")
showEMA      = input.bool(true,  "📉 EMA 200 + EMA 50",           group="🔧 Features")
showSweeps   = input.bool(true,  "💧 Liquidity Sweeps",           group="🔧 Features")
showFVG      = input.bool(true,  "📦 Fair Value Gaps",            group="🔧 Features")
showDiv      = input.bool(true,  "📐 RSI Divergence",             group="🔧 Features")
showBOS      = input.bool(true,  "🔨 Break of Structure",         group="🔧 Features")
showOB       = input.bool(true,  "🟦 Order Blocks",               group="🔧 Features")
showSignals  = input.bool(true,  "🚦 Buy/Sell Signals",           group="🔧 Features")
fvgMinIn     = input.float(0.5,  "FVG Min Size (x ATR)",          group="🔧 Features", minval=0.1)
modeOvr      = input.string("Auto", "Trading Mode Override",
     options=["Auto","Scalping","Day Trade","Swing Trade"],      group="🎯 Signal Quality")
minBonusOvr  = input.int(0,      "Min Bonus Override (0=auto)",   group="🎯 Signal Quality", minval=0, maxval=7)
adxLen       = input.int(14,     "ADX Period",                    group="📊 Market Filter")
adxThr       = input.int(25,     "ADX Minimum",                   group="📊 Market Filter")
accountSize  = input.float(1000, "💰 Account Size ($)",           group="💵 Risk")
riskPct      = input.float(1.0,  "⚠️ Risk Per Trade (%)",        group="💵 Risk")
atrLen       = input.int(14,     "ATR Period",                    group="💵 Risk")

// ═══════════════════════════════════════════
// BIAS TIME SETTINGS
// ═══════════════════════════════════════════
isStandard   = goldBiasTime == "1AM / 5AM (Standard Time)"
biasHour1    = isStandard ? 1 : 2
biasHour2    = isStandard ? 5 : 6
biasLbl      = isStandard ? "⚡ 1AM+5AM" : "⚡ 2AM+6AM"
dailyLbl     = isStandard ? "🟢 1AM" : "🟢 2AM"

// ═══════════════════════════════════════════
// TIMEFRAME DETECTION
// ═══════════════════════════════════════════
tfMins      = timeframe.in_seconds() / 60
isScalpTF   = tfMins <= 5
isDayTF     = tfMins > 5   and tfMins <= 30
isSwingTF   = tfMins > 30  and tfMins <= 240
autoStr     = isScalpTF ? "Scalping" : isDayTF ? "Day Trade" : isSwingTF ? "Swing Trade" : "Long Term"
isScalp     = modeOvr == "Auto" ? isScalpTF  : modeOvr == "Scalping"
isSwing     = modeOvr == "Auto" ? isSwingTF  : modeOvr == "Swing Trade"
modeStr     = modeOvr == "Auto" ? autoStr + " (Auto)" : modeOvr
tfLbl       = tfMins == 1 ? "1M" : tfMins == 5 ? "5M" : tfMins == 15 ? "15M" : tfMins == 30 ? "30M" : tfMins == 60 ? "1H" : tfMins == 240 ? "4H" : tfMins == 1440 ? "D" : str.tostring(math.round(tfMins)) + "M"
minBonus    = minBonusOvr > 0 ? minBonusOvr : isScalp ? 2 : isSwing ? 4 : 3
bosRst      = isScalp ? 50 : isSwing ? 20 : 30
fvgMin      = isScalp ? 0.3 : isSwing ? 0.7 : fvgMinIn
minStr2     = isScalp ? 2 : isSwing ? 3 : 2
cooldown    = tfMins <= 1 ? 15 : tfMins <= 5 ? 12 : tfMins <= 15 ? 10 : tfMins <= 30 ? 8 : 5

// ═══════════════════════════════════════════
// TIME
// ═══════════════════════════════════════════
gmtH       = hour(time, "GMT+0")
gmtDow     = dayofweek(time, "GMT+0")
isNewDay   = ta.change(time("D")) != 0
isNewWeek  = ta.change(time("W")) != 0
isNewMonth = ta.change(time("M")) != 0
isNewISOWeek = ta.change(weekofyear(time, "GMT+0")) != 0

// ═══════════════════════════════════════════
// SESSION
// ═══════════════════════════════════════════
london     = gmtH >= 7  and gmtH < 12
ny         = gmtH >= 13 and gmtH < 17
overlap    = gmtH >= 12 and gmtH < 13
activeSess = london or ny or overlap
sessName   = overlap ? "⚡ Overlap — BEST" : london ? "🇬🇧 London" : ny ? "🇺🇸 New York" : "😴 Asian — BLOCKED"
sessCol    = overlap ? color.yellow : london ? color.green : ny ? color.blue : color.gray

// ═══════════════════════════════════════════
// ATR + EMA
// ═══════════════════════════════════════════
atr        = ta.atr(atrLen)
ema200     = ta.ema(close, 200)
ema50      = ta.ema(close, 50)
aboveE200  = close > ema200
belowE200  = close < ema200
bullMkt    = ema50 > ema200
pctBelow   = (ema200 - close) / ema200 * 100
extBear    = pctBelow > 20
regStr     = extBear ? "🚨 EXTREME BEAR" : bullMkt ? "🟢 BULL MARKET" : "🔴 BEAR MARKET"
regCol     = extBear ? color.red : bullMkt ? color.green : color.red
plot(showEMA ? ema200 : na, "EMA 200", color=aboveE200 ? color.new(color.green,0) : color.new(color.red,0), linewidth=2)
plot(showEMA ? ema50  : na, "EMA 50",  color=bullMkt   ? color.new(color.green,50) : color.new(color.red,50), linewidth=1)

// ═══════════════════════════════════════════
// ADX
// ═══════════════════════════════════════════
[diP, diM, adxVal] = ta.dmi(adxLen, adxLen)
trending   = adxVal >= adxThr
adxStr     = trending ? "📈 TRENDING (" + str.tostring(math.round(adxVal,1)) + ")" : "📉 RANGING (" + str.tostring(math.round(adxVal,1)) + ")"
adxCol     = trending ? color.green : color.red

// ═══════════════════════════════════════════
// HTF EMA 200
// ═══════════════════════════════════════════
c1H    = request.security(syminfo.tickerid, "60",  close,              lookahead=barmerge.lookahead_off)
e1H    = request.security(syminfo.tickerid, "60",  ta.ema(close, 200), lookahead=barmerge.lookahead_off)
c4H    = request.security(syminfo.tickerid, "240", close,              lookahead=barmerge.lookahead_off)
e4H    = request.security(syminfo.tickerid, "240", ta.ema(close, 200), lookahead=barmerge.lookahead_off)
cD     = request.security(syminfo.tickerid, "D",   close,              lookahead=barmerge.lookahead_off)
eD     = request.security(syminfo.tickerid, "D",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)
cW     = request.security(syminfo.tickerid, "W",   close,              lookahead=barmerge.lookahead_off)
eW     = request.security(syminfo.tickerid, "W",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)

bull1H = c1H > e1H
bull4H = c4H > e4H
bullD  = cD  > eD
bullW  = cW  > eW
bear1H = c1H < e1H
bear4H = c4H < e4H
bearD  = cD  < eD
bearW  = cW  < eW

// ═══════════════════════════════════════════
// GATE 1 — WEEKLY BIAS via 4H EMA 200
// ═══════════════════════════════════════════
weekBull  = bull4H
weekBear  = bear4H
wkBiasStr = weekBull ? "🟢 BULL — 4H above EMA200" : weekBear ? "🔴 BEAR — 4H below EMA200" : "⏳ FLAT"

// ═══════════════════════════════════════════
// GATE 2 — DAILY BIAS via 1H EMA 200
// ═══════════════════════════════════════════
dayBull   = bull1H
dayBear   = bear1H
dBiasStr  = dayBull ? "🟢 BULL — 1H above EMA200" : dayBear ? "🔴 BEAR — 1H below EMA200" : "⏳ FLAT"

// ═══════════════════════════════════════════
// GATE 3 — HTF CONFIRMATION
// ═══════════════════════════════════════════
htfBullOk = isScalpTF ? (bull1H and bull4H and bullD) : isDayTF ? (bull1H and bull4H) : isSwingTF ? bull4H : (bullD and bullW)
htfBearOk = isScalpTF ? (bear1H and bear4H and bearD) : isDayTF ? (bear1H and bear4H) : isSwingTF ? bear4H : (bearD and bearW)
htfNeeded = isScalpTF ? "1H+4H+D" : isDayTF ? "1H+4H" : isSwingTF ? "4H" : "D+W"
e1He      = bull1H ? "🟢" : "🔴"
e4He      = bull4H ? "🟢" : "🔴"
eDe       = bullD  ? "🟢" : "🔴"
eWe       = bullW  ? "🟢" : "🔴"
htfStr    = htfBullOk ? "🟢 " + htfNeeded + " BULL ✓" : htfBearOk ? "🔴 " + htfNeeded + " BEAR ✓" : "⏳ 1H:" + e1He + " 4H:" + e4He + " D:" + eDe + " W:" + eWe

// ═══════════════════════════════════════════
// MONTHLY LEVELS
// ═══════════════════════════════════════════
mH = ta.valuewhen(isNewMonth, high[1], 0)
mL = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, mH, bar_index+100, mH, color=color.new(color.purple,0), width=2)
    mLL := line.new(bar_index, mL, bar_index+100, mL, color=color.new(color.purple,0), width=2)
    label.new(bar_index, mH, "M High", color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, mL, "M Low",  color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
mBullAlign = close > mL and aboveE200
mBearAlign = close < mH and belowE200

// ═══════════════════════════════════════════
// WEEKLY LEVELS
// ═══════════════════════════════════════════
wH = ta.valuewhen(isNewWeek, high[1], 0)
wL = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, wH, bar_index+100, wH, color=color.new(color.blue,0), width=2)
    wLL := line.new(bar_index, wL, bar_index+100, wL, color=color.new(color.blue,0), width=2)
    label.new(bar_index, wH, "W High", color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, wL, "W Low",  color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY LEVELS — 3 days fading
// ═══════════════════════════════════════════
d1H = ta.valuewhen(isNewDay, high[1], 0)
d1L = ta.valuewhen(isNewDay, low[1],  0)
d2H = ta.valuewhen(isNewDay, high[1], 1)
d2L = ta.valuewhen(isNewDay, low[1],  1)
d3H = ta.valuewhen(isNewDay, high[1], 2)
d3L = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, d1H, bar_index+100, d1H, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, d1L, bar_index+100, d1L, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, d2H, bar_index+100, d2H, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, d2L, bar_index+100, d2L, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, d3H, bar_index+100, d3H, color=color.new(color.black,70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, d3L, bar_index+100, d3L, color=color.new(color.black,70), width=1, style=line.style_dotted)
    label.new(bar_index, d1H, "D1H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d1L, "D1L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d2H, "D2H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d2L, "D2L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d3H, "D3H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d3L, "D3L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// EQUAL HIGHS AND LOWS
// ═══════════════════════════════════════════
sH      = ta.pivothigh(high, 5, 5)
sL      = ta.pivotlow(low,   5, 5)
psH     = ta.valuewhen(not na(sH), sH, 1)
psL     = ta.valuewhen(not na(sL), sL, 1)
isEQH   = not na(sH) and not na(psH) and math.abs(sH - psH) <= atr * 0.1
isEQL   = not na(sL) and not na(psL) and math.abs(sL - psL) <= atr * 0.1
if showEQL and isEQH
    line.new(bar_index-5, sH, bar_index+20, sH, color=color.new(color.red,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, sH, "EQH — Liquidity Above", color=color.new(color.red,70), textcolor=color.red, style=label.style_label_right, size=size.tiny)
if showEQL and isEQL
    line.new(bar_index-5, sL, bar_index+20, sL, color=color.new(color.green,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, sL, "EQL — Liquidity Below", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_right, size=size.tiny)
eqhSwept = not na(sH) and high > sH and close < sH
eqlSwept = not na(sL) and low  < sL and close > sL

// ═══════════════════════════════════════════
// PREMIUM AND DISCOUNT ZONE
// ═══════════════════════════════════════════
wRange = wH - wL
wMid   = wL + wRange / 2
inDisc = close < wMid and not na(wMid)
inPrem = close > wMid and not na(wMid)
if showZone and isNewWeek and not na(wH) and not na(wL)
    line.new(bar_index, wMid, bar_index+100, wMid, color=color.new(color.gray,50), style=line.style_dashed, width=1)
    label.new(bar_index, wMid, "Equilibrium", color=color.new(color.gray,70), textcolor=color.gray, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// YELLOW LINES — FIXED FINAL VERSION
// Named functions detect exact 1AM and 5AM
// Monday candles on the 4H timeframe
// Price levels are identical on ALL timeframes
// Daylight saving toggle included
// ═══════════════════════════════════════════
f_wbH1() =>
    isMon1 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    isMon1 ? high : float(na)

f_wbL1() =>
    isMon1 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    isMon1 ? low : float(na)

f_wbH2() =>
    isMon2 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour2 and minute(time, "GMT+0") == 0
    isMon2 ? high : float(na)

f_wbL2() =>
    isMon2 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour2 and minute(time, "GMT+0") == 0
    isMon2 ? low : float(na)

wb4H_H1 = request.security(syminfo.tickerid, "240", f_wbH1(), lookahead=barmerge.lookahead_off)
wb4H_L1 = request.security(syminfo.tickerid, "240", f_wbL1(), lookahead=barmerge.lookahead_off)
wb4H_H2 = request.security(syminfo.tickerid, "240", f_wbH2(), lookahead=barmerge.lookahead_off)
wb4H_L2 = request.security(syminfo.tickerid, "240", f_wbL2(), lookahead=barmerge.lookahead_off)

// Store values for this week — reset on new ISO week
var float wkH1 = na
var float wkL1 = na
var float wkH2 = na
var float wkL2 = na

if isNewISOWeek
    wkH1 := na
    wkL1 := na
    wkH2 := na
    wkL2 := na

if not na(wb4H_H1)
    wkH1 := wb4H_H1
    wkL1 := wb4H_L1
if not na(wb4H_H2)
    wkH2 := wb4H_H2
    wkL2 := wb4H_L2

// Build the range from the two candles
wbFinalH = not na(wkH1) and not na(wkH2) ? math.max(wkH1, wkH2) : not na(wkH1) ? wkH1 : wkH2
wbFinalL = not na(wkL1) and not na(wkL2) ? math.min(wkL1, wkL2) : not na(wkL1) ? wkL1 : wkL2

var line yHL = na
var line yLL = na
var float lastYH = na
var float lastYL = na

if showYellow and not na(wbFinalH) and not na(wbFinalL)
    if wbFinalH != lastYH or wbFinalL != lastYL
        lastYH := wbFinalH
        lastYL := wbFinalL
        line.delete(yHL)
        line.delete(yLL)
        yHL := line.new(bar_index, wbFinalH, bar_index+500, wbFinalH, color=color.new(color.yellow,0), width=2)
        yLL := line.new(bar_index, wbFinalL, bar_index+500, wbFinalL, color=color.new(color.yellow,0), width=2)
        label.new(bar_index, wbFinalH, biasLbl + " High", color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
        label.new(bar_index, wbFinalL, biasLbl + " Low",  color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// GREEN LINES — FIXED FINAL VERSION
// Named function detects exact 1AM candle
// on the 4H timeframe every day
// Price level identical on ALL timeframes
// Daylight saving toggle included
// ═══════════════════════════════════════════
f_dbH() =>
    is1am = hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    is1am ? high : float(na)

f_dbL() =>
    is1am = hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    is1am ? low : float(na)

db4H_H = request.security(syminfo.tickerid, "240", f_dbH(), lookahead=barmerge.lookahead_off)
db4H_L = request.security(syminfo.tickerid, "240", f_dbL(), lookahead=barmerge.lookahead_off)

var float dayBiasH = na
var float dayBiasL = na
var line  gHL      = na
var line  gLL      = na
var float lastGH   = na
var float lastGL   = na

isNewGMTDay = ta.change(dayofmonth(time, "GMT+0")) != 0

if isNewGMTDay
    dayBiasH := na
    dayBiasL := na

if not na(db4H_H) and not na(db4H_L)
    dayBiasH := db4H_H
    dayBiasL := db4H_L

if showGreen and not na(dayBiasH) and not na(dayBiasL)
    if dayBiasH != lastGH or dayBiasL != lastGL
        lastGH := dayBiasH
        lastGL := dayBiasL
        line.delete(gHL)
        line.delete(gLL)
        gHL := line.new(bar_index, dayBiasH, bar_index+100, dayBiasH, color=color.new(color.green,0), width=2, style=line.style_dashed)
        gLL := line.new(bar_index, dayBiasL, bar_index+100, dayBiasL, color=color.new(color.green,0), width=2, style=line.style_dashed)
        label.new(bar_index, dayBiasH, dailyLbl + " H", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
        label.new(bar_index, dayBiasL, dailyLbl + " L", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swLen     = 5
swHigh    = ta.highest(high, swLen*2+1)[swLen]
swLow     = ta.lowest(low,   swLen*2+1)[swLen]
strongBod = math.abs(close - open) > atr * 0.7
bullBOS   = close > swHigh and strongBod and aboveE200 and close[1] <= swHigh
bearBOS   = close < swLow  and strongBod and belowE200 and close[1] >= swLow

var bool recBullBOS   = false
var bool recBearBOS   = false
var int  bosTimer     = 0
var bool bosBullFired = false
var bool bosBearFired = false
var float bosLevel    = na

if bullBOS
    recBullBOS   := true
    recBearBOS   := false
    bosTimer     := bar_index
    bosBullFired := true
    bosBearFired := false
    bosLevel     := swHigh
if bearBOS
    recBearBOS   := true
    recBullBOS   := false
    bosTimer     := bar_index
    bosBearFired := true
    bosBullFired := false
    bosLevel     := swLow
if bar_index - bosTimer > bosRst
    recBullBOS := false
    recBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low - atr*0.5, "🔨 BOS ↑", color=color.new(color.green,20), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index-1, swHigh, bar_index+10, swHigh, color=color.new(color.green,40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr*0.5, "🔨 BOS ↓", color=color.new(color.red,20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index-1, swLow, bar_index+10, swLow, color=color.new(color.red,40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweep = low < d1L and close > d1L and (d1L - low) >= atr * 0.3
bearSweep = high > d1H and close < d1H and (high - d1H) >= atr * 0.3

var float lastBullSwL = na
var float lastBearSwH = na
if bullSweep
    lastBullSwL := low
if bearSweep
    lastBearSwH := high

bullSwRej = bullSweep and (close - low)  > atr * 1.0
bearSwRej = bearSweep and (high - close) > atr * 1.0

plotshape(showSweeps and bullSweep, "Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime,0), size=size.small, text="Sweep")
plotshape(showSweeps and bearSweep, "Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,0),  size=size.small, text="Sweep")

intBullSw = low  < ta.lowest(low[1],   3)[1] and close > ta.lowest(low[1],   3)[1]
intBearSw = high > ta.highest(high[1], 3)[1] and close < ta.highest(high[1], 3)[1]
intBullOk = intBullSw and aboveE200
intBearOk = intBearSw and belowE200

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
imp3     = 3
bullImp  = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImp  = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBC  = bullImp and close[imp3] < open[imp3] and math.abs(close[imp3]-open[imp3]) > atr*0.5
bearOBC  = bearImp and close[imp3] > open[imp3] and math.abs(close[imp3]-open[imp3]) > atr*0.5

var float bOBT   = na
var float bOBB   = na
var float brOBT  = na
var float brOBB  = na
var bool  bOBAl  = false
var bool  brOBAl = false
var box   bOBBx  = na
var box   brOBBx = na
var bool  bOBRsp = false
var bool  brOBRsp= false
var int   bOBTch = 0
var int   brOBTch= 0

if bullOBC and aboveE200 and showOB
    bOBT   := high[imp3]
    bOBB   := low[imp3]
    bOBAl  := true
    bOBRsp := false
    bOBTch := 0
    box.delete(bOBBx)
    bOBBx := box.new(bar_index-imp3, bOBT, bar_index+100, bOBB, bgcolor=color.new(color.blue,80), border_color=color.new(color.blue,30), border_width=1)
    label.new(bar_index-imp3, bOBT, "🟦 Bull OB", color=color.new(color.blue,60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBC and belowE200 and showOB
    brOBT  := high[imp3]
    brOBB  := low[imp3]
    brOBAl := true
    brOBRsp:= false
    brOBTch:= 0
    box.delete(brOBBx)
    brOBBx := box.new(bar_index-imp3, brOBT, bar_index+100, brOBB, bgcolor=color.new(color.orange,80), border_color=color.new(color.orange,30), border_width=1)
    label.new(bar_index-imp3, brOBB, "🟧 Bear OB", color=color.new(color.orange,60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bOBAl and bOBTch == 0 and low <= bOBT and close >= bOBB
    bOBRsp := true
    bOBTch += 1
    box.set_bgcolor(bOBBx, color.new(color.green,70))
    box.set_border_color(bOBBx, color.new(color.green,20))
    label.new(bar_index, bOBB, "✅ Bull OB Entry", color=color.new(color.green,30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if brOBAl and brOBTch == 0 and high >= brOBB and close <= brOBT
    brOBRsp := true
    brOBTch += 1
    box.set_bgcolor(brOBBx, color.new(color.red,70))
    box.set_border_color(brOBBx, color.new(color.red,20))
    label.new(bar_index, brOBT, "✅ Bear OB Entry", color=color.new(color.red,30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bOBAl and close < bOBB
    bOBAl  := false
    bOBRsp := false
    box.set_bgcolor(bOBBx, color.new(color.gray,90))
if brOBAl and close > brOBT
    brOBAl  := false
    brOBRsp := false
    box.set_bgcolor(brOBBx, color.new(color.gray,90))

// ═══════════════════════════════════════════
// FVG
// ═══════════════════════════════════════════
cBody    = math.abs(close[1] - open[1])
strCdl   = cBody > atr * 1.0
dispCdl  = math.abs(close - open) > atr * 1.5
bFVGSz   = low - high[2]
brFVGSz  = low[2] - high
bFVGraw  = low > high[2] and strCdl and close[1] > open[1] and aboveE200 and bFVGSz  >= atr * fvgMin
brFVGraw = high < low[2] and strCdl and close[1] < open[1] and belowE200 and brFVGSz >= atr * fvgMin

var float bFVGT   = na
var float bFVGB   = na
var float brFVGT  = na
var float brFVGB  = na
var bool  bFVGAl  = false
var bool  brFVGAl = false
var box   bFVGBx  = na
var box   brFVGBx = na
var bool  bEZ     = false
var bool  brEZ    = false
var box   bEZBx   = na
var box   brEZBx  = na

if bFVGraw and showFVG
    bFVGT  := low
    bFVGB  := high[2]
    bFVGAl := true
    bEZ    := false
    box.delete(bFVGBx)
    bFVGBx := box.new(bar_index-2, bFVGT, bar_index+100, bFVGB, bgcolor=color.new(color.green,85), border_color=color.new(color.green,40), border_width=1)
    label.new(bar_index, bFVGB, "📦 Bull FVG", color=color.new(color.green,50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if brFVGraw and showFVG
    brFVGT  := low[2]
    brFVGB  := high
    brFVGAl := true
    brEZ    := false
    box.delete(brFVGBx)
    brFVGBx := box.new(bar_index-2, brFVGT, bar_index+100, brFVGB, bgcolor=color.new(color.red,85), border_color=color.new(color.red,40), border_width=1)
    label.new(bar_index, brFVGT, "📦 Bear FVG", color=color.new(color.red,50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bFVGAl and low <= bFVGT and close >= bFVGB
    box.set_bgcolor(bFVGBx, color.new(color.green,70))
    if dispCdl and aboveE200 and weekBull and dayBull and not bEZ
        bEZ := true
        box.delete(bEZBx)
        bEZBx := box.new(bar_index, bFVGT, bar_index+30, bFVGB, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, bFVGB, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_up, size=size.small)

if brFVGAl and high >= brFVGB and close <= brFVGT
    box.set_bgcolor(brFVGBx, color.new(color.red,70))
    if dispCdl and belowE200 and weekBear and dayBear and not brEZ
        brEZ := true
        box.delete(brEZBx)
        brEZBx := box.new(bar_index, brFVGT, bar_index+30, brFVGB, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, brFVGT, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_down, size=size.small)

if bFVGAl and close < bFVGB
    bFVGAl := false
    bEZ    := false
    box.set_bgcolor(bFVGBx, color.new(color.gray,90))
if brFVGAl and close > brFVGT
    brFVGAl := false
    brEZ    := false
    box.set_bgcolor(brFVGBx, color.new(color.gray,90))

// ═══════════════════════════════════════════
// RSI
// ═══════════════════════════════════════════
rsi      = ta.rsi(close, 14)
lkb      = 7
pLL      = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
pHL      = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
pHH      = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
pLH      = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)
rHL      = ta.lowest(rsi,   lkb) > ta.lowest(rsi[lkb],   lkb)
rLL      = ta.lowest(rsi,   lkb) < ta.lowest(rsi[lkb],   lkb)
rLH      = ta.highest(rsi,  lkb) < ta.highest(rsi[lkb],  lkb)
rHH      = ta.highest(rsi,  lkb) > ta.highest(rsi[lkb],  lkb)
rsiBullX = rsi < 40
rsiBearX = rsi > 60
sBullDiv = pLL and rHL
hBullDiv = pHL and rLL
sBearDiv = pHH and rLH
hBearDiv = pLH and rHH
anyBullD = sBullDiv or hBullDiv
anyBearD = sBearDiv or hBearDiv

if showDiv and sBullDiv
    label.new(bar_index, low  - atr*0.3, "◆ Strong Bull Div", color=color.new(color.green,0),  textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and hBullDiv
    label.new(bar_index, low  - atr*0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,0),   textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and sBearDiv
    label.new(bar_index, high + atr*0.3, "◆ Strong Bear Div", color=color.new(color.red,0),    textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hBearDiv
    label.new(bar_index, high + atr*0.3, "◇ Hidden Bear Div", color=color.new(color.orange,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdL, sigL, histL] = ta.macd(close, 12, 26, 9)
macdBull = macdL > sigL
macdBear = macdL < sigL

// ═══════════════════════════════════════════
// TIER 2 — STRUCTURE
// ═══════════════════════════════════════════
s2Long  = 0
s2Long  := s2Long  + (aboveE200  ? 1 : 0)
s2Long  := s2Long  + (recBullBOS ? 1 : 0)
s2Long  := s2Long  + (bullSweep  ? 1 : 0)
s2Long  := s2Long  + (inDisc     ? 1 : 0)
s2Long  := s2Long  + (mBullAlign ? 1 : 0)

s2Short = 0
s2Short := s2Short + (belowE200  ? 1 : 0)
s2Short := s2Short + (recBearBOS ? 1 : 0)
s2Short := s2Short + (bearSweep  ? 1 : 0)
s2Short := s2Short + (inPrem     ? 1 : 0)
s2Short := s2Short + (mBearAlign ? 1 : 0)

// ═══════════════════════════════════════════
// TIER 3 — 7 HIGH QUALITY ENTRY TRIGGERS
// ═══════════════════════════════════════════

// T1 — FVG Retracement
t1Bull = bEZ
t1Bear = brEZ

// T2 — Order Block Respected
t2Bull = bOBRsp
t2Bear = brOBRsp

// T3 — Sweep Rejection 1.0x ATR
t3Bull = bullSwRej
t3Bear = bearSwRej

// T4 — Pin Bar at Key Level
bullPinWick = low - math.min(open, close)
bearPinWick = math.max(open, close) - high
pinBody     = math.abs(close - open)
nearD1L     = math.abs(close - d1L) <= atr * 0.5
nearD1H     = math.abs(close - d1H) <= atr * 0.5
nearE200    = math.abs(close - ema200) <= atr * 0.5
nearBullOB  = bOBAl  and math.abs(close - bOBB)  <= atr * 0.5
nearBearOB  = brOBAl and math.abs(close - brOBT) <= atr * 0.5
nearBullFVG = bFVGAl  and close >= bFVGB  and close <= bFVGT
nearBearFVG = brFVGAl and close >= brFVGB and close <= brFVGT
atKeyBull   = nearD1L or nearE200 or nearBullOB or nearBullFVG
atKeyBear   = nearD1H or nearE200 or nearBearOB or nearBearFVG
t4Bull = bullPinWick >= 2.5 * pinBody and pinBody >= atr * 0.1 and aboveE200 and atKeyBull and weekBull and dayBull
t4Bear = bearPinWick >= 2.5 * pinBody and pinBody >= atr * 0.1 and belowE200 and atKeyBear and weekBear and dayBear

if showSignals and t4Bull
    label.new(bar_index, low - atr*0.2, "📍 Bull Pin", color=color.new(color.green,40), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showSignals and t4Bear
    label.new(bar_index, high + atr*0.2, "📍 Bear Pin", color=color.new(color.red,40), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// T5 — Previous Day High or Low Break and Close
t5Bull = close > d1H and close[1] <= d1H and aboveE200 and weekBull and dayBull
t5Bear = close < d1L and close[1] >= d1L and belowE200 and weekBear and dayBear

if showSignals and t5Bull
    label.new(bar_index, low - atr*0.3, "📈 PDH Break", color=color.new(color.green,30), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showSignals and t5Bear
    label.new(bar_index, high + atr*0.3, "📉 PDL Break", color=color.new(color.red,30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// T6 — BOS with Retest
var bool bosRetestReady = false
var bool bosRetestDir   = false
var float bosRetestLvl  = na

if bosBullFired
    bosRetestReady := true
    bosRetestDir   := true
    bosRetestLvl   := bosLevel
if bosBearFired
    bosRetestReady := true
    bosRetestDir   := false
    bosRetestLvl   := bosLevel

bosRetestTouch = bosRetestReady and not na(bosRetestLvl) and math.abs(close - bosRetestLvl) <= atr * 0.3
t6Bull = bosRetestTouch and bosRetestDir      and close > open and aboveE200
t6Bear = bosRetestTouch and not bosRetestDir  and close < open and belowE200

if t6Bull or t6Bear
    bosRetestReady := false

// T7 — Change of Character
lh1    = high[1] < high[2] and high[2] < high[3] and high[3] < high[4]
hl1    = low[1]  > low[2]  and low[2]  > low[3]  and low[3]  > low[4]
t7Bull = lh1 and high > high[1] and aboveE200 and weekBull and dayBull
t7Bear = hl1 and low  < low[1]  and belowE200 and weekBear and dayBear

if showSignals and t7Bull
    label.new(bar_index, low - atr*0.4, "🔄 CHoCH Bull", color=color.new(color.green,20), textcolor=color.white, style=label.style_label_up, size=size.tiny)
if showSignals and t7Bear
    label.new(bar_index, high + atr*0.4, "🔄 CHoCH Bear", color=color.new(color.red,20), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// Combine all triggers
anyTrigBull = t1Bull or t2Bull or t3Bull or t4Bull or t5Bull or t6Bull or t7Bull
anyTrigBear = t1Bear or t2Bear or t3Bear or t4Bear or t5Bear or t6Bear or t7Bear

tStrBull = t1Bull ? "🎯 FVG Retracement" : t2Bull ? "🟦 OB Respected" : t3Bull ? "💧 Sweep Rejection" : t4Bull ? "📍 Pin Bar" : t5Bull ? "📈 PDH Break" : t6Bull ? "🔨 BOS Retest" : t7Bull ? "🔄 CHoCH" : "⏳ None"
tStrBear = t1Bear ? "🎯 FVG Retracement" : t2Bear ? "🟧 OB Respected" : t3Bear ? "💧 Sweep Rejection" : t4Bear ? "📍 Pin Bar" : t5Bear ? "📉 PDL Break" : t6Bear ? "🔨 BOS Retest" : t7Bear ? "🔄 CHoCH" : "⏳ None"

// ═══════════════════════════════════════════
// TIER 4 — BONUS
// ═══════════════════════════════════════════
bonL = 0
bonL := bonL + (sBullDiv  ? 2 : hBullDiv ? 1 : 0)
bonL := bonL + (macdBull  ? 1 : 0)
bonL := bonL + (rsiBullX  ? 1 : 0)
bonL := bonL + (trending  ? 1 : 0)
bonL := bonL + (intBullOk ? 1 : 0)
bonL := bonL + (eqlSwept  ? 1 : 0)

bonS = 0
bonS := bonS + (sBearDiv  ? 2 : hBearDiv ? 1 : 0)
bonS := bonS + (macdBear  ? 1 : 0)
bonS := bonS + (rsiBearX  ? 1 : 0)
bonS := bonS + (trending  ? 1 : 0)
bonS := bonS + (intBearOk ? 1 : 0)
bonS := bonS + (eqhSwept  ? 1 : 0)

// ═══════════════════════════════════════════
// SIGNAL FIRING
// ═══════════════════════════════════════════
g1L  = weekBull
g2L  = dayBull
g3L  = htfBullOk
g4L  = activeSess
g5L  = not extBear
allL = g1L and g2L and g3L and g4L and g5L

g1S  = weekBear
g2S  = dayBear
g3S  = htfBearOk
g4S  = activeSess
g5S  = not extBear
allS = g1S and g2S and g3S and g4S and g5S

var int lastBullB = 0
var int lastBearB = 0

rawBull = allL and s2Long  >= minStr2 and anyTrigBull and bonL >= minBonus
rawBear = allS and s2Short >= minStr2 and anyTrigBear and bonS >= minBonus

bullSig = rawBull and (bar_index - lastBullB) > cooldown
bearSig = rawBear and (bar_index - lastBearB) > cooldown

if bullSig
    lastBullB := bar_index
if bearSig
    lastBearB := bar_index

// ═══════════════════════════════════════════
// SIGNAL QUALITY
// ═══════════════════════════════════════════
qLong  = bonL >= 7 ? "🔥 A+ HIGH CONFIDENCE" : bonL >= 5 ? "💪 STRONG" : "✅ VALID"
qShort = bonS >= 7 ? "🔥 A+ HIGH CONFIDENCE" : bonS >= 5 ? "💪 STRONG" : "✅ VALID"

// ═══════════════════════════════════════════
// RISK MANAGEMENT
// ═══════════════════════════════════════════
riskAmt  = accountSize * (riskPct / 100)
slBufL   = bonL >= 6 ? atr * 0.2 : atr * 0.5
slBufS   = bonS >= 6 ? atr * 0.2 : atr * 0.5
longSL   = not na(lastBullSwL) ? lastBullSwL - slBufL : close - atr * 1.5
longRisk = math.max(close - longSL, 0.0001)
longTP   = close + longRisk * 3
longSz   = riskAmt / longRisk
shortSL  = not na(lastBearSwH) ? lastBearSwH + slBufS : close + atr * 1.5
shortRsk = math.max(shortSL - close, 0.0001)
shortTP  = close - shortRsk * 3
shortSz  = riskAmt / shortRsk

// ═══════════════════════════════════════════
// SIGNAL LABELS
// ═══════════════════════════════════════════
lLbl =
     "⬆ BUY — " + tfLbl + " — " + qLong + "\n" +
     "Mode: " + modeStr + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "GATES: Wk🟢 D🟢 HTF🟢 Sess✅\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "STRUCTURE (" + str.tostring(s2Long) + "/5):\n" +
     (aboveE200  ? "✅" : "❌") + " EMA  " + (recBullBOS ? "✅" : "❌") + " BOS  " + (bullSweep ? "✅" : "❌") + " Sweep\n" +
     (inDisc     ? "✅" : "❌") + " Discount  " + (mBullAlign ? "✅" : "❌") + " Monthly\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRIGGER: " + tStrBull + "\n" +
     "BONUS: " + str.tostring(bonL) + "/7\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL: " + str.tostring(math.round(longSL,2)) + "\n" +
     "🟢 TP: " + str.tostring(math.round(longTP,2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmt,2)) + "  Target: $" + str.tostring(math.round(riskAmt*3,2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(longSz,4)) + " units  RR 1:3"

sLbl =
     "⬇ SELL — " + tfLbl + " — " + qShort + "\n" +
     "Mode: " + modeStr + "\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "GATES: Wk🔴 D🔴 HTF🔴 Sess✅\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "STRUCTURE (" + str.tostring(s2Short) + "/5):\n" +
     (belowE200  ? "✅" : "❌") + " EMA  " + (recBearBOS ? "✅" : "❌") + " BOS  " + (bearSweep ? "✅" : "❌") + " Sweep\n" +
     (inPrem     ? "✅" : "❌") + " Premium  " + (mBearAlign ? "✅" : "❌") + " Monthly\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "TRIGGER: " + tStrBear + "\n" +
     "BONUS: " + str.tostring(bonS) + "/7\n" +
     "━━━━━━━━━━━━━━━━━━━\n" +
     "🔴 SL: " + str.tostring(math.round(shortSL,2)) + "\n" +
     "🟢 TP: " + str.tostring(math.round(shortTP,2)) + "\n" +
     "💰 Risk: $" + str.tostring(math.round(riskAmt,2)) + "  Target: $" + str.tostring(math.round(riskAmt*3,2)) + "\n" +
     "📊 Size: " + str.tostring(math.round(shortSz,4)) + " units  RR 1:3"

if showSignals and bullSig
    label.new(bar_index, low,  lLbl, color=color.new(color.green,10), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index, longSL, bar_index+40, longSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP, bar_index+40, longTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, longSL, "🔴 SL " + str.tostring(math.round(longSL,2)), color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, longTP, "🟢 TP " + str.tostring(math.round(longTP,2)), color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearSig
    label.new(bar_index, high, sLbl, color=color.new(color.red,10),   textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index+40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index+40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, shortSL, "🔴 SL " + str.tostring(math.round(shortSL,2)), color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, shortTP, "🟢 TP " + str.tostring(math.round(shortTP,2)), color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// ALERTS
// ═══════════════════════════════════════════
alertcondition(bullSig, title="🟢 GOLD BUY — High Confidence",  message="⬆ BUY SIGNAL XAUUSD — Check chart")
alertcondition(bearSig, title="🔴 GOLD SELL — High Confidence", message="⬇ SELL SIGNAL XAUUSD — Check chart")
alertcondition(bullSig or bearSig, title="🚦 GOLD Signal",      message="Gold signal fired on XAUUSD — Check chart")

// ═══════════════════════════════════════════
// WAIT REASON
// ═══════════════════════════════════════════
wr1L = not g1L ? "⏳ WAIT — Weekly Bias Bearish (4H below EMA200)"    : ""
wr2L = not g2L ? "⏳ WAIT — Daily Bias Bearish (1H below EMA200)"     : ""
wr3L = not g3L ? "⏳ WAIT — HTF " + htfNeeded + " Not Bullish"        : ""
wr4L = not g4L ? "⏳ WAIT — Asian Session Blocked"                     : ""
wr5L = not g5L ? "⏳ WAIT — Extreme Bear Blocked"                      : ""
wr6L = s2Long < minStr2 ? "⏳ WAIT — Need " + str.tostring(minStr2 - s2Long) + " More Structure"  : ""
wr7L = not anyTrigBull  ? "⏳ WAIT — Need Entry Trigger"               : ""
wr8L = bonL < minBonus  ? "⏳ WAIT — Need " + str.tostring(minBonus - bonL) + " More Bonus Pts"   : ""
waitL = not g1L ? wr1L : not g2L ? wr2L : not g3L ? wr3L : not g4L ? wr4L : not g5L ? wr5L : s2Long < minStr2 ? wr6L : not anyTrigBull ? wr7L : bonL < minBonus ? wr8L : "✅ LONG CONDITIONS MET"

wr1S = not g1S ? "⏳ WAIT — Weekly Bias Bullish (4H above EMA200)"    : ""
wr2S = not g2S ? "⏳ WAIT — Daily Bias Bullish (1H above EMA200)"     : ""
wr3S = not g3S ? "⏳ WAIT — HTF " + htfNeeded + " Not Bearish"        : ""
wr4S = not g4S ? "⏳ WAIT — Asian Session Blocked"                     : ""
wr5S = not g5S ? "⏳ WAIT — Extreme Bear Blocked"                      : ""
wr6S = s2Short < minStr2 ? "⏳ WAIT — Need " + str.tostring(minStr2 - s2Short) + " More Structure" : ""
wr7S = not anyTrigBear   ? "⏳ WAIT — Need Entry Trigger"              : ""
wr8S = bonS < minBonus   ? "⏳ WAIT — Need " + str.tostring(minBonus - bonS) + " More Bonus Pts"   : ""
waitS = not g1S ? wr1S : not g2S ? wr2S : not g3S ? wr3S : not g4S ? wr4S : not g5S ? wr5S : s2Short < minStr2 ? wr6S : not anyTrigBear ? wr7S : bonS < minBonus ? wr8S : "✅ SHORT CONDITIONS MET"

// ═══════════════════════════════════════════
// DASHBOARD
// ═══════════════════════════════════════════
var table dash = table.new(position.top_right, 3, 42, bgcolor=color.new(color.black,50), border_color=color.new(color.gray,30), border_width=1)
hBG  = color.new(color.blue,40)
sBG  = color.new(color.gray,60)
sCol = color.yellow

table.cell(dash, 0, 0, "📊 CONDITION",  text_color=color.white,  text_size=size.small, bgcolor=hBG)
table.cell(dash, 1, 0, "STATUS",        text_color=color.white,  text_size=size.small, bgcolor=hBG)
table.cell(dash, 2, 0, "🥇 " + tfLbl,  text_color=color.yellow, text_size=size.small, bgcolor=hBG)

table.cell(dash, 0, 1, "── MARKET ──",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 1, "────────────",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 1, "──────",        text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 2, "Regime",        text_color=color.white, text_size=size.small)
table.cell(dash, 1, 2, regStr,          text_color=regCol,      text_size=size.small)
table.cell(dash, 2, 2, extBear ? "🚫" : "✅", text_color=extBear ? color.red : color.green, text_size=size.small)
table.cell(dash, 0, 3, "ADX",           text_color=color.white, text_size=size.small)
table.cell(dash, 1, 3, adxStr,          text_color=adxCol,      text_size=size.small)
table.cell(dash, 2, 3, trending ? "✅" : "❌", text_color=trending ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 4, "Session",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 4, sessName,        text_color=sessCol,     text_size=size.small)
table.cell(dash, 2, 4, activeSess ? "✅" : "❌", text_color=activeSess ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 5, "TF Mode",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 5, modeStr,         text_color=color.aqua,  text_size=size.small)
table.cell(dash, 2, 5, "B≥" + str.tostring(minBonus) + " S≥" + str.tostring(minStr2), text_color=color.aqua, text_size=size.tiny)
table.cell(dash, 0, 6, "HTF " + htfNeeded, text_color=color.white, text_size=size.small)
table.cell(dash, 1, 6, htfStr, text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.tiny)
table.cell(dash, 2, 6, htfBullOk ? "✅🟢" : htfBearOk ? "✅🔴" : "❌", text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.small)

table.cell(dash, 0, 7, "── TIER 1 GATES ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 7, "────────────────",   text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 7, "──────",             text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 8, "Weekly Bias ⚡",    text_color=color.white, text_size=size.small)
table.cell(dash, 1, 8, wkBiasStr, text_color=weekBull ? color.green : weekBear ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 8, weekBull or weekBear ? "✅" : "❌", text_color=weekBull or weekBear ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 9, "Daily Bias 🟢",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 9, dBiasStr, text_color=dayBull ? color.green : dayBear ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 9, dayBull or dayBear ? "✅" : "❌", text_color=dayBull or dayBear ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 10, "HTF Aligned",      text_color=color.white, text_size=size.small)
table.cell(dash, 1, 10, htfStr, text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.tiny)
table.cell(dash, 2, 10, htfBullOk or htfBearOk ? "✅" : "❌", text_color=htfBullOk or htfBearOk ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 11, "Session Gate",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 11, sessName,           text_color=sessCol,     text_size=size.small)
table.cell(dash, 2, 11, activeSess ? "✅" : "❌", text_color=activeSess ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 12, "Extreme Bear",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 12, extBear ? "🚨 BLOCKED" : "✅ SAFE", text_color=extBear ? color.red : color.green, text_size=size.small)
table.cell(dash, 2, 12, extBear ? "❌" : "✅", text_color=extBear ? color.red : color.green, text_size=size.small)

table.cell(dash, 0, 13, "── TIER 2 STRUCTURE ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 13, "────────────────────",   text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 13, "──────",                 text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 14, "EMA 200",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 14, aboveE200 ? "🟢 BULLISH" : "🔴 BEARISH", text_color=aboveE200 ? color.green : color.red, text_size=size.small)
table.cell(dash, 2, 14, "✅", text_color=color.green, text_size=size.small)
table.cell(dash, 0, 15, "BOS",           text_color=color.white, text_size=size.small)
table.cell(dash, 1, 15, recBullBOS ? "🟢 BULL" : recBearBOS ? "🔴 BEAR" : "⏳ NONE", text_color=recBullBOS ? color.green : recBearBOS ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 15, recBullBOS or recBearBOS ? "✅" : "❌", text_color=recBullBOS or recBearBOS ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 16, "Sweep",         text_color=color.white, text_size=size.small)
table.cell(dash, 1, 16, bullSweep ? "🟢 BULL" : bearSweep ? "🔴 BEAR" : "⏳ NONE", text_color=bullSweep or bearSweep ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 16, bullSweep or bearSweep ? "✅" : "❌", text_color=bullSweep or bearSweep ? color.green : color.red, text_size=size.small)
table.cell(dash, 0, 17, "Zone",          text_color=color.white, text_size=size.small)
table.cell(dash, 1, 17, inDisc ? "🟢 DISCOUNT" : inPrem ? "🔴 PREMIUM" : "⚪ EQUILIB", text_color=inDisc ? color.green : inPrem ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 17, inDisc or inPrem ? "✅" : "➖", text_color=inDisc or inPrem ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 18, "Monthly",       text_color=color.white, text_size=size.small)
table.cell(dash, 1, 18, mBullAlign ? "🟢 BULL" : mBearAlign ? "🔴 BEAR" : "⏳ WAIT", text_color=mBullAlign ? color.green : mBearAlign ? color.red : color.yellow, text_size=size.small)
table.cell(dash, 2, 18, mBullAlign or mBearAlign ? "✅" : "❌", text_color=mBullAlign or mBearAlign ? color.green : color.red, text_size=size.small)
s2disp = belowE200 ? s2Short : s2Long
table.cell(dash, 0, 19, "Str Score",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 19, str.tostring(s2disp) + "/5 (need " + str.tostring(minStr2) + ")", text_color=s2disp >= minStr2 ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 19, s2disp >= minStr2 ? "✅" : "❌", text_color=s2disp >= minStr2 ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 20, "── TIER 3 TRIGGERS ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 20, "────────────────────",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 20, "──────",                text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 21, "T1 FVG Zone",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 21, t1Bull or t1Bear ? "🎯 ACTIVE" : "⏳ NONE", text_color=t1Bull or t1Bear ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 21, t1Bull or t1Bear ? "✅" : "➖", text_color=t1Bull or t1Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 22, "T2 OB Respect", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 22, t2Bull ? "✅ BULL" : t2Bear ? "✅ BEAR" : bOBAl or brOBAl ? "⏳ LIVE" : "⏳ NONE", text_color=t2Bull or t2Bear ? color.green : bOBAl or brOBAl ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 22, t2Bull or t2Bear ? "✅" : "➖", text_color=t2Bull or t2Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 23, "T3 Sw Reject",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 23, t3Bull ? "🟢 BULL" : t3Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t3Bull or t3Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 23, t3Bull or t3Bear ? "✅" : "➖", text_color=t3Bull or t3Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 24, "T4 Pin Bar",    text_color=color.white, text_size=size.small)
table.cell(dash, 1, 24, t4Bull ? "🟢 BULL" : t4Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t4Bull or t4Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 24, t4Bull or t4Bear ? "✅" : "➖", text_color=t4Bull or t4Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 25, "T5 PDH/L Break",text_color=color.white, text_size=size.small)
table.cell(dash, 1, 25, t5Bull ? "🟢 BULL" : t5Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t5Bull or t5Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 25, t5Bull or t5Bear ? "✅" : "➖", text_color=t5Bull or t5Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 26, "T6 BOS Retest", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 26, t6Bull ? "🟢 BULL" : t6Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t6Bull or t6Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 26, t6Bull or t6Bear ? "✅" : "➖", text_color=t6Bull or t6Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 27, "T7 CHoCH",      text_color=color.white, text_size=size.small)
table.cell(dash, 1, 27, t7Bull ? "🟢 BULL" : t7Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t7Bull or t7Bear ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 27, t7Bull or t7Bear ? "✅" : "➖", text_color=t7Bull or t7Bear ? color.green : color.gray, text_size=size.small)
trigActive = anyTrigBull or anyTrigBear
table.cell(dash, 0, 28, "Active Trigger",text_color=color.white, text_size=size.small)
table.cell(dash, 1, 28, trigActive ? (belowE200 ? tStrBear : tStrBull) : "⏳ WAITING", text_color=trigActive ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 28, trigActive ? "✅" : "❌", text_color=trigActive ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 29, "── TIER 4 BONUS ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 29, "──────────────────", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 29, "──────",             text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 30, "RSI Divergence",text_color=color.white, text_size=size.small)
table.cell(dash, 1, 30, sBullDiv ? "🟢 Strong(+2)" : hBullDiv ? "🟢 Hidden(+1)" : sBearDiv ? "🔴 Strong(+2)" : hBearDiv ? "🔴 Hidden(+1)" : "⏳ NONE", text_color=anyBullD ? color.green : anyBearD ? color.red : color.gray, text_size=size.small)
table.cell(dash, 2, 30, anyBullD or anyBearD ? "✅" : "➖", text_color=anyBullD or anyBearD ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 31, "MACD",          text_color=color.white, text_size=size.small)
table.cell(dash, 1, 31, macdBull ? "🟢 BULL" : "🔴 BEAR", text_color=macdBull ? color.green : color.red, text_size=size.small)
table.cell(dash, 2, 31, "✅", text_color=color.green, text_size=size.small)
table.cell(dash, 0, 32, "RSI Level",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 32, str.tostring(math.round(rsi,1)) + (rsiBullX ? " 🟢 OS" : rsiBearX ? " 🔴 OB" : " ⚪ N"), text_color=rsiBullX ? color.green : rsiBearX ? color.red : color.white, text_size=size.small)
table.cell(dash, 2, 32, rsiBullX or rsiBearX ? "✅" : "➖", text_color=rsiBullX or rsiBearX ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 33, "Int Sweep",     text_color=color.white, text_size=size.small)
table.cell(dash, 1, 33, intBullOk ? "🟢 BULL" : intBearOk ? "🔴 BEAR" : "⏳ NONE", text_color=intBullOk or intBearOk ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 33, intBullOk or intBearOk ? "✅" : "➖", text_color=intBullOk or intBearOk ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 34, "EQH/EQL Swept", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 34, eqhSwept ? "🔴 EQH" : eqlSwept ? "🟢 EQL" : "⏳ NONE", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 34, eqhSwept or eqlSwept ? "✅" : "➖", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)
bonDisp = belowE200 ? bonS : bonL
table.cell(dash, 0, 35, "Bonus Score",   text_color=color.white, text_size=size.small)
table.cell(dash, 1, 35, str.tostring(bonDisp) + "/7 (need " + str.tostring(minBonus) + ")", text_color=bonDisp >= minBonus ? color.green : color.yellow, text_size=size.small)
table.cell(dash, 2, 35, bonDisp >= minBonus ? "✅" : "❌", text_color=bonDisp >= minBonus ? color.green : color.red, text_size=size.small)

table.cell(dash, 0, 36, "── RISK ──",   text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 36, "────────────", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 36, "──────",       text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 37, "Risk/Target",  text_color=color.white, text_size=size.small)
table.cell(dash, 1, 37, "💰 $" + str.tostring(math.round(riskAmt,2)) + " → 🎯 $" + str.tostring(math.round(riskAmt*3,2)) + " (1:3)", text_color=color.yellow, text_size=size.small)
table.cell(dash, 2, 37, "1:3", text_color=color.yellow, text_size=size.small)

table.cell(dash, 0, 38, "── BIAS LINES ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 1, 38, "────────────────", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 2, 38, "──────",           text_color=sCol, text_size=size.tiny, bgcolor=sBG)
table.cell(dash, 0, 39, "⚡ Yellow Lines", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 39, not na(wbFinalH) ? "H:" + str.tostring(math.round(wbFinalH,2)) + " L:" + str.tostring(math.round(wbFinalL,2)) : "⏳ Waiting for Monday", text_color=not na(wbFinalH) ? color.yellow : color.gray, text_size=size.small)
table.cell(dash, 2, 39, not na(wbFinalH) ? "✅" : "⏳", text_color=not na(wbFinalH) ? color.green : color.gray, text_size=size.small)
table.cell(dash, 0, 40, "🟢 Green Lines", text_color=color.white, text_size=size.small)
table.cell(dash, 1, 40, not na(dayBiasH) ? "H:" + str.tostring(math.round(dayBiasH,2)) + " L:" + str.tostring(math.round(dayBiasL,2)) : "⏳ Waiting for 1AM candle", text_color=not na(dayBiasH) ? color.green : color.gray, text_size=size.small)
table.cell(dash, 2, 40, not na(dayBiasH) ? "✅" : "⏳", text_color=not na(dayBiasH) ? color.green : color.gray, text_size=size.small)

waitStr = belowE200 ? waitS : waitL
sigStr  = bullSig ? "⬆ BUY " + tfLbl + " — " + qLong : bearSig ? "⬇ SELL " + tfLbl + " — " + qShort : waitStr
sigCol  = bullSig ? color.green : bearSig ? color.red : color.yellow
sig3    = bullSig ? "🟢 BUY" : bearSig ? "🔴 SELL" : "⏳ WAIT"
sig3C   = bullSig ? color.green : bearSig ? color.red : color.yellow
table.cell(dash, 0, 41, "🚦 SIGNAL",    text_color=color.white, text_size=size.normal, bgcolor=color.new(color.gray,30))
table.cell(dash, 1, 41, sigStr,         text_color=sigCol,      text_size=size.small)
table.cell(dash, 2, 41, sig3,           text_color=sig3C,       text_size=size.normal)
```

```python
// ============================================================
// Manny's Gold Strategy V2 — STEVE BUILD FINAL
// Pine Script v6 — XAUUSD Only
// Chart timezone must be set to UTC
// ============================================================
//@version=6
indicator("Manny's Gold Strategy V2", overlay=true, max_lines_count=500, max_labels_count=500, max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
compactDash  = input.bool(false, "📱 Compact Dashboard (Phone Mode)", group="📊 Dashboard")
dashPos      = input.string("Top Right", "Dashboard Position",
     options=["Top Right", "Bottom Right", "Top Left", "Bottom Left"],
     group="📊 Dashboard")
showMonthly  = input.bool(true,  "📅 Monthly High/Low",           group="📊 Chart Levels")
showWeekly   = input.bool(true,  "📅 Weekly High/Low",            group="📊 Chart Levels")
showDaily    = input.bool(true,  "📅 Daily High/Low (3 Days)",    group="📊 Chart Levels")
showEQL      = input.bool(true,  "📊 Equal Highs and Lows",       group="📊 Chart Levels")
showZone     = input.bool(true,  "📊 Premium/Discount Zone",      group="📊 Chart Levels")
showYellow   = input.bool(true,  "⚡ Monday Bias Lines (Yellow)", group="🧭 Bias Visuals")
showGreen    = input.bool(true,  "🟢 Daily 1AM Lines (Green)",    group="🧭 Bias Visuals")
goldBiasTime = input.string("1AM / 5AM (Standard Time)", "🥇 Bias Candle Time",
     options=["1AM / 5AM (Standard Time)", "2AM / 6AM (Daylight Saving)"],
     group="🧭 Bias Visuals",
     tooltip="Standard = winter. Daylight Saving = summer. Switch manually when clocks change.")
showEMA      = input.bool(true,  "📉 EMA 200 + EMA 50",           group="🔧 Features")
showSweeps   = input.bool(true,  "💧 Liquidity Sweeps",           group="🔧 Features")
showFVG      = input.bool(true,  "📦 Fair Value Gaps",            group="🔧 Features")
showDiv      = input.bool(true,  "📐 RSI Divergence",             group="🔧 Features")
showBOS      = input.bool(true,  "🔨 Break of Structure",         group="🔧 Features")
showOB       = input.bool(true,  "🟦 Order Blocks",               group="🔧 Features")
showSignals  = input.bool(true,  "🚦 Buy/Sell Signals",           group="🔧 Features")
fvgMinIn     = input.float(0.5,  "FVG Min Size (x ATR)",          group="🔧 Features", minval=0.1)
modeOvr      = input.string("Auto", "Trading Mode Override",
     options=["Auto","Scalping","Day Trade","Swing Trade"],      group="🎯 Signal Quality")
minBonusOvr  = input.int(0,      "Min Bonus Override (0=auto)",   group="🎯 Signal Quality", minval=0, maxval=7)
adxLen       = input.int(14,     "ADX Period",                    group="📊 Market Filter")
adxThr       = input.int(25,     "ADX Minimum",                   group="📊 Market Filter")
accountSize  = input.float(1000, "💰 Account Size ($)",           group="💵 Risk")
riskPct      = input.float(1.0,  "⚠️ Risk Per Trade (%)",        group="💵 Risk")
atrLen       = input.int(14,     "ATR Period",                    group="💵 Risk")

// ═══════════════════════════════════════════
// DASHBOARD POSITION
// ═══════════════════════════════════════════
dashPosVal = dashPos == "Top Right"    ? position.top_right    :
             dashPos == "Bottom Right" ? position.bottom_right :
             dashPos == "Top Left"     ? position.top_left     :
             position.bottom_left

// ═══════════════════════════════════════════
// BIAS TIME SETTINGS
// ═══════════════════════════════════════════
isStandard   = goldBiasTime == "1AM / 5AM (Standard Time)"
biasHour1    = isStandard ? 1 : 2
biasHour2    = isStandard ? 5 : 6
biasLbl      = isStandard ? "⚡ 1AM+5AM" : "⚡ 2AM+6AM"
dailyLbl     = isStandard ? "🟢 1AM" : "🟢 2AM"

// ═══════════════════════════════════════════
// TIMEFRAME DETECTION
// ═══════════════════════════════════════════
tfMins      = timeframe.in_seconds() / 60
isScalpTF   = tfMins <= 5
isDayTF     = tfMins > 5   and tfMins <= 30
isSwingTF   = tfMins > 30  and tfMins <= 240
autoStr     = isScalpTF ? "Scalping" : isDayTF ? "Day Trade" : isSwingTF ? "Swing Trade" : "Long Term"
isScalp     = modeOvr == "Auto" ? isScalpTF  : modeOvr == "Scalping"
isSwing     = modeOvr == "Auto" ? isSwingTF  : modeOvr == "Swing Trade"
modeStr     = modeOvr == "Auto" ? autoStr + " (Auto)" : modeOvr
tfLbl       = tfMins == 1 ? "1M" : tfMins == 5 ? "5M" : tfMins == 15 ? "15M" : tfMins == 30 ? "30M" : tfMins == 60 ? "1H" : tfMins == 240 ? "4H" : tfMins == 1440 ? "D" : str.tostring(math.round(tfMins)) + "M"
minBonus    = minBonusOvr > 0 ? minBonusOvr : isScalp ? 2 : isSwing ? 4 : 3
bosRst      = isScalp ? 50 : isSwing ? 20 : 30
fvgMin      = isScalp ? 0.3 : isSwing ? 0.7 : fvgMinIn
minStr2     = isScalp ? 2 : isSwing ? 3 : 2
cooldown    = tfMins <= 1 ? 15 : tfMins <= 5 ? 12 : tfMins <= 15 ? 10 : tfMins <= 30 ? 8 : 5

// ═══════════════════════════════════════════
// TIME
// ═══════════════════════════════════════════
gmtH         = hour(time, "GMT+0")
gmtDow       = dayofweek(time, "GMT+0")
isNewDay     = ta.change(time("D")) != 0
isNewWeek    = ta.change(time("W")) != 0
isNewMonth   = ta.change(time("M")) != 0
isNewISOWeek = ta.change(weekofyear(time, "GMT+0")) != 0

// ═══════════════════════════════════════════
// SESSION
// ═══════════════════════════════════════════
london     = gmtH >= 7  and gmtH < 12
ny         = gmtH >= 13 and gmtH < 17
overlap    = gmtH >= 12 and gmtH < 13
activeSess = london or ny or overlap
sessName   = overlap ? "⚡ Overlap" : london ? "🇬🇧 London" : ny ? "🇺🇸 New York" : "😴 Asian — BLOCKED"
sessCol    = overlap ? color.yellow : london ? color.green : ny ? color.blue : color.gray

// ═══════════════════════════════════════════
// ATR + EMA
// ═══════════════════════════════════════════
atr        = ta.atr(atrLen)
ema200     = ta.ema(close, 200)
ema50      = ta.ema(close, 50)
aboveE200  = close > ema200
belowE200  = close < ema200
bullMkt    = ema50 > ema200
pctBelow   = (ema200 - close) / ema200 * 100
extBear    = pctBelow > 20
regStr     = extBear ? "🚨 EXTREME BEAR" : bullMkt ? "🟢 BULL MARKET" : "🔴 BEAR MARKET"
regCol     = extBear ? color.red : bullMkt ? color.green : color.red
plot(showEMA ? ema200 : na, "EMA 200", color=aboveE200 ? color.new(color.green,0) : color.new(color.red,0), linewidth=2)
plot(showEMA ? ema50  : na, "EMA 50",  color=bullMkt   ? color.new(color.green,50) : color.new(color.red,50), linewidth=1)

// ═══════════════════════════════════════════
// ADX
// ═══════════════════════════════════════════
[diP, diM, adxVal] = ta.dmi(adxLen, adxLen)
trending   = adxVal >= adxThr
adxStr     = trending ? "📈 TRENDING (" + str.tostring(math.round(adxVal,1)) + ")" : "📉 RANGING (" + str.tostring(math.round(adxVal,1)) + ")"
adxCol     = trending ? color.green : color.red

// ═══════════════════════════════════════════
// HTF EMA 200
// ═══════════════════════════════════════════
c1H    = request.security(syminfo.tickerid, "60",  close,              lookahead=barmerge.lookahead_off)
e1H    = request.security(syminfo.tickerid, "60",  ta.ema(close, 200), lookahead=barmerge.lookahead_off)
c4H    = request.security(syminfo.tickerid, "240", close,              lookahead=barmerge.lookahead_off)
e4H    = request.security(syminfo.tickerid, "240", ta.ema(close, 200), lookahead=barmerge.lookahead_off)
cD     = request.security(syminfo.tickerid, "D",   close,              lookahead=barmerge.lookahead_off)
eD     = request.security(syminfo.tickerid, "D",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)
cW     = request.security(syminfo.tickerid, "W",   close,              lookahead=barmerge.lookahead_off)
eW     = request.security(syminfo.tickerid, "W",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)

bull1H = c1H > e1H
bull4H = c4H > e4H
bullD  = cD  > eD
bullW  = cW  > eW
bear1H = c1H < e1H
bear4H = c4H < e4H
bearD  = cD  < eD
bearW  = cW  < eW

// ═══════════════════════════════════════════
// GATE 1 — WEEKLY BIAS via 4H EMA 200
// ═══════════════════════════════════════════
weekBull  = bull4H
weekBear  = bear4H
wkBiasStr = weekBull ? "🟢 BULL 4H EMA" : weekBear ? "🔴 BEAR 4H EMA" : "⏳ FLAT"

// ═══════════════════════════════════════════
// GATE 2 — DAILY BIAS via 1H EMA 200
// ═══════════════════════════════════════════
dayBull   = bull1H
dayBear   = bear1H
dBiasStr  = dayBull ? "🟢 BULL 1H EMA" : dayBear ? "🔴 BEAR 1H EMA" : "⏳ FLAT"

// ═══════════════════════════════════════════
// GATE 3 — HTF CONFIRMATION
// ═══════════════════════════════════════════
htfBullOk = isScalpTF ? (bull1H and bull4H and bullD) : isDayTF ? (bull1H and bull4H) : isSwingTF ? bull4H : (bullD and bullW)
htfBearOk = isScalpTF ? (bear1H and bear4H and bearD) : isDayTF ? (bear1H and bear4H) : isSwingTF ? bear4H : (bearD and bearW)
htfNeeded = isScalpTF ? "1H+4H+D" : isDayTF ? "1H+4H" : isSwingTF ? "4H" : "D+W"
e1He      = bull1H ? "🟢" : "🔴"
e4He      = bull4H ? "🟢" : "🔴"
eDe       = bullD  ? "🟢" : "🔴"
eWe       = bullW  ? "🟢" : "🔴"
htfStr    = htfBullOk ? "🟢 " + htfNeeded + " ✓" : htfBearOk ? "🔴 " + htfNeeded + " ✓" : "⏳ 1H:" + e1He + " 4H:" + e4He + " D:" + eDe + " W:" + eWe

// ═══════════════════════════════════════════
// MONTHLY LEVELS
// ═══════════════════════════════════════════
mH = ta.valuewhen(isNewMonth, high[1], 0)
mL = ta.valuewhen(isNewMonth, low[1],  0)
var line mHL = na
var line mLL = na
if isNewMonth and showMonthly
    line.delete(mHL)
    line.delete(mLL)
    mHL := line.new(bar_index, mH, bar_index+100, mH, color=color.new(color.purple,0), width=2)
    mLL := line.new(bar_index, mL, bar_index+100, mL, color=color.new(color.purple,0), width=2)
    label.new(bar_index, mH, "M High", color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, mL, "M Low",  color=color.new(color.purple,70), textcolor=color.purple, style=label.style_label_left, size=size.tiny)
mBullAlign = close > mL and aboveE200
mBearAlign = close < mH and belowE200

// ═══════════════════════════════════════════
// WEEKLY LEVELS
// ═══════════════════════════════════════════
wH = ta.valuewhen(isNewWeek, high[1], 0)
wL = ta.valuewhen(isNewWeek, low[1],  0)
var line wHL = na
var line wLL = na
if isNewWeek and showWeekly
    line.delete(wHL)
    line.delete(wLL)
    wHL := line.new(bar_index, wH, bar_index+100, wH, color=color.new(color.blue,0), width=2)
    wLL := line.new(bar_index, wL, bar_index+100, wL, color=color.new(color.blue,0), width=2)
    label.new(bar_index, wH, "W High", color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, wL, "W Low",  color=color.new(color.blue,70), textcolor=color.blue, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// DAILY LEVELS — 3 days fading
// ═══════════════════════════════════════════
d1H = ta.valuewhen(isNewDay, high[1], 0)
d1L = ta.valuewhen(isNewDay, low[1],  0)
d2H = ta.valuewhen(isNewDay, high[1], 1)
d2L = ta.valuewhen(isNewDay, low[1],  1)
d3H = ta.valuewhen(isNewDay, high[1], 2)
d3L = ta.valuewhen(isNewDay, low[1],  2)
var line d1HL = na
var line d1LL = na
var line d2HL = na
var line d2LL = na
var line d3HL = na
var line d3LL = na
if isNewDay and showDaily
    line.delete(d1HL)
    line.delete(d1LL)
    line.delete(d2HL)
    line.delete(d2LL)
    line.delete(d3HL)
    line.delete(d3LL)
    d1HL := line.new(bar_index, d1H, bar_index+100, d1H, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d1LL := line.new(bar_index, d1L, bar_index+100, d1L, color=color.new(color.black,0),  width=2, style=line.style_solid)
    d2HL := line.new(bar_index, d2H, bar_index+100, d2H, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d2LL := line.new(bar_index, d2L, bar_index+100, d2L, color=color.new(color.black,40), width=1, style=line.style_dashed)
    d3HL := line.new(bar_index, d3H, bar_index+100, d3H, color=color.new(color.black,70), width=1, style=line.style_dotted)
    d3LL := line.new(bar_index, d3L, bar_index+100, d3L, color=color.new(color.black,70), width=1, style=line.style_dotted)
    label.new(bar_index, d1H, "D1H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d1L, "D1L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d2H, "D2H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d2L, "D2L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d3H, "D3H", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)
    label.new(bar_index, d3L, "D3L", color=color.new(color.gray,80), textcolor=color.black, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// EQUAL HIGHS AND LOWS
// ═══════════════════════════════════════════
sH      = ta.pivothigh(high, 5, 5)
sL      = ta.pivotlow(low,   5, 5)
psH     = ta.valuewhen(not na(sH), sH, 1)
psL     = ta.valuewhen(not na(sL), sL, 1)
isEQH   = not na(sH) and not na(psH) and math.abs(sH - psH) <= atr * 0.1
isEQL   = not na(sL) and not na(psL) and math.abs(sL - psL) <= atr * 0.1
if showEQL and isEQH
    line.new(bar_index-5, sH, bar_index+20, sH, color=color.new(color.red,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, sH, "EQH", color=color.new(color.red,70), textcolor=color.red, style=label.style_label_right, size=size.tiny)
if showEQL and isEQL
    line.new(bar_index-5, sL, bar_index+20, sL, color=color.new(color.green,30), style=line.style_dashed, width=1)
    label.new(bar_index-5, sL, "EQL", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_right, size=size.tiny)
eqhSwept = not na(sH) and high > sH and close < sH
eqlSwept = not na(sL) and low  < sL and close > sL

// ═══════════════════════════════════════════
// PREMIUM AND DISCOUNT ZONE
// ═══════════════════════════════════════════
wRange = wH - wL
wMid   = wL + wRange / 2
inDisc = close < wMid and not na(wMid)
inPrem = close > wMid and not na(wMid)
if showZone and isNewWeek and not na(wH) and not na(wL)
    line.new(bar_index, wMid, bar_index+100, wMid, color=color.new(color.gray,50), style=line.style_dashed, width=1)
    label.new(bar_index, wMid, "Equilibrium", color=color.new(color.gray,70), textcolor=color.gray, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// YELLOW LINES — 1AM and 5AM Monday 4H candles
// ═══════════════════════════════════════════
f_wbH1() =>
    isMon1 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    isMon1 ? high : float(na)
f_wbL1() =>
    isMon1 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    isMon1 ? low : float(na)
f_wbH2() =>
    isMon2 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour2 and minute(time, "GMT+0") == 0
    isMon2 ? high : float(na)
f_wbL2() =>
    isMon2 = dayofweek(time, "GMT+0") == 2 and hour(time, "GMT+0") == biasHour2 and minute(time, "GMT+0") == 0
    isMon2 ? low : float(na)

wb4H_H1 = request.security(syminfo.tickerid, "240", f_wbH1(), lookahead=barmerge.lookahead_off)
wb4H_L1 = request.security(syminfo.tickerid, "240", f_wbL1(), lookahead=barmerge.lookahead_off)
wb4H_H2 = request.security(syminfo.tickerid, "240", f_wbH2(), lookahead=barmerge.lookahead_off)
wb4H_L2 = request.security(syminfo.tickerid, "240", f_wbL2(), lookahead=barmerge.lookahead_off)

var float wkH1 = na
var float wkL1 = na
var float wkH2 = na
var float wkL2 = na

if isNewISOWeek
    wkH1 := na
    wkL1 := na
    wkH2 := na
    wkL2 := na

if not na(wb4H_H1)
    wkH1 := wb4H_H1
    wkL1 := wb4H_L1
if not na(wb4H_H2)
    wkH2 := wb4H_H2
    wkL2 := wb4H_L2

wbFinalH = not na(wkH1) and not na(wkH2) ? math.max(wkH1, wkH2) : not na(wkH1) ? wkH1 : wkH2
wbFinalL = not na(wkL1) and not na(wkL2) ? math.min(wkL1, wkL2) : not na(wkL1) ? wkL1 : wkL2

var line yHL      = na
var line yLL      = na
var float lastYH  = na
var float lastYL  = na

if showYellow and not na(wbFinalH) and not na(wbFinalL)
    if wbFinalH != lastYH or wbFinalL != lastYL
        lastYH := wbFinalH
        lastYL := wbFinalL
        line.delete(yHL)
        line.delete(yLL)
        yHL := line.new(bar_index, wbFinalH, bar_index+500, wbFinalH, color=color.new(color.yellow,0), width=2)
        yLL := line.new(bar_index, wbFinalL, bar_index+500, wbFinalL, color=color.new(color.yellow,0), width=2)
        label.new(bar_index, wbFinalH, biasLbl + " H", color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)
        label.new(bar_index, wbFinalL, biasLbl + " L", color=color.new(color.yellow,70), textcolor=color.yellow, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// GREEN LINES — 1AM 4H candle every day
// ═══════════════════════════════════════════
f_dbH() =>
    is1am = hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    is1am ? high : float(na)
f_dbL() =>
    is1am = hour(time, "GMT+0") == biasHour1 and minute(time, "GMT+0") == 0
    is1am ? low : float(na)

db4H_H = request.security(syminfo.tickerid, "240", f_dbH(), lookahead=barmerge.lookahead_off)
db4H_L = request.security(syminfo.tickerid, "240", f_dbL(), lookahead=barmerge.lookahead_off)

var float dayBiasH = na
var float dayBiasL = na
var line  gHL      = na
var line  gLL      = na
var float lastGH   = na
var float lastGL   = na

isNewGMTDay = ta.change(dayofmonth(time, "GMT+0")) != 0

if isNewGMTDay
    dayBiasH := na
    dayBiasL := na

if not na(db4H_H) and not na(db4H_L)
    dayBiasH := db4H_H
    dayBiasL := db4H_L

if showGreen and not na(dayBiasH) and not na(dayBiasL)
    if dayBiasH != lastGH or dayBiasL != lastGL
        lastGH := dayBiasH
        lastGL := dayBiasL
        line.delete(gHL)
        line.delete(gLL)
        gHL := line.new(bar_index, dayBiasH, bar_index+100, dayBiasH, color=color.new(color.green,0), width=2, style=line.style_dashed)
        gLL := line.new(bar_index, dayBiasL, bar_index+100, dayBiasL, color=color.new(color.green,0), width=2, style=line.style_dashed)
        label.new(bar_index, dayBiasH, dailyLbl + " H", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)
        label.new(bar_index, dayBiasL, dailyLbl + " L", color=color.new(color.green,70), textcolor=color.green, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// BREAK OF STRUCTURE
// ═══════════════════════════════════════════
swLen     = 5
swHigh    = ta.highest(high, swLen*2+1)[swLen]
swLow     = ta.lowest(low,   swLen*2+1)[swLen]
strongBod = math.abs(close - open) > atr * 0.7
bullBOS   = close > swHigh and strongBod and aboveE200 and close[1] <= swHigh
bearBOS   = close < swLow  and strongBod and belowE200 and close[1] >= swLow

var bool recBullBOS   = false
var bool recBearBOS   = false
var int  bosTimer     = 0
var bool bosBullFired = false
var bool bosBearFired = false
var float bosLevel    = na

if bullBOS
    recBullBOS   := true
    recBearBOS   := false
    bosTimer     := bar_index
    bosBullFired := true
    bosBearFired := false
    bosLevel     := swHigh
if bearBOS
    recBearBOS   := true
    recBullBOS   := false
    bosTimer     := bar_index
    bosBearFired := true
    bosBullFired := false
    bosLevel     := swLow
if bar_index - bosTimer > bosRst
    recBullBOS := false
    recBearBOS := false

if showBOS and bullBOS
    label.new(bar_index, low - atr*0.5, "🔨 BOS ↑", color=color.new(color.green,20), textcolor=color.white, style=label.style_label_up, size=size.small)
    line.new(bar_index-1, swHigh, bar_index+10, swHigh, color=color.new(color.green,40), width=1, style=line.style_dotted)
if showBOS and bearBOS
    label.new(bar_index, high + atr*0.5, "🔨 BOS ↓", color=color.new(color.red,20), textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index-1, swLow, bar_index+10, swLow, color=color.new(color.red,40), width=1, style=line.style_dotted)

// ═══════════════════════════════════════════
// LIQUIDITY SWEEPS
// ═══════════════════════════════════════════
bullSweep = low < d1L and close > d1L and (d1L - low) >= atr * 0.3
bearSweep = high > d1H and close < d1H and (high - d1H) >= atr * 0.3

var float lastBullSwL = na
var float lastBearSwH = na
if bullSweep
    lastBullSwL := low
if bearSweep
    lastBearSwH := high

bullSwRej = bullSweep and (close - low)  > atr * 1.0
bearSwRej = bearSweep and (high - close) > atr * 1.0

plotshape(showSweeps and bullSweep, "Bull Sweep", shape.triangleup,   location.belowbar, color.new(color.lime,0), size=size.small, text="Sweep")
plotshape(showSweeps and bearSweep, "Bear Sweep", shape.triangledown, location.abovebar, color.new(color.red,0),  size=size.small, text="Sweep")

intBullSw = low  < ta.lowest(low[1],   3)[1] and close > ta.lowest(low[1],   3)[1]
intBearSw = high > ta.highest(high[1], 3)[1] and close < ta.highest(high[1], 3)[1]
intBullOk = intBullSw and aboveE200
intBearOk = intBearSw and belowE200

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
imp3     = 3
bullImp  = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImp  = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBC  = bullImp and close[imp3] < open[imp3] and math.abs(close[imp3]-open[imp3]) > atr*0.5
bearOBC  = bearImp and close[imp3] > open[imp3] and math.abs(close[imp3]-open[imp3]) > atr*0.5

var float bOBT   = na
var float bOBB   = na
var float brOBT  = na
var float brOBB  = na
var bool  bOBAl  = false
var bool  brOBAl = false
var box   bOBBx  = na
var box   brOBBx = na
var bool  bOBRsp = false
var bool  brOBRsp= false
var int   bOBTch = 0
var int   brOBTch= 0

if bullOBC and aboveE200 and showOB
    bOBT   := high[imp3]
    bOBB   := low[imp3]
    bOBAl  := true
    bOBRsp := false
    bOBTch := 0
    box.delete(bOBBx)
    bOBBx := box.new(bar_index-imp3, bOBT, bar_index+100, bOBB, bgcolor=color.new(color.blue,80), border_color=color.new(color.blue,30), border_width=1)
    label.new(bar_index-imp3, bOBT, "🟦 Bull OB", color=color.new(color.blue,60), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bearOBC and belowE200 and showOB
    brOBT  := high[imp3]
    brOBB  := low[imp3]
    brOBAl := true
    brOBRsp:= false
    brOBTch:= 0
    box.delete(brOBBx)
    brOBBx := box.new(bar_index-imp3, brOBT, bar_index+100, brOBB, bgcolor=color.new(color.orange,80), border_color=color.new(color.orange,30), border_width=1)
    label.new(bar_index-imp3, brOBB, "🟧 Bear OB", color=color.new(color.orange,60), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if bOBAl and bOBTch == 0 and low <= bOBT and close >= bOBB
    bOBRsp := true
    bOBTch += 1
    box.set_bgcolor(bOBBx, color.new(color.green,70))
    box.set_border_color(bOBBx, color.new(color.green,20))
    label.new(bar_index, bOBB, "✅ Bull OB Entry", color=color.new(color.green,30), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if brOBAl and brOBTch == 0 and high >= brOBB and close <= brOBT
    brOBRsp := true
    brOBTch += 1
    box.set_bgcolor(brOBBx, color.new(color.red,70))
    box.set_border_color(brOBBx, color.new(color.red,20))
    label.new(bar_index, brOBT, "✅ Bear OB Entry", color=color.new(color.red,30), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bOBAl and close < bOBB
    bOBAl  := false
    bOBRsp := false
    box.set_bgcolor(bOBBx, color.new(color.gray,90))
if brOBAl and close > brOBT
    brOBAl  := false
    brOBRsp := false
    box.set_bgcolor(brOBBx, color.new(color.gray,90))

// ═══════════════════════════════════════════
// FVG
// ═══════════════════════════════════════════
cBody    = math.abs(close[1] - open[1])
strCdl   = cBody > atr * 1.0
dispCdl  = math.abs(close - open) > atr * 1.5
bFVGSz   = low - high[2]
brFVGSz  = low[2] - high
bFVGraw  = low > high[2] and strCdl and close[1] > open[1] and aboveE200 and bFVGSz  >= atr * fvgMin
brFVGraw = high < low[2] and strCdl and close[1] < open[1] and belowE200 and brFVGSz >= atr * fvgMin

var float bFVGT   = na
var float bFVGB   = na
var float brFVGT  = na
var float brFVGB  = na
var bool  bFVGAl  = false
var bool  brFVGAl = false
var box   bFVGBx  = na
var box   brFVGBx = na
var bool  bEZ     = false
var bool  brEZ    = false
var box   bEZBx   = na
var box   brEZBx  = na

if bFVGraw and showFVG
    bFVGT  := low
    bFVGB  := high[2]
    bFVGAl := true
    bEZ    := false
    box.delete(bFVGBx)
    bFVGBx := box.new(bar_index-2, bFVGT, bar_index+100, bFVGB, bgcolor=color.new(color.green,85), border_color=color.new(color.green,40), border_width=1)
    label.new(bar_index, bFVGB, "📦 Bull FVG", color=color.new(color.green,50), textcolor=color.white, style=label.style_label_up, size=size.tiny)

if brFVGraw and showFVG
    brFVGT  := low[2]
    brFVGB  := high
    brFVGAl := true
    brEZ    := false
    box.delete(brFVGBx)
    brFVGBx := box.new(bar_index-2, brFVGT, bar_index+100, brFVGB, bgcolor=color.new(color.red,85), border_color=color.new(color.red,40), border_width=1)
    label.new(bar_index, brFVGT, "📦 Bear FVG", color=color.new(color.red,50), textcolor=color.white, style=label.style_label_down, size=size.tiny)

if bFVGAl and low <= bFVGT and close >= bFVGB
    box.set_bgcolor(bFVGBx, color.new(color.green,70))
    if dispCdl and aboveE200 and weekBull and dayBull and not bEZ
        bEZ := true
        box.delete(bEZBx)
        bEZBx := box.new(bar_index, bFVGT, bar_index+30, bFVGB, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, bFVGB, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_up, size=size.small)

if brFVGAl and high >= brFVGB and close <= brFVGT
    box.set_bgcolor(brFVGBx, color.new(color.red,70))
    if dispCdl and belowE200 and weekBear and dayBear and not brEZ
        brEZ := true
        box.delete(brEZBx)
        brEZBx := box.new(bar_index, brFVGT, bar_index+30, brFVGB, bgcolor=color.new(color.yellow,60), border_color=color.new(color.yellow,0), border_width=2)
        label.new(bar_index, brFVGT, "🎯 FVG ENTRY ZONE", color=color.new(color.yellow,20), textcolor=color.black, style=label.style_label_down, size=size.small)

if bFVGAl and close < bFVGB
    bFVGAl := false
    bEZ    := false
    box.set_bgcolor(bFVGBx, color.new(color.gray,90))
if brFVGAl and close > brFVGT
    brFVGAl := false
    brEZ    := false
    box.set_bgcolor(brFVGBx, color.new(color.gray,90))

// ═══════════════════════════════════════════
// RSI
// ═══════════════════════════════════════════
rsi      = ta.rsi(close, 14)
lkb      = 7
pLL      = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
pHL      = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
pHH      = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
pLH      = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)
rHL      = ta.lowest(rsi,   lkb) > ta.lowest(rsi[lkb],   lkb)
rLL      = ta.lowest(rsi,   lkb) < ta.lowest(rsi[lkb],   lkb)
rLH      = ta.highest(rsi,  lkb) < ta.highest(rsi[lkb],  lkb)
rHH      = ta.highest(rsi,  lkb) > ta.highest(rsi[lkb],  lkb)
rsiBullX = rsi < 40
rsiBearX = rsi > 60
sBullDiv = pLL and rHL
hBullDiv = pHL and rLL
sBearDiv = pHH and rLH
hBearDiv = pLH and rHH
anyBullD = sBullDiv or hBullDiv
anyBearD = sBearDiv or hBearDiv

if showDiv and sBullDiv
    label.new(bar_index, low  - atr*0.3, "◆ Strong Bull Div", color=color.new(color.green,0),  textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and hBullDiv
    label.new(bar_index, low  - atr*0.3, "◇ Hidden Bull Div", color=color.new(color.aqua,0),   textcolor=color.white, style=label.style_label_up,   size=size.tiny)
if showDiv and sBearDiv
    label.new(bar_index, high + atr*0.3, "◆ Strong Bear Div", color=color.new(color.red,0),    textcolor=color.white, style=label.style_label_down, size=size.tiny)
if showDiv and hBearDiv
    label.new(bar_index, high + atr*0.3, "◇ Hidden Bear Div", color=color.new(color.orange,0), textcolor=color.white, style=label.style_label_down, size=size.tiny)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdL, sigL, histL] = ta.macd(close, 12, 26, 9)
macdBull = macdL > sigL
macdBear = macdL < sigL

// ═══════════════════════════════════════════
// TIER 2 — STRUCTURE
// ═══════════════════════════════════════════
s2Long  = 0
s2Long  := s2Long  + (aboveE200  ? 1 : 0)
s2Long  := s2Long  + (recBullBOS ? 1 : 0)
s2Long  := s2Long  + (bullSweep  ? 1 : 0)
s2Long  := s2Long  + (inDisc     ? 1 : 0)
s2Long  := s2Long  + (mBullAlign ? 1 : 0)

s2Short = 0
s2Short := s2Short + (belowE200  ? 1 : 0)
s2Short := s2Short + (recBearBOS ? 1 : 0)
s2Short := s2Short + (bearSweep  ? 1 : 0)
s2Short := s2Short + (inPrem     ? 1 : 0)
s2Short := s2Short + (mBearAlign ? 1 : 0)

// ═══════════════════════════════════════════
// TIER 3 — 7 ENTRY TRIGGERS
// ═══════════════════════════════════════════
t1Bull = bEZ
t1Bear = brEZ
t2Bull = bOBRsp
t2Bear = brOBRsp
t3Bull = bullSwRej
t3Bear = bearSwRej

bullPinWick = low - math.min(open, close)
bearPinWick = math.max(open, close) - high
pinBody     = math.abs(close - open)
atKeyBull   = math.abs(close - d1L) <= atr*0.5 or math.abs(close - ema200) <= atr*0.5 or (bOBAl and math.abs(close - bOBB) <= atr*0.5) or (bFVGAl and close >= bFVGB and close <= bFVGT)
atKeyBear   = math.abs(close - d1H) <= atr*0.5 or math.abs(close - ema200) <= atr*0.5 or (brOBAl and math.abs(close - brOBT) <= atr*0.5) or (brFVGAl and close >= brFVGB and close <= brFVGT)
t4Bull = bullPinWick >= 2.5*pinBody and pinBody >= atr*0.1 and aboveE200 and atKeyBull and weekBull and dayBull
t4Bear = bearPinWick >= 2.5*pinBody and pinBody >= atr*0.1 and belowE200 and atKeyBear and weekBear and dayBear

t5Bull = close > d1H and close[1] <= d1H and aboveE200 and weekBull and dayBull
t5Bear = close < d1L and close[1] >= d1L and belowE200 and weekBear and dayBear

var bool bosRetestReady = false
var bool bosRetestDir   = false
var float bosRetestLvl  = na
if bosBullFired
    bosRetestReady := true
    bosRetestDir   := true
    bosRetestLvl   := bosLevel
if bosBearFired
    bosRetestReady := true
    bosRetestDir   := false
    bosRetestLvl   := bosLevel
bosRetestTouch = bosRetestReady and not na(bosRetestLvl) and math.abs(close - bosRetestLvl) <= atr*0.3
t6Bull = bosRetestTouch and bosRetestDir     and close > open and aboveE200
t6Bear = bosRetestTouch and not bosRetestDir and close < open and belowE200
if t6Bull or t6Bear
    bosRetestReady := false

lh1    = high[1] < high[2] and high[2] < high[3] and high[3] < high[4]
hl1    = low[1]  > low[2]  and low[2]  > low[3]  and low[3]  > low[4]
t7Bull = lh1 and high > high[1] and aboveE200 and weekBull and dayBull
t7Bear = hl1 and low  < low[1]  and belowE200 and weekBear and dayBear

anyTrigBull = t1Bull or t2Bull or t3Bull or t4Bull or t5Bull or t6Bull or t7Bull
anyTrigBear = t1Bear or t2Bear or t3Bear or t4Bear or t5Bear or t6Bear or t7Bear

tStrBull = t1Bull ? "🎯 FVG" : t2Bull ? "🟦 OB" : t3Bull ? "💧 Sweep" : t4Bull ? "📍 Pin" : t5Bull ? "📈 PDH" : t6Bull ? "🔨 BOS Retest" : t7Bull ? "🔄 CHoCH" : "⏳ None"
tStrBear = t1Bear ? "🎯 FVG" : t2Bear ? "🟧 OB" : t3Bear ? "💧 Sweep" : t4Bear ? "📍 Pin" : t5Bear ? "📉 PDL" : t6Bear ? "🔨 BOS Retest" : t7Bear ? "🔄 CHoCH" : "⏳ None"

// ═══════════════════════════════════════════
// TIER 4 — BONUS
// ═══════════════════════════════════════════
bonL = 0
bonL := bonL + (sBullDiv  ? 2 : hBullDiv ? 1 : 0)
bonL := bonL + (macdBull  ? 1 : 0)
bonL := bonL + (rsiBullX  ? 1 : 0)
bonL := bonL + (trending  ? 1 : 0)
bonL := bonL + (intBullOk ? 1 : 0)
bonL := bonL + (eqlSwept  ? 1 : 0)

bonS = 0
bonS := bonS + (sBearDiv  ? 2 : hBearDiv ? 1 : 0)
bonS := bonS + (macdBear  ? 1 : 0)
bonS := bonS + (rsiBearX  ? 1 : 0)
bonS := bonS + (trending  ? 1 : 0)
bonS := bonS + (intBearOk ? 1 : 0)
bonS := bonS + (eqhSwept  ? 1 : 0)

// ═══════════════════════════════════════════
// SIGNAL FIRING
// ═══════════════════════════════════════════
g1L  = weekBull
g2L  = dayBull
g3L  = htfBullOk
g4L  = activeSess
g5L  = not extBear
allL = g1L and g2L and g3L and g4L and g5L

g1S  = weekBear
g2S  = dayBear
g3S  = htfBearOk
g4S  = activeSess
g5S  = not extBear
allS = g1S and g2S and g3S and g4S and g5S

var int lastBullB = 0
var int lastBearB = 0

rawBull = allL and s2Long  >= minStr2 and anyTrigBull and bonL >= minBonus
rawBear = allS and s2Short >= minStr2 and anyTrigBear and bonS >= minBonus

bullSig = rawBull and (bar_index - lastBullB) > cooldown
bearSig = rawBear and (bar_index - lastBearB) > cooldown

if bullSig
    lastBullB := bar_index
if bearSig
    lastBearB := bar_index

// ═══════════════════════════════════════════
// SIGNAL QUALITY
// ═══════════════════════════════════════════
qLong  = bonL >= 7 ? "🔥 A+" : bonL >= 5 ? "💪 STRONG" : "✅ VALID"
qShort = bonS >= 7 ? "🔥 A+" : bonS >= 5 ? "💪 STRONG" : "✅ VALID"

// ═══════════════════════════════════════════
// RISK MANAGEMENT
// ═══════════════════════════════════════════
riskAmt  = accountSize * (riskPct / 100)
slBufL   = bonL >= 6 ? atr * 0.2 : atr * 0.5
slBufS   = bonS >= 6 ? atr * 0.2 : atr * 0.5
longSL   = not na(lastBullSwL) ? lastBullSwL - slBufL : close - atr * 1.5
longRisk = math.max(close - longSL, 0.0001)
longTP   = close + longRisk * 3
longSz   = riskAmt / longRisk
shortSL  = not na(lastBearSwH) ? lastBearSwH + slBufS : close + atr * 1.5
shortRsk = math.max(shortSL - close, 0.0001)
shortTP  = close - shortRsk * 3
shortSz  = riskAmt / shortRsk

// ═══════════════════════════════════════════
// SIGNAL LABELS
// ═══════════════════════════════════════════
lLbl =
     "⬆ BUY — " + tfLbl + " — " + qLong + "\n" +
     "Wk🟢 D🟢 HTF🟢 Sess✅\n" +
     "Str:" + str.tostring(s2Long) + "/5 Trig:" + tStrBull + " Bon:" + str.tostring(bonL) + "/7\n" +
     "SL:" + str.tostring(math.round(longSL,2)) + " TP:" + str.tostring(math.round(longTP,2)) + " RR:1:3\n" +
     "Risk:$" + str.tostring(math.round(riskAmt,2)) + " Target:$" + str.tostring(math.round(riskAmt*3,2))

sLbl =
     "⬇ SELL — " + tfLbl + " — " + qShort + "\n" +
     "Wk🔴 D🔴 HTF🔴 Sess✅\n" +
     "Str:" + str.tostring(s2Short) + "/5 Trig:" + tStrBear + " Bon:" + str.tostring(bonS) + "/7\n" +
     "SL:" + str.tostring(math.round(shortSL,2)) + " TP:" + str.tostring(math.round(shortTP,2)) + " RR:1:3\n" +
     "Risk:$" + str.tostring(math.round(riskAmt,2)) + " Target:$" + str.tostring(math.round(riskAmt*3,2))

if showSignals and bullSig
    label.new(bar_index, low,  lLbl, color=color.new(color.green,10), textcolor=color.white, style=label.style_label_up,   size=size.small)
    line.new(bar_index, longSL, bar_index+40, longSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, longTP, bar_index+40, longTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, longSL, "🔴 SL " + str.tostring(math.round(longSL,2)), color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, longTP, "🟢 TP " + str.tostring(math.round(longTP,2)), color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

if showSignals and bearSig
    label.new(bar_index, high, sLbl, color=color.new(color.red,10),   textcolor=color.white, style=label.style_label_down, size=size.small)
    line.new(bar_index, shortSL, bar_index+40, shortSL, color=color.red,   width=2, style=line.style_dashed)
    line.new(bar_index, shortTP, bar_index+40, shortTP, color=color.green, width=2, style=line.style_dashed)
    label.new(bar_index+40, shortSL, "🔴 SL " + str.tostring(math.round(shortSL,2)), color=color.new(color.red,0),   textcolor=color.white, style=label.style_label_left, size=size.tiny)
    label.new(bar_index+40, shortTP, "🟢 TP " + str.tostring(math.round(shortTP,2)), color=color.new(color.green,0), textcolor=color.white, style=label.style_label_left, size=size.tiny)

// ═══════════════════════════════════════════
// ALERTS
// ═══════════════════════════════════════════
alertcondition(bullSig, title="🟢 GOLD BUY — High Confidence",  message="⬆ BUY SIGNAL XAUUSD — Check chart")
alertcondition(bearSig, title="🔴 GOLD SELL — High Confidence", message="⬇ SELL SIGNAL XAUUSD — Check chart")
alertcondition(bullSig or bearSig, title="🚦 GOLD Signal",      message="Gold signal fired on XAUUSD — Check chart")

// ═══════════════════════════════════════════
// WAIT REASON
// ═══════════════════════════════════════════
wr1L = not g1L ? "⏳ Weekly Bearish"                                              : ""
wr2L = not g2L ? "⏳ Daily Bearish"                                               : ""
wr3L = not g3L ? "⏳ HTF " + htfNeeded + " Not Bullish"                           : ""
wr4L = not g4L ? "⏳ Asian Session"                                               : ""
wr5L = not g5L ? "⏳ Extreme Bear"                                                : ""
wr6L = s2Long < minStr2 ? "⏳ Need " + str.tostring(minStr2 - s2Long) + " Str"   : ""
wr7L = not anyTrigBull  ? "⏳ Need Trigger"                                       : ""
wr8L = bonL < minBonus  ? "⏳ Need " + str.tostring(minBonus - bonL) + " Bonus"  : ""
waitL = not g1L ? wr1L : not g2L ? wr2L : not g3L ? wr3L : not g4L ? wr4L : not g5L ? wr5L : s2Long < minStr2 ? wr6L : not anyTrigBull ? wr7L : bonL < minBonus ? wr8L : "✅ LONG READY"

wr1S = not g1S ? "⏳ Weekly Bullish"                                              : ""
wr2S = not g2S ? "⏳ Daily Bullish"                                               : ""
wr3S = not g3S ? "⏳ HTF " + htfNeeded + " Not Bearish"                           : ""
wr4S = not g4S ? "⏳ Asian Session"                                               : ""
wr5S = not g5S ? "⏳ Extreme Bear"                                                : ""
wr6S = s2Short < minStr2 ? "⏳ Need " + str.tostring(minStr2 - s2Short) + " Str" : ""
wr7S = not anyTrigBear   ? "⏳ Need Trigger"                                      : ""
wr8S = bonS < minBonus   ? "⏳ Need " + str.tostring(minBonus - bonS) + " Bonus" : ""
waitS = not g1S ? wr1S : not g2S ? wr2S : not g3S ? wr3S : not g4S ? wr4S : not g5S ? wr5S : s2Short < minStr2 ? wr6S : not anyTrigBear ? wr7S : bonS < minBonus ? wr8S : "✅ SHORT READY"

// ═══════════════════════════════════════════
// SIGNAL ROW VALUES — used in both dashboards
// ═══════════════════════════════════════════
waitStr  = belowE200 ? waitS : waitL
sigStr   = bullSig ? "⬆ BUY " + tfLbl + " — " + qLong : bearSig ? "⬇ SELL " + tfLbl + " — " + qShort : waitStr
sigCol   = bullSig ? color.green : bearSig ? color.red : color.yellow
sig3     = bullSig ? "🟢 BUY" : bearSig ? "🔴 SELL" : "⏳ WAIT"
sig3C    = bullSig ? color.green : bearSig ? color.red : color.yellow
bonDisp  = belowE200 ? bonS : bonL
s2disp   = belowE200 ? s2Short : s2Long
tActive  = anyTrigBull or anyTrigBear
tActStr  = tActive ? (belowE200 ? tStrBear : tStrBull) : "⏳ None"
tActCol  = tActive ? color.green : color.yellow

// ═══════════════════════════════════════════
// COMPACT DASHBOARD — Phone Mode
// Only 8 rows — signal, bias, htf,
// session, structure, trigger, bonus, risk
// ═══════════════════════════════════════════
var table dashC = table.new(dashPosVal, 3, 9, bgcolor=color.new(color.black,30), border_color=color.new(color.gray,40), border_width=1)

if compactDash
    hBGc = color.new(color.blue,40)
    table.cell(dashC, 0, 0, "🥇 " + tfLbl, text_color=color.yellow, text_size=size.small, bgcolor=hBGc)
    table.cell(dashC, 1, 0, modeStr,        text_color=color.aqua,   text_size=size.small, bgcolor=hBGc)
    table.cell(dashC, 2, 0, sessName,        text_color=sessCol,      text_size=size.small, bgcolor=hBGc)

    table.cell(dashC, 0, 1, "Weekly Bias",  text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 1, wkBiasStr,      text_color=weekBull ? color.green : weekBear ? color.red : color.yellow, text_size=size.small)
    table.cell(dashC, 2, 1, weekBull or weekBear ? "✅" : "❌", text_color=weekBull or weekBear ? color.green : color.red, text_size=size.small)

    table.cell(dashC, 0, 2, "Daily Bias",   text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 2, dBiasStr,       text_color=dayBull ? color.green : dayBear ? color.red : color.yellow, text_size=size.small)
    table.cell(dashC, 2, 2, dayBull or dayBear ? "✅" : "❌", text_color=dayBull or dayBear ? color.green : color.red, text_size=size.small)

    table.cell(dashC, 0, 3, "HTF " + htfNeeded, text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 3, htfStr,         text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.tiny)
    table.cell(dashC, 2, 3, htfBullOk or htfBearOk ? "✅" : "❌", text_color=htfBullOk or htfBearOk ? color.green : color.red, text_size=size.small)

    table.cell(dashC, 0, 4, "Structure",    text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 4, str.tostring(s2disp) + "/5 (need " + str.tostring(minStr2) + ")", text_color=s2disp >= minStr2 ? color.green : color.yellow, text_size=size.small)
    table.cell(dashC, 2, 4, s2disp >= minStr2 ? "✅" : "❌", text_color=s2disp >= minStr2 ? color.green : color.red, text_size=size.small)

    table.cell(dashC, 0, 5, "Trigger",      text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 5, tActStr,        text_color=tActCol,     text_size=size.small)
    table.cell(dashC, 2, 5, tActive ? "✅" : "❌", text_color=tActive ? color.green : color.red, text_size=size.small)

    table.cell(dashC, 0, 6, "Bonus",        text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 6, str.tostring(bonDisp) + "/7 (need " + str.tostring(minBonus) + ")", text_color=bonDisp >= minBonus ? color.green : color.yellow, text_size=size.small)
    table.cell(dashC, 2, 6, bonDisp >= minBonus ? "✅" : "❌", text_color=bonDisp >= minBonus ? color.green : color.red, text_size=size.small)

    table.cell(dashC, 0, 7, "Risk/TP",      text_color=color.white, text_size=size.small)
    table.cell(dashC, 1, 7, "$" + str.tostring(math.round(riskAmt,2)) + " → $" + str.tostring(math.round(riskAmt*3,2)), text_color=color.yellow, text_size=size.small)
    table.cell(dashC, 2, 7, "1:3",          text_color=color.yellow, text_size=size.small)

    table.cell(dashC, 0, 8, "🚦 SIGNAL",   text_color=color.white, text_size=size.normal, bgcolor=color.new(color.gray,30))
    table.cell(dashC, 1, 8, sigStr,         text_color=sigCol,      text_size=size.small)
    table.cell(dashC, 2, 8, sig3,           text_color=sig3C,       text_size=size.normal)

// ═══════════════════════════════════════════
// FULL DASHBOARD — Desktop Mode
// ═══════════════════════════════════════════
var table dash = table.new(dashPosVal, 3, 42, bgcolor=color.new(color.black,50), border_color=color.new(color.gray,30), border_width=1)
hBG  = color.new(color.blue,40)
sBG  = color.new(color.gray,60)
sCol = color.yellow

if not compactDash
    table.cell(dash, 0, 0, "📊 CONDITION",  text_color=color.white,  text_size=size.small, bgcolor=hBG)
    table.cell(dash, 1, 0, "STATUS",        text_color=color.white,  text_size=size.small, bgcolor=hBG)
    table.cell(dash, 2, 0, "🥇 " + tfLbl,  text_color=color.yellow, text_size=size.small, bgcolor=hBG)

    table.cell(dash, 0, 1, "── MARKET ──",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 1, "────────────",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 1, "──────",        text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 2, "Regime",        text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 2, regStr,          text_color=regCol,      text_size=size.small)
    table.cell(dash, 2, 2, extBear ? "🚫" : "✅", text_color=extBear ? color.red : color.green, text_size=size.small)
    table.cell(dash, 0, 3, "ADX",           text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 3, adxStr,          text_color=adxCol,      text_size=size.small)
    table.cell(dash, 2, 3, trending ? "✅" : "❌", text_color=trending ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 4, "Session",       text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 4, sessName,        text_color=sessCol,     text_size=size.small)
    table.cell(dash, 2, 4, activeSess ? "✅" : "❌", text_color=activeSess ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 5, "TF Mode",       text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 5, modeStr,         text_color=color.aqua,  text_size=size.small)
    table.cell(dash, 2, 5, "B≥" + str.tostring(minBonus) + " S≥" + str.tostring(minStr2), text_color=color.aqua, text_size=size.tiny)
    table.cell(dash, 0, 6, "HTF " + htfNeeded, text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 6, htfStr, text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.tiny)
    table.cell(dash, 2, 6, htfBullOk ? "✅🟢" : htfBearOk ? "✅🔴" : "❌", text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.small)

    table.cell(dash, 0, 7, "── TIER 1 GATES ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 7, "────────────────",   text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 7, "──────",             text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 8, "Weekly Bias ⚡",    text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 8, wkBiasStr, text_color=weekBull ? color.green : weekBear ? color.red : color.yellow, text_size=size.small)
    table.cell(dash, 2, 8, weekBull or weekBear ? "✅" : "❌", text_color=weekBull or weekBear ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 9, "Daily Bias 🟢",     text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 9, dBiasStr, text_color=dayBull ? color.green : dayBear ? color.red : color.yellow, text_size=size.small)
    table.cell(dash, 2, 9, dayBull or dayBear ? "✅" : "❌", text_color=dayBull or dayBear ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 10, "HTF Aligned",      text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 10, htfStr, text_color=htfBullOk ? color.green : htfBearOk ? color.red : color.yellow, text_size=size.tiny)
    table.cell(dash, 2, 10, htfBullOk or htfBearOk ? "✅" : "❌", text_color=htfBullOk or htfBearOk ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 11, "Session Gate",     text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 11, sessName,           text_color=sessCol,     text_size=size.small)
    table.cell(dash, 2, 11, activeSess ? "✅" : "❌", text_color=activeSess ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 12, "Extreme Bear",     text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 12, extBear ? "🚨 BLOCKED" : "✅ SAFE", text_color=extBear ? color.red : color.green, text_size=size.small)
    table.cell(dash, 2, 12, extBear ? "❌" : "✅", text_color=extBear ? color.red : color.green, text_size=size.small)

    table.cell(dash, 0, 13, "── TIER 2 STRUCTURE ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 13, "────────────────────",   text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 13, "──────",                 text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 14, "EMA 200",       text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 14, aboveE200 ? "🟢 BULLISH" : "🔴 BEARISH", text_color=aboveE200 ? color.green : color.red, text_size=size.small)
    table.cell(dash, 2, 14, "✅", text_color=color.green, text_size=size.small)
    table.cell(dash, 0, 15, "BOS",           text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 15, recBullBOS ? "🟢 BULL" : recBearBOS ? "🔴 BEAR" : "⏳ NONE", text_color=recBullBOS ? color.green : recBearBOS ? color.red : color.gray, text_size=size.small)
    table.cell(dash, 2, 15, recBullBOS or recBearBOS ? "✅" : "❌", text_color=recBullBOS or recBearBOS ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 16, "Sweep",         text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 16, bullSweep ? "🟢 BULL" : bearSweep ? "🔴 BEAR" : "⏳ NONE", text_color=bullSweep or bearSweep ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 16, bullSweep or bearSweep ? "✅" : "❌", text_color=bullSweep or bearSweep ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 17, "Zone",          text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 17, inDisc ? "🟢 DISCOUNT" : inPrem ? "🔴 PREMIUM" : "⚪ EQUILIB", text_color=inDisc ? color.green : inPrem ? color.red : color.gray, text_size=size.small)
    table.cell(dash, 2, 17, inDisc or inPrem ? "✅" : "➖", text_color=inDisc or inPrem ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 18, "Monthly",       text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 18, mBullAlign ? "🟢 BULL" : mBearAlign ? "🔴 BEAR" : "⏳ WAIT", text_color=mBullAlign ? color.green : mBearAlign ? color.red : color.yellow, text_size=size.small)
    table.cell(dash, 2, 18, mBullAlign or mBearAlign ? "✅" : "❌", text_color=mBullAlign or mBearAlign ? color.green : color.red, text_size=size.small)
    table.cell(dash, 0, 19, "Str Score",     text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 19, str.tostring(s2disp) + "/5 (need " + str.tostring(minStr2) + ")", text_color=s2disp >= minStr2 ? color.green : color.yellow, text_size=size.small)
    table.cell(dash, 2, 19, s2disp >= minStr2 ? "✅" : "❌", text_color=s2disp >= minStr2 ? color.green : color.red, text_size=size.small)

    table.cell(dash, 0, 20, "── TIER 3 TRIGGERS ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 20, "────────────────────",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 20, "──────",                text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 21, "T1 FVG Zone",   text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 21, t1Bull or t1Bear ? "🎯 ACTIVE" : "⏳ NONE", text_color=t1Bull or t1Bear ? color.yellow : color.gray, text_size=size.small)
    table.cell(dash, 2, 21, t1Bull or t1Bear ? "✅" : "➖", text_color=t1Bull or t1Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 22, "T2 OB Respect", text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 22, t2Bull ? "✅ BULL" : t2Bear ? "✅ BEAR" : bOBAl or brOBAl ? "⏳ LIVE" : "⏳ NONE", text_color=t2Bull or t2Bear ? color.green : bOBAl or brOBAl ? color.yellow : color.gray, text_size=size.small)
    table.cell(dash, 2, 22, t2Bull or t2Bear ? "✅" : "➖", text_color=t2Bull or t2Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 23, "T3 Sw Reject",  text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 23, t3Bull ? "🟢 BULL" : t3Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t3Bull or t3Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 23, t3Bull or t3Bear ? "✅" : "➖", text_color=t3Bull or t3Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 24, "T4 Pin Bar",    text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 24, t4Bull ? "🟢 BULL" : t4Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t4Bull or t4Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 24, t4Bull or t4Bear ? "✅" : "➖", text_color=t4Bull or t4Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 25, "T5 PDH/L",      text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 25, t5Bull ? "🟢 BULL" : t5Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t5Bull or t5Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 25, t5Bull or t5Bear ? "✅" : "➖", text_color=t5Bull or t5Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 26, "T6 BOS Retest", text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 26, t6Bull ? "🟢 BULL" : t6Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t6Bull or t6Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 26, t6Bull or t6Bear ? "✅" : "➖", text_color=t6Bull or t6Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 27, "T7 CHoCH",      text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 27, t7Bull ? "🟢 BULL" : t7Bear ? "🔴 BEAR" : "⏳ NONE", text_color=t7Bull or t7Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 27, t7Bull or t7Bear ? "✅" : "➖", text_color=t7Bull or t7Bear ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 28, "Active Trigger", text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 28, tActStr,          text_color=tActCol,     text_size=size.small)
    table.cell(dash, 2, 28, tActive ? "✅" : "❌", text_color=tActive ? color.green : color.red, text_size=size.small)

    table.cell(dash, 0, 29, "── TIER 4 BONUS ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 29, "──────────────────", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 29, "──────",             text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 30, "RSI Div",       text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 30, sBullDiv ? "🟢 Strong(+2)" : hBullDiv ? "🟢 Hidden(+1)" : sBearDiv ? "🔴 Strong(+2)" : hBearDiv ? "🔴 Hidden(+1)" : "⏳ NONE", text_color=anyBullD ? color.green : anyBearD ? color.red : color.gray, text_size=size.small)
    table.cell(dash, 2, 30, anyBullD or anyBearD ? "✅" : "➖", text_color=anyBullD or anyBearD ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 31, "MACD",          text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 31, macdBull ? "🟢 BULL" : "🔴 BEAR", text_color=macdBull ? color.green : color.red, text_size=size.small)
    table.cell(dash, 2, 31, "✅", text_color=color.green, text_size=size.small)
    table.cell(dash, 0, 32, "RSI Level",     text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 32, str.tostring(math.round(rsi,1)) + (rsiBullX ? " 🟢 OS" : rsiBearX ? " 🔴 OB" : " ⚪ N"), text_color=rsiBullX ? color.green : rsiBearX ? color.red : color.white, text_size=size.small)
    table.cell(dash, 2, 32, rsiBullX or rsiBearX ? "✅" : "➖", text_color=rsiBullX or rsiBearX ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 33, "Int Sweep",     text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 33, intBullOk ? "🟢 BULL" : intBearOk ? "🔴 BEAR" : "⏳ NONE", text_color=intBullOk or intBearOk ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 33, intBullOk or intBearOk ? "✅" : "➖", text_color=intBullOk or intBearOk ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 34, "EQH/EQL",       text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 34, eqhSwept ? "🔴 EQH" : eqlSwept ? "🟢 EQL" : "⏳ NONE", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 34, eqhSwept or eqlSwept ? "✅" : "➖", text_color=eqhSwept or eqlSwept ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 35, "Bonus Score",   text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 35, str.tostring(bonDisp) + "/7 (need " + str.tostring(minBonus) + ")", text_color=bonDisp >= minBonus ? color.green : color.yellow, text_size=size.small)
    table.cell(dash, 2, 35, bonDisp >= minBonus ? "✅" : "❌", text_color=bonDisp >= minBonus ? color.green : color.red, text_size=size.small)

    table.cell(dash, 0, 36, "── RISK ──",    text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 36, "────────────",  text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 36, "──────",        text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 37, "Risk/Target",   text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 37, "💰 $" + str.tostring(math.round(riskAmt,2)) + " → 🎯 $" + str.tostring(math.round(riskAmt*3,2)) + " (1:3)", text_color=color.yellow, text_size=size.small)
    table.cell(dash, 2, 37, "1:3",           text_color=color.yellow, text_size=size.small)

    table.cell(dash, 0, 38, "── BIAS LINES ──", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 1, 38, "────────────────", text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 2, 38, "──────",           text_color=sCol, text_size=size.tiny, bgcolor=sBG)
    table.cell(dash, 0, 39, "⚡ Yellow",        text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 39, not na(wbFinalH) ? "H:" + str.tostring(math.round(wbFinalH,2)) + " L:" + str.tostring(math.round(wbFinalL,2)) : "⏳ Waiting for Monday", text_color=not na(wbFinalH) ? color.yellow : color.gray, text_size=size.small)
    table.cell(dash, 2, 39, not na(wbFinalH) ? "✅" : "⏳", text_color=not na(wbFinalH) ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 0, 40, "🟢 Green",         text_color=color.white, text_size=size.small)
    table.cell(dash, 1, 40, not na(dayBiasH) ? "H:" + str.tostring(math.round(dayBiasH,2)) + " L:" + str.tostring(math.round(dayBiasL,2)) : "⏳ Waiting for 1AM", text_color=not na(dayBiasH) ? color.green : color.gray, text_size=size.small)
    table.cell(dash, 2, 40, not na(dayBiasH) ? "✅" : "⏳", text_color=not na(dayBiasH) ? color.green : color.gray, text_size=size.small)

    table.cell(dash, 0, 41, "🚦 SIGNAL",     text_color=color.white, text_size=size.normal, bgcolor=color.new(color.gray,30))
    table.cell(dash, 1, 41, sigStr,          text_color=sigCol,      text_size=size.small)
    table.cell(dash, 2, 41, sig3,            text_color=sig3C,       text_size=size.normal)
```

```python
// ============================================================
// Manny's Gold Strategy V2 — BACKTEST
// Pine Script v6 — XAUUSD Only
// Add to chart then open Strategy Tester tab
// ============================================================
//@version=6
strategy("Manny's Gold Strategy V2 — Backtest", overlay=true,
     initial_capital=1000,
     default_qty_type=strategy.fixed,
     default_qty_value=1,
     commission_type=strategy.commission.percent,
     commission_value=0.01,
     slippage=2,
     max_lines_count=500,
     max_labels_count=500,
     max_boxes_count=500)

// ═══════════════════════════════════════════
// SETTINGS
// ═══════════════════════════════════════════
goldBiasTime = input.string("1AM / 5AM (Standard Time)", "🥇 Bias Candle Time",
     options=["1AM / 5AM (Standard Time)", "2AM / 6AM (Daylight Saving)"],
     group="🧭 Bias")
modeOvr      = input.string("Auto", "Trading Mode Override",
     options=["Auto","Scalping","Day Trade","Swing Trade"],
     group="🎯 Signal Quality")
minBonusOvr  = input.int(0,      "Min Bonus Override (0=auto)",   group="🎯 Signal Quality", minval=0, maxval=7)
adxLen       = input.int(14,     "ADX Period",                    group="📊 Filters")
adxThr       = input.int(25,     "ADX Minimum",                   group="📊 Filters")
accountSize  = input.float(1000, "💰 Account Size ($)",           group="💵 Risk")
riskPct      = input.float(1.0,  "⚠️ Risk Per Trade (%)",        group="💵 Risk")
atrLen       = input.int(14,     "ATR Period",                    group="💵 Risk")
fvgMinIn     = input.float(0.5,  "FVG Min Size (x ATR)",          group="📊 Filters", minval=0.1)

// ═══════════════════════════════════════════
// BIAS TIME
// ═══════════════════════════════════════════
isStandard = goldBiasTime == "1AM / 5AM (Standard Time)"
biasHour1  = isStandard ? 1 : 2
biasHour2  = isStandard ? 5 : 6

// ═══════════════════════════════════════════
// TIMEFRAME DETECTION
// ═══════════════════════════════════════════
tfMins    = timeframe.in_seconds() / 60
isScalpTF = tfMins <= 5
isDayTF   = tfMins > 5   and tfMins <= 30
isSwingTF = tfMins > 30  and tfMins <= 240
autoStr   = isScalpTF ? "Scalping" : isDayTF ? "Day Trade" : isSwingTF ? "Swing Trade" : "Long Term"
isScalp   = modeOvr == "Auto" ? isScalpTF : modeOvr == "Scalping"
isSwing   = modeOvr == "Auto" ? isSwingTF : modeOvr == "Swing Trade"
tfLbl     = tfMins == 1 ? "1M" : tfMins == 5 ? "5M" : tfMins == 15 ? "15M" : tfMins == 30 ? "30M" : tfMins == 60 ? "1H" : tfMins == 240 ? "4H" : tfMins == 1440 ? "D" : str.tostring(math.round(tfMins)) + "M"
minBonus  = minBonusOvr > 0 ? minBonusOvr : isScalp ? 2 : isSwing ? 4 : 3
bosRst    = isScalp ? 50 : isSwing ? 20 : 30
fvgMin    = isScalp ? 0.3 : isSwing ? 0.7 : fvgMinIn
minStr2   = isScalp ? 2 : isSwing ? 3 : 2
cooldown  = tfMins <= 1 ? 15 : tfMins <= 5 ? 12 : tfMins <= 15 ? 10 : tfMins <= 30 ? 8 : 5

// ═══════════════════════════════════════════
// TIME
// ═══════════════════════════════════════════
gmtH         = hour(time, "GMT+0")
isNewDay     = ta.change(time("D")) != 0
isNewWeek    = ta.change(time("W")) != 0
isNewMonth   = ta.change(time("M")) != 0
isNewISOWeek = ta.change(weekofyear(time, "GMT+0")) != 0
isNewGMTDay  = ta.change(dayofmonth(time, "GMT+0")) != 0

// ═══════════════════════════════════════════
// SESSION
// ═══════════════════════════════════════════
london     = gmtH >= 7  and gmtH < 12
ny         = gmtH >= 13 and gmtH < 17
overlap    = gmtH >= 12 and gmtH < 13
activeSess = london or ny or overlap

// ═══════════════════════════════════════════
// ATR + EMA
// ═══════════════════════════════════════════
atr       = ta.atr(atrLen)
ema200    = ta.ema(close, 200)
ema50     = ta.ema(close, 50)
aboveE200 = close > ema200
belowE200 = close < ema200
bullMkt   = ema50 > ema200
pctBelow  = (ema200 - close) / ema200 * 100
extBear   = pctBelow > 20

// ═══════════════════════════════════════════
// ADX
// ═══════════════════════════════════════════
[diP, diM, adxVal] = ta.dmi(adxLen, adxLen)
trending = adxVal >= adxThr

// ═══════════════════════════════════════════
// HTF EMA 200
// ═══════════════════════════════════════════
c1H    = request.security(syminfo.tickerid, "60",  close,              lookahead=barmerge.lookahead_off)
e1H    = request.security(syminfo.tickerid, "60",  ta.ema(close, 200), lookahead=barmerge.lookahead_off)
c4H    = request.security(syminfo.tickerid, "240", close,              lookahead=barmerge.lookahead_off)
e4H    = request.security(syminfo.tickerid, "240", ta.ema(close, 200), lookahead=barmerge.lookahead_off)
cD     = request.security(syminfo.tickerid, "D",   close,              lookahead=barmerge.lookahead_off)
eD     = request.security(syminfo.tickerid, "D",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)
cW     = request.security(syminfo.tickerid, "W",   close,              lookahead=barmerge.lookahead_off)
eW     = request.security(syminfo.tickerid, "W",   ta.ema(close, 200), lookahead=barmerge.lookahead_off)

bull1H = c1H > e1H
bull4H = c4H > e4H
bullD  = cD  > eD
bullW  = cW  > eW
bear1H = c1H < e1H
bear4H = c4H < e4H
bearD  = cD  < eD
bearW  = cW  < eW

// ═══════════════════════════════════════════
// GATES 1 AND 2 — BIAS
// ═══════════════════════════════════════════
weekBull = bull4H
weekBear = bear4H
dayBull  = bull1H
dayBear  = bear1H

// ═══════════════════════════════════════════
// GATE 3 — HTF
// ═══════════════════════════════════════════
htfBullOk = isScalpTF ? (bull1H and bull4H and bullD) : isDayTF ? (bull1H and bull4H) : isSwingTF ? bull4H : (bullD and bullW)
htfBearOk = isScalpTF ? (bear1H and bear4H and bearD) : isDayTF ? (bear1H and bear4H) : isSwingTF ? bear4H : (bearD and bearW)

// ═══════════════════════════════════════════
// LEVELS
// ═══════════════════════════════════════════
d1H = ta.valuewhen(isNewDay,   high[1], 0)
d1L = ta.valuewhen(isNewDay,   low[1],  0)
mH  = ta.valuewhen(isNewMonth, high[1], 0)
mL  = ta.valuewhen(isNewMonth, low[1],  0)
wH  = ta.valuewhen(isNewWeek,  high[1], 0)
wL  = ta.valuewhen(isNewWeek,  low[1],  0)

wRange = wH - wL
wMid   = wL + wRange / 2
inDisc = close < wMid and not na(wMid)
inPrem = close > wMid and not na(wMid)

mBullAlign = close > mL and aboveE200
mBearAlign = close < mH and belowE200

// ═══════════════════════════════════════════
// BOS
// ═══════════════════════════════════════════
swLen     = 5
swHigh    = ta.highest(high, swLen*2+1)[swLen]
swLow     = ta.lowest(low,   swLen*2+1)[swLen]
strongBod = math.abs(close - open) > atr * 0.7
bullBOS   = close > swHigh and strongBod and aboveE200 and close[1] <= swHigh
bearBOS   = close < swLow  and strongBod and belowE200 and close[1] >= swLow

var bool recBullBOS   = false
var bool recBearBOS   = false
var int  bosTimer     = 0
var bool bosBullFired = false
var bool bosBearFired = false
var float bosLevel    = na

if bullBOS
    recBullBOS   := true
    recBearBOS   := false
    bosTimer     := bar_index
    bosBullFired := true
    bosBearFired := false
    bosLevel     := swHigh
if bearBOS
    recBearBOS   := true
    recBullBOS   := false
    bosTimer     := bar_index
    bosBearFired := true
    bosBullFired := false
    bosLevel     := swLow
if bar_index - bosTimer > bosRst
    recBullBOS := false
    recBearBOS := false

// ═══════════════════════════════════════════
// SWEEPS
// ═══════════════════════════════════════════
bullSweep = low < d1L and close > d1L and (d1L - low) >= atr * 0.3
bearSweep = high > d1H and close < d1H and (high - d1H) >= atr * 0.3

var float lastBullSwL = na
var float lastBearSwH = na
if bullSweep
    lastBullSwL := low
if bearSweep
    lastBearSwH := high

bullSwRej = bullSweep and (close - low)  > atr * 1.0
bearSwRej = bearSweep and (high - close) > atr * 1.0

intBullSw = low  < ta.lowest(low[1],   3)[1] and close > ta.lowest(low[1],   3)[1]
intBearSw = high > ta.highest(high[1], 3)[1] and close < ta.highest(high[1], 3)[1]
intBullOk = intBullSw and aboveE200
intBearOk = intBearSw and belowE200

// ═══════════════════════════════════════════
// ORDER BLOCKS
// ═══════════════════════════════════════════
imp3    = 3
bullImp = close > close[1] and close[1] > close[2] and close[2] > close[3]
bearImp = close < close[1] and close[1] < close[2] and close[2] < close[3]
bullOBC = bullImp and close[imp3] < open[imp3] and math.abs(close[imp3]-open[imp3]) > atr*0.5
bearOBC = bearImp and close[imp3] > open[imp3] and math.abs(close[imp3]-open[imp3]) > atr*0.5

var float bOBT    = na
var float bOBB    = na
var float brOBT   = na
var float brOBB   = na
var bool  bOBAl   = false
var bool  brOBAl  = false
var bool  bOBRsp  = false
var bool  brOBRsp = false
var int   bOBTch  = 0
var int   brOBTch = 0

if bullOBC and aboveE200
    bOBT   := high[imp3]
    bOBB   := low[imp3]
    bOBAl  := true
    bOBRsp := false
    bOBTch := 0
if bearOBC and belowE200
    brOBT  := high[imp3]
    brOBB  := low[imp3]
    brOBAl := true
    brOBRsp:= false
    brOBTch:= 0

if bOBAl and bOBTch == 0 and low <= bOBT and close >= bOBB
    bOBRsp := true
    bOBTch += 1
if brOBAl and brOBTch == 0 and high >= brOBB and close <= brOBT
    brOBRsp := true
    brOBTch += 1
if bOBAl and close < bOBB
    bOBAl  := false
    bOBRsp := false
if brOBAl and close > brOBT
    brOBAl  := false
    brOBRsp := false

// ═══════════════════════════════════════════
// FVG
// ═══════════════════════════════════════════
cBody    = math.abs(close[1] - open[1])
strCdl   = cBody > atr * 1.0
dispCdl  = math.abs(close - open) > atr * 1.5
bFVGSz   = low - high[2]
brFVGSz  = low[2] - high
bFVGraw  = low > high[2] and strCdl and close[1] > open[1] and aboveE200 and bFVGSz  >= atr * fvgMin
brFVGraw = high < low[2] and strCdl and close[1] < open[1] and belowE200 and brFVGSz >= atr * fvgMin

var float bFVGT   = na
var float bFVGB   = na
var float brFVGT  = na
var float brFVGB  = na
var bool  bFVGAl  = false
var bool  brFVGAl = false
var bool  bEZ     = false
var bool  brEZ    = false

if bFVGraw
    bFVGT  := low
    bFVGB  := high[2]
    bFVGAl := true
    bEZ    := false
if brFVGraw
    brFVGT  := low[2]
    brFVGB  := high
    brFVGAl := true
    brEZ    := false

if bFVGAl and low <= bFVGT and close >= bFVGB
    if dispCdl and aboveE200 and weekBull and dayBull and not bEZ
        bEZ := true
if brFVGAl and high >= brFVGB and close <= brFVGT
    if dispCdl and belowE200 and weekBear and dayBear and not brEZ
        brEZ := true

if bFVGAl and close < bFVGB
    bFVGAl := false
    bEZ    := false
if brFVGAl and close > brFVGT
    brFVGAl := false
    brEZ    := false

// ═══════════════════════════════════════════
// RSI
// ═══════════════════════════════════════════
rsi      = ta.rsi(close, 14)
lkb      = 7
pLL      = ta.lowest(low,   lkb) < ta.lowest(low[lkb],   lkb)
pHL      = ta.lowest(low,   lkb) > ta.lowest(low[lkb],   lkb)
pHH      = ta.highest(high, lkb) > ta.highest(high[lkb], lkb)
pLH      = ta.highest(high, lkb) < ta.highest(high[lkb], lkb)
rHL      = ta.lowest(rsi,   lkb) > ta.lowest(rsi[lkb],   lkb)
rLL      = ta.lowest(rsi,   lkb) < ta.lowest(rsi[lkb],   lkb)
rLH      = ta.highest(rsi,  lkb) < ta.highest(rsi[lkb],  lkb)
rHH      = ta.highest(rsi,  lkb) > ta.highest(rsi[lkb],  lkb)
rsiBullX = rsi < 40
rsiBearX = rsi > 60
sBullDiv = pLL and rHL
hBullDiv = pHL and rLL
sBearDiv = pHH and rLH
hBearDiv = pLH and rHH
eqhSwept = not na(ta.pivothigh(high,5,5)) and high > ta.valuewhen(not na(ta.pivothigh(high,5,5)), ta.pivothigh(high,5,5), 1) and close < ta.valuewhen(not na(ta.pivothigh(high,5,5)), ta.pivothigh(high,5,5), 1)
eqlSwept = not na(ta.pivotlow(low,5,5))   and low  < ta.valuewhen(not na(ta.pivotlow(low,5,5)),   ta.pivotlow(low,5,5),   1) and close > ta.valuewhen(not na(ta.pivotlow(low,5,5)),   ta.pivotlow(low,5,5),   1)

// ═══════════════════════════════════════════
// MACD
// ═══════════════════════════════════════════
[macdL, sigL, histL] = ta.macd(close, 12, 26, 9)
macdBull = macdL > sigL
macdBear = macdL < sigL

// ═══════════════════════════════════════════
// TIER 2 STRUCTURE
// ═══════════════════════════════════════════
s2Long  = 0
s2Long  := s2Long  + (aboveE200  ? 1 : 0)
s2Long  := s2Long  + (recBullBOS ? 1 : 0)
s2Long  := s2Long  + (bullSweep  ? 1 : 0)
s2Long  := s2Long  + (inDisc     ? 1 : 0)
s2Long  := s2Long  + (mBullAlign ? 1 : 0)

s2Short = 0
s2Short := s2Short + (belowE200  ? 1 : 0)
s2Short := s2Short + (recBearBOS ? 1 : 0)
s2Short := s2Short + (bearSweep  ? 1 : 0)
s2Short := s2Short + (inPrem     ? 1 : 0)
s2Short := s2Short + (mBearAlign ? 1 : 0)

// ═══════════════════════════════════════════
// TIER 3 TRIGGERS
// ═══════════════════════════════════════════
t1Bull = bEZ
t1Bear = brEZ
t2Bull = bOBRsp
t2Bear = brOBRsp
t3Bull = bullSwRej
t3Bear = bearSwRej

bullPinWick = low - math.min(open, close)
bearPinWick = math.max(open, close) - high
pinBody     = math.abs(close - open)
atKeyBull   = math.abs(close - d1L) <= atr*0.5 or math.abs(close - ema200) <= atr*0.5 or (bOBAl and math.abs(close - bOBB) <= atr*0.5) or (bFVGAl and close >= bFVGB and close <= bFVGT)
atKeyBear   = math.abs(close - d1H) <= atr*0.5 or math.abs(close - ema200) <= atr*0.5 or (brOBAl and math.abs(close - brOBT) <= atr*0.5) or (brFVGAl and close >= brFVGB and close <= brFVGT)
t4Bull = bullPinWick >= 2.5*pinBody and pinBody >= atr*0.1 and aboveE200 and atKeyBull and weekBull and dayBull
t4Bear = bearPinWick >= 2.5*pinBody and pinBody >= atr*0.1 and belowE200 and atKeyBear and weekBear and dayBear
t5Bull = close > d1H and close[1] <= d1H and aboveE200 and weekBull and dayBull
t5Bear = close < d1L and close[1] >= d1L and belowE200 and weekBear and dayBear

var bool bosRetestReady = false
var bool bosRetestDir   = false
var float bosRetestLvl  = na
if bosBullFired
    bosRetestReady := true
    bosRetestDir   := true
    bosRetestLvl   := bosLevel
if bosBearFired
    bosRetestReady := true
    bosRetestDir   := false
    bosRetestLvl   := bosLevel
bosRetestTouch = bosRetestReady and not na(bosRetestLvl) and math.abs(close - bosRetestLvl) <= atr*0.3
t6Bull = bosRetestTouch and bosRetestDir     and close > open and aboveE200
t6Bear = bosRetestTouch and not bosRetestDir and close < open and belowE200
if t6Bull or t6Bear
    bosRetestReady := false

lh1    = high[1] < high[2] and high[2] < high[3] and high[3] < high[4]
hl1    = low[1]  > low[2]  and low[2]  > low[3]  and low[3]  > low[4]
t7Bull = lh1 and high > high[1] and aboveE200 and weekBull and dayBull
t7Bear = hl1 and low  < low[1]  and belowE200 and weekBear and dayBear

anyTrigBull = t1Bull or t2Bull or t3Bull or t4Bull or t5Bull or t6Bull or t7Bull
anyTrigBear = t1Bear or t2Bear or t3Bear or t4Bear or t5Bear or t6Bear or t7Bear

// ═══════════════════════════════════════════
// TIER 4 BONUS
// ═══════════════════════════════════════════
bonL = 0
bonL := bonL + (sBullDiv  ? 2 : hBullDiv ? 1 : 0)
bonL := bonL + (macdBull  ? 1 : 0)
bonL := bonL + (rsiBullX  ? 1 : 0)
bonL := bonL + (trending  ? 1 : 0)
bonL := bonL + (intBullOk ? 1 : 0)
bonL := bonL + (eqlSwept  ? 1 : 0)

bonS = 0
bonS := bonS + (sBearDiv  ? 2 : hBearDiv ? 1 : 0)
bonS := bonS + (macdBear  ? 1 : 0)
bonS := bonS + (rsiBearX  ? 1 : 0)
bonS := bonS + (trending  ? 1 : 0)
bonS := bonS + (intBearOk ? 1 : 0)
bonS := bonS + (eqhSwept  ? 1 : 0)

// ═══════════════════════════════════════════
// SIGNAL CONDITIONS
// ═══════════════════════════════════════════
g1L  = weekBull
g2L  = dayBull
g3L  = htfBullOk
g4L  = activeSess
g5L  = not extBear
allL = g1L and g2L and g3L and g4L and g5L

g1S  = weekBear
g2S  = dayBear
g3S  = htfBearOk
g4S  = activeSess
g5S  = not extBear
allS = g1S and g2S and g3S and g4S and g5S

var int lastBullB = 0
var int lastBearB = 0

rawBull = allL and s2Long  >= minStr2 and anyTrigBull and bonL >= minBonus
rawBear = allS and s2Short >= minStr2 and anyTrigBear and bonS >= minBonus

bullSig = rawBull and (bar_index - lastBullB) > cooldown
bearSig = rawBear and (bar_index - lastBearB) > cooldown

if bullSig
    lastBullB := bar_index
if bearSig
    lastBearB := bar_index

// ═══════════════════════════════════════════
// RISK AND POSITION SIZING
// ═══════════════════════════════════════════
riskAmt  = accountSize * (riskPct / 100)
slBufL   = bonL >= 6 ? atr * 0.2 : atr * 0.5
slBufS   = bonS >= 6 ? atr * 0.2 : atr * 0.5
longSL   = not na(lastBullSwL) ? lastBullSwL - slBufL : close - atr * 1.5
longRisk = math.max(close - longSL, 0.0001)
longTP   = close + longRisk * 3
longSz   = math.max(math.round(riskAmt / longRisk, 4), 0.001)

shortSL  = not na(lastBearSwH) ? lastBearSwH + slBufS : close + atr * 1.5
shortRsk = math.max(shortSL - close, 0.0001)
shortTP  = close - shortRsk * 3
shortSz  = math.max(math.round(riskAmt / shortRsk, 4), 0.001)

// ═══════════════════════════════════════════
// STRATEGY ENTRIES AND EXITS
// ═══════════════════════════════════════════
if bullSig and strategy.position_size == 0
    strategy.entry("BUY", strategy.long, qty=longSz)
    strategy.exit("BUY EXIT", "BUY", stop=longSL, limit=longTP)

if bearSig and strategy.position_size == 0
    strategy.entry("SELL", strategy.short, qty=shortSz)
    strategy.exit("SELL EXIT", "SELL", stop=shortSL, limit=shortTP)

// ═══════════════════════════════════════════
// PLOT SIGNALS ON CHART
// ═══════════════════════════════════════════
plotshape(bullSig, "BUY",  shape.triangleup,   location.belowbar, color.new(color.green,0), size=size.small, text="BUY")
plotshape(bearSig, "SELL", shape.triangledown, location.abovebar, color.new(color.red,0),   size=size.small, text="SELL")

plot(strategy.position_size > 0 ? longSL  : na, "Long SL",  color=color.new(color.red,0),   linewidth=1, style=plot.style_linebr)
plot(strategy.position_size > 0 ? longTP  : na, "Long TP",  color=color.new(color.green,0), linewidth=1, style=plot.style_linebr)
plot(strategy.position_size < 0 ? shortSL : na, "Short SL", color=color.new(color.red,0),   linewidth=1, style=plot.style_linebr)
plot(strategy.position_size < 0 ? shortTP : na, "Short TP", color=color.new(color.green,0), linewidth=1, style=plot.style_linebr)
```
