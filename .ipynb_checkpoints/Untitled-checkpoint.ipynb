If Statement 

```python
if <condition>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>
```

```python
age = 24

if age >=18:
    print("You're an adult!")
elif age>=13:
    print("You're a Teenager")
else:
    print("You're a kid")
```

For Loop

```python
for <variable> in <list>:
    <code>
```

```python
countries
```

```python
for country in countries:
    print(country)
```

```python
for i, country in enumerate(countries):
    print(i)
    print(country)
```

```python
my_data = {'name':'Frank', 'age':26}
my_data
```

```python
for key, value in my_data.items():
    print(key)
    print(value)
```
