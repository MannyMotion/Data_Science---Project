If Statement 

```python
if <condition>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>
```

```python
age = 24

if age >=18:
    print("You're an adult!")
elif age>=13:
    print("You're a Teenager")
else:
    print("You're a kid")
```
